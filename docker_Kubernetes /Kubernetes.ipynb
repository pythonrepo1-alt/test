# ☸️ Introduction

>**Kubernetes (K8s)**: is an open-source container orchestration platform that automates the deployment, scaling, networking, and management of containerized applications.
>
>👉 Kubernetes helps manage containers at scale across multiple machines.
>
>👉 In simple terms:
> * Docker runs containers.
> * Kubernetes manages containers.
>
>👉 **Kubernetes (K8s)**   
>> Why K8s?  
>> K + 8 letters + s = K8s

---
## 🤔 Why Do We Need Kubernetes?
Docker makes it easy to package and run applications inside containers. However, as applications grow, managing hundreds or thousands of containers across multiple servers becomes difficult.

Common challenges:
   ```
   ❌ Container failures   
   ❌ Manual scaling   
   ❌ Load balancing   
   ❌ Service discovery    
   ❌ Application updates    
   ❌ Multi-server management  
   ```
Example: Imagine an application running with 3 containers
  ```  
  Container-1 ✅
  Container-2 ✅
  Container-3 ✅
  ```
If one container crashes:
```
Container-1 ✅
Container-2 ❌
Container-3 ✅
```
Without Kubernetes:
* Manual intervention is required.

With Kubernetes:
* Automatically detects failure.
* Creates a new container.
* Maintains the desired state.


##🎯 Desired State Management
One of the most important concepts in Kubernetes is Desired State.  
You define how the application should look:
   ```
   Run 3 application instances
   Scale automatically
   Recover from failures
   Perform zero-downtime updates
   ```
Kubernetes continuously monitors the current state and takes corrective actions to match the desired state.

Example:  
* Desired State:`3 Running Pods`  
* Current State:`2 Running Pods`  
* Kubernetes Action:`Create 1 New Pod`  

**Note**: Kubernetes follows a Declarative Model.
* Instead of telling Kubernetes HOW to perform a task, you tell it WHAT the final state should be.

* Example:
  * "I want 3 replicas running"
  * Kubernetes figures out how to achieve that state.

# 🏗️ Kubernetes Architecture

>A Kubernetes cluster consists of two major parts:
> 1. Control Plane – the brain of the cluster
> 2. Worker Nodes  – the machines that run your applications
>
> Every cluster needs at least one worker node in order to run Pods.
>
>![](https://kubernetes.io/images/docs/kubernetes-cluster-architecture.svg)




## 🧠 Control Plane
>The Control Plane is the brain of Kubernetes.
>
>It manages the entire cluster and makes decisions about:
> * Scheduling workloads
> * Maintaining desired state
> * Scaling applications
> * Monitoring cluster health

>Example: starting up a new pod when a Deployment's replicas field is unsatisfied.



>### Components of Control Plane:
>```
>+--------------------------+
>|      Control Plane       |
>+--------------------------+
>| API Server               |       API Server               = Front Door
>| etcd                     |       etcd                     = Database
>| Scheduler                |       Scheduler                = Placement Engine
>| Controller Manager       |       Controller Manager       = State Manager
>| Cloud Controller Manager |       Cloud Controller Manager = Cloud Integration Layer
>+------------+-------------+
>```

### 🌐 API Server

> The **API Server** is the central component and primary entry point of a Kubernetes cluster.
>
> Think of it as: **API Server = Front Door of Kubernetes**

All requests to the cluster—whether from users, automation tools, or internal Kubernetes components—pass through the API Server.

### Example Commands

```bash
kubectl get pods                  # List all Pods in the current namespace

kubectl apply -f deployment.yaml  # Create or update resources from a YAML file

kubectl create deployment myapp   # Create a new Deployment named "myapp"

kubectl delete pod pod-1          # Delete the Pod named "pod-1"
```

#### How It Works
```
User
  │
  ▼
kubectl
  │
  │  (API Request)
  ▼
API Server
````
**kubectl** is the command-line tool used to communicate with Kubernetes.
  
When a command is executed:

* kubectl sends a request to the API Server.
* The API Server validates the request.
* The API Server updates or retrieves cluster information.
* Other Kubernetes components act on the request if needed.

---
####Responsibilities
✅ Receives all cluster requests  
✅ Authenticates and authorizes users  
✅ Validates requests  
✅ Updates cluster state  
✅ Communicates with other Control Plane   components  
✅ Acts as the central communication hub of the cluster   


### 🗄️ etcd

> **etcd** is a distributed key-value database used by Kubernetes to store all cluster data.
>
> Think of it as: **etcd = Database of Kubernetes**

>etcd acts as the **single source of truth** for the Kubernetes cluster.

>Every important piece of cluster information is ultimately stored in etcd.



#### What Does etcd Store?

```text
Pods
Nodes
Deployments
Services
Secrets
ConfigMaps
Namespaces
Cluster Configuration
```
#### How It Works
```
  User                   Example:
    │                        kubectl apply -f deployment.yaml
    ▼                                  │
  kubectl                              ▼
    │                              API Server
    ▼                                  │
  API Server                           ▼
    │                                 etcd
    ▼                                  │
  etcd                                 ▼
                              Stores Deployment Configuration

```
When a change is made to the cluster:

* The API Server receives the request.
* The cluster state is stored in etcd.
* Other Kubernetes components read the required information from the API Server.

>**Note**: Users and Worker Nodes do not communicate directly with etcd. All interactions happen through the API Server.

---

####Responsibilities
✅ Stores cluster state  
✅ Stores configuration data  
✅ Maintains desired state information  
✅ Acts as the source of truth for the cluster  
✅ Provides high availability through distributed storage  

---

#### Why Is etcd Important?
Without etcd, Kubernetes would have no memory of:

* Running Pods
* Deployments
* Services
* Cluster configuration
* Node Information

If etcd is lost, the cluster state is lost.

### 📍 Scheduler

> The **Scheduler** is a Control Plane component responsible for deciding which Worker Node should run a newly created Pod.
>
> Think of it as: **Scheduler = Placement Engine of Kubernetes**

> The Scheduler does not create Pods. It simply decides **where a Pod should run**.

The Scheduler continuously watches for Pods that do not yet have a node assigned and selects the most suitable Worker Node for them.

---
#### How It Works

```text
   User
     │
     ▼
kubectl apply -f deployment.yaml
     │
     ▼
API Server
     │
     ▼
Scheduler
     │
     │ Select Best Node
     ▼
Worker Node
```

Example:

```text
Available Nodes

Node-1   CPU Usage: 80%
Node-2   CPU Usage: 30%
Node-3   CPU Usage: 50%
            │
            ▼
      Scheduler Decision

            │
            ▼
    Pod Assigned to Node-2
```

The Scheduler evaluates available Worker Nodes and selects the most suitable one based on resource availability and scheduling rules.

---

#### What Does the Scheduler Consider?

✅ CPU Availability   
✅ Memory Availability  
✅ Resource Requests   
✅ Resource Limits   
✅ Node Labels   
✅ Node Affinity Rules   
✅ Taints and Tolerations   
✅ Pod Distribution Constraints  

---

#### Responsibilities

✅ Selects a Worker Node for newly created Pods  
✅ Distributes workloads across nodes  
✅ Optimizes resource utilization   
✅ Prevents node overloading  
✅ Ensures scheduling rules are followed   

---

#### Why Is the Scheduler Important?

Without the Scheduler:

* New Pods would not know where to run.
* Some nodes could become overloaded.
* Cluster resources would be used inefficiently.

The Scheduler helps balance workloads across the cluster.




### 🎮 Controller Manager

> The **Controller Manager** is a Control Plane component responsible for ensuring that the current state of the cluster matches the desired state.    
>  
> Think of it as: **Controller Manager = State Manager of Kubernetes**  

The Controller Manager continuously monitors the cluster and takes corrective actions whenever the actual state differs from the desired state.

---

#### How It Works

```text

eployment Says:
Run 3 Pods
       │
       ▼
Current State:
Only 2 Pods Running
       │
       ▼
Controller Manager Detects Difference
       │
       ▼
Creates 1 New Pod
       │
       ▼
Current State = Desired State
(3 Pods Running)
```

---

#### Responsibilities

✅ Maintains the desired state of the cluster  
✅ Monitors cluster resources   
✅ Replaces failed Pods   
✅ Handles Node failures   
✅ Ensures Deployments, ReplicaSets, and Nodes are in the expected state  

---

#### Why Is the Controller Manager Important?

Without the Controller Manager:

* Failed Pods would not be replaced.
* Deployments would not maintain the required number of replicas.
* The cluster could drift away from its desired state.

Example:
* Desired Pods = 3
* Current Pods = 2
* Controller Manager Action: Create 1 New Pod

----

#### Scheduler vs Controller Manager

**Controller Manager**
- Ensures the cluster's **desired state** matches the **current state**.
- Creates, replaces, or removes resources when needed.
- Example: If 3 Pods are desired but only 2 are running, it creates 1 more Pod.

**Scheduler**
- Decides **which Worker Node** should run a Pod.
- Assigns Pods to the most suitable node based on available resources and scheduling rules.
- Example: Chooses whether a Pod should run on Node-1, Node-2, or Node-3.

```text
Controller Manager → "Do we have the correct number of Pods?"

Scheduler → "Which node should run these Pods?"
```

### ☁️ Cloud Controller Manager

> The **Cloud Controller Manager (CCM)** is a Control Plane component that enables Kubernetes to interact with cloud provider services.
>
> Think of it as: **Cloud Controller Manager = Cloud Integration Layer**

The Cloud Controller Manager separates cloud-specific functionality from the core Kubernetes components.

It allows Kubernetes to communicate with cloud providers such as:

```text
AWS
Azure
Google Cloud
```

---

#### Why Do We Need It?

Kubernetes itself does not know how to:

* Create a Load Balancer in AWS
* Attach a Disk in Azure
* Manage VM Instances in GCP

The Cloud Controller Manager handles these cloud-specific operations.

---

#### How It Works

```text
Kubernetes Cluster
        │
        ▼
Cloud Controller Manager
        │
        ▼
Cloud Provider APIs
        │
 ┌──────┼──────┐
 ▼      ▼      ▼
VMs  Load Balancers  Storage
```

Example:

```text
User Creates Service
(Type = LoadBalancer)
          │
          ▼
Cloud Controller Manager
          │
          ▼
Cloud Provider API
          │
          ▼
Creates Cloud Load Balancer
```

---

#### Responsibilities

✅ Manages cloud-based Nodes  
✅ Creates external Load Balancers  
✅ Manages cloud networking  
✅ Handles cloud storage integration  
✅ Communicates with cloud provider APIs  

---

#### Common Controllers Inside CCM

```text
Node Controller
Route Controller
Service Controller
```

These controllers interact directly with the cloud provider.

---

#### Why Is It Important?

Without the Cloud Controller Manager:

* Kubernetes would not be able to manage cloud resources automatically.
* LoadBalancer Services would not create cloud load balancers.
* Cloud node information would not be synchronized with Kubernetes.

---

#### Controller Manager vs Cloud Controller Manager

**Controller Manager**
- Manages Kubernetes resources.
- Maintains desired state.
- Handles Pods, ReplicaSets, Nodes, and Deployments.

**Cloud Controller Manager**
- Manages cloud-provider resources.
- Handles Load Balancers, Routes, and Cloud Nodes.
- Communicates with AWS, Azure, or GCP APIs.

```text
Controller Manager
      │
      ▼
Manages Kubernetes Resources

Cloud Controller Manager
      │
      ▼
Manages Cloud Resources
```

---

---
## 👷 Worker Nodes

> Worker Nodes are the machines where Kubernetes runs application workloads.
>
> Think of them as: **Worker Nodes = Execution Layer of Kubernetes**
>
> Every Kubernetes cluster must have at least one Worker Node.
>
> A Worker Node can be:
>
> * A Physical Machine
> * A Virtual Machine (VM)
> * A Cloud Instance (AWS EC2, Azure VM, GCP VM)
>
> The Worker Node is managed by the Control Plane and is responsible for running Pods and Containers.

---

### A Closer Look at a Worker Node

```text
                     Control Plane
                           │
                           ▼
+------------------------------------------------+
|                 Worker Node                    |
|------------------------------------------------|                                               
|  Kubelet     Container Runtime    Kube Proxy   |  🤖 Kubelet = Node Manager
|  +----------+      +----------+                |  📦 Container Runtime = Container Execution Engine
|  |  Pod 1   |      |  Pod 2   |                |  🌐 Kube Proxy = Network Manager
|  |----------|      |----------|                |  📦 Pods = Smallest Deployable Units in Kubernetes
|  |Container |      |Container |                |
|  |Volumes   |      |Volumes   |                |
|  +----------+      +----------+                |                                        
+------------------------------------------------+
```

---
### Control Plane vs Worker Node

```text
Control Plane              Worker Node
---------------------      ---------------------
Makes Decisions            Executes Decisions
Manages the Cluster        Runs Applications
Maintains Desired State    Runs Pods & Containers
Schedules Workloads        Hosts Workloads
```

---

### Key Points

✅ Worker Nodes run application workloads  
✅ Worker Nodes can be physical machines, virtual machines, or cloud instances  
✅ A Worker Node can host multiple Pods   
✅ Pods contain one or more Containers along with their resources   
✅ Kubelet communicates with the Control Plane   
✅ Container Runtime runs Containers inside Pods  
✅ Kube Proxy manages networking and traffic   routing   
✅ Worker Nodes execute instructions received from the Control Plane    
✅ Worker Nodes are managed by the Control Plane

---



### 🤖 Kubelet

> The **Kubelet** is the primary agent that runs on every Worker Node.
>
> Think of it as: **Kubelet = Node Manager**
>
> The Kubelet is responsible for managing everything on a Worker Node.
>
> It continuously communicates with the **Control Plane (API Server)** and ensures that Pods and Containers are running as expected on the node.
>
> The Kubelet receives instructions from the Control Plane and works with the **Container Runtime** to start, stop, and monitor containers.
>
> In simple terms, the Kubelet acts as the **local representative of Kubernetes on each Worker Node**.

---

#### How It Works

```text

                      Example:
Control Plane          Deployment Created
      │                         │
      ▼                         ▼
  API Server           Scheduler Assigns Pod to Worker Node                   
      │                         │
      ▼                         ▼
    Kubelet                   Kubelet
      │                         │
      ▼                         ▼
Container Runtime      Uses Container Runtime to Start Containers
      │                         │
      ▼                         ▼
     Pods                    Pod Running
```

The Kubelet continuously checks whether the Pods are healthy and running as expected.



---

#### Responsibilities

✅ Registers the Worker Node with the cluster  
✅ Receives instructions from the Control Plane  
✅ Works with the Container Runtime to start Pods  
✅ Monitors Pod and Container health  
✅ Reports Node and Pod status back to the Control Plane   
✅ Ensures containers are running as expected  

---

#### Why Is Kubelet Important?

Without the Kubelet:

* Worker Nodes could not communicate with the Control Plane.
* Pods would not be started on the node.
* Kubernetes would not know the health status of Pods and Nodes.

* Example:

  ```text
  Pod Crashes
      │
      ▼
  Kubelet Detects Failure
      │
      ▼
  Reports Status to
  Control Plane
  ```

---

#### Scheduler vs Kubelet

**Scheduler**
- Decides which Worker Node should run a Pod.

**Kubelet**
- Runs and manages Pods on the assigned Worker Node.

```text
Scheduler → "Which node should run this Pod?"

Kubelet → "I'll make sure the Pod runs on that node."
```

---

### Key Takeaway

The Kubelet is the primary agent running on every Worker Node. It communicates with the Control Plane, manages Pods on the node, monitors their health, and ensures they remain in the desired state


### 📦 Container Runtime

> The **Container Runtime** is the software responsible for running containers on a Worker Node.
>
> Think of it as: **Container Runtime = Container Execution Engine**
>
> The Container Runtime is responsible for pulling container images, creating containers, starting them, stopping them, and managing their lifecycle.
>
> It works closely with the **Kubelet**, which instructs it on what containers need to run on the node.

---

#### Common Container Runtimes

```text
containerd
CRI-O
```

> Historically, Docker was used as the container runtime in Kubernetes. Today, Kubernetes typically uses containerd or CRI-O.

---

#### How It Works

```text
                         Example:
Control Plane                Pod Created
      │                          │
      ▼                          ▼
   Kubelet                  Kubelet Receives Instruction
      │                          │
      ▼                          ▼
Container Runtime           Container Runtime Pulls nginx Image
      │                          │
      ▼                          ▼
 Pull Image                 Container Created
      │                          │
      ▼                          ▼
Create Container            Container Starts
      │                          │
      ▼                          ▼
 Start Container              Pod Running
      │                          
      ▼                          
    Pod Running               
```

---

#### Responsibilities

✅ Pulls container images from registries   
✅ Creates containers   
✅ Starts and stops containers  
✅ Manages container lifecycle   
✅ Provides container execution environment   
✅ Works with Kubelet to run Pods   

---

#### Why Is Container Runtime Important?

Without a Container Runtime:

* Containers cannot be created.
* Pods cannot run applications.
* Worker Nodes cannot execute workloads.

---

#### Kubelet vs Container Runtime

**Kubelet**
- Decides what containers should run on the node.
- Monitors and manages Pods.

**Container Runtime**
- Actually runs the containers.
- Handles container lifecycle operations.

```text
Kubelet → "Run this container."

Container Runtime → "I'll create and start it."
```

---

#### Key Takeaway

The Container Runtime is the software that runs containers on a Worker Node. It works with the Kubelet to pull images, create containers, and manage their lifecycle.

### 🌐 Kube Proxy

> The **Kube Proxy** is a network component that runs on every Worker Node.
>
> Think of it as: **Kube Proxy = Network Manager**
>
> Kube Proxy is responsible for managing network communication between Pods, Services, and external traffic.
>
>It ensures that network requests are routed to the correct Pods, even when Pods are created, deleted, or moved between nodes.

---

#### How It Works

```text
User Request
      │
      ▼
   Service
      │
      ▼
 Kube Proxy
      │
      ▼
 Selects a Pod
      │
      ▼
 Pod Receives Request
```

Example:

```text
             Service
                │
                ▼
          Kube Proxy
                │
      ┌─────────┼─────────┐
      ▼         ▼         ▼
    Pod-1     Pod-2     Pod-3

Kube Proxy routes traffic to one
of the available Pods.
```

---

#### Responsibilities

✅ Routes traffic to the correct Pods  
✅ Maintains network rules on the node  
✅ Enables communication between Services and Pods   
✅ Supports load balancing across Pods  
✅ Handles internal and external traffic routing   

---

#### Why Is Kube Proxy Important?

Without Kube Proxy:

* Services would not be able to reach Pods.
* Network traffic could not be routed correctly.
* Applications would become inaccessible.


---

#### Kube Proxy vs Kubelet

**Kubelet**
- Manages Pods and Containers.
- Ensures workloads are running.

**Kube Proxy**
- Manages network communication.
- Ensures traffic reaches the correct Pods.

```text
Kubelet    → "Make sure the Pod is running."

Kube Proxy → "Make sure traffic reaches the Pod."
```

---

#### Key Takeaway

Kube Proxy is the network component running on every Worker Node. It manages traffic routing, Service networking, and load balancing to ensure requests reach the correct Pods.
