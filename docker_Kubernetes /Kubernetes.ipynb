# ☸️ Introduction

>

>
>👉 In simple terms:
> * Docker runs containers.

## 🤔 Why Do We Need Kubernetes?
Docker makes it easy to package and run applications inside containers. However, as applications grow, managing hundreds or thousands of containers across multiple servers becomes difficult.

Common challenges:
   ```
   ❌ Container failures   
   ❌ Manual scaling   
   ❌ Load balancing   
   ❌ Service discovery    
   ❌ Application updates    
   ❌ Multi-server management  
   ```
Example: Imagine an application running with 3 containers
  ```  
  Container-1 ✅
  Container-2 ✅
  Container-3 ✅
  ```
If one container crashes:
```
Container-1 ✅
Container-2 ❌
Container-3 ✅
```
Without Kubernetes:
* Manual intervention is required.

With Kubernetes:
* Automatically detects failure.
* Creates a new container.
* Maintains the desired state.


##🎯 Desired State Management
One of the most important concepts in Kubernetes is Desired State.  
You define how the application should look:
   ```
   Run 3 application instances
   Scale automatically
   Recover from failures
   Perform zero-downtime updates
   ```
Kubernetes continuously monitors the current state and takes corrective actions to match the desired state.

Example:  
* Desired State:`3 Running Pods`  
* Current State:`2 Running Pods`  
* Kubernetes Action:`Create 1 New Pod`  

**Note**: Kubernetes follows a Declarative Model.
* Instead of telling Kubernetes HOW to perform a task, you tell it WHAT the final state should be.

* Example:
  * "I want 3 replicas running"
  * Kubernetes figures out how to achieve that state.

> 1. Control Plane – the brain of the cluster

>The Control Plane is the brain of Kubernetes.
>
>It manages the entire cluster and makes decisions about:
> * Scheduling workloads
> * Maintaining desired state
> * Scaling applications
> * Monitoring cluster health

>Example: starting up a new pod when a Deployment's replicas field is unsatisfied.

### 🌐 API Server

> The **API Server** is the central component and primary entry point of a Kubernetes cluster.
>
> Think of it as: **API Server = Front Door of Kubernetes**

All requests to the cluster—whether from users, automation tools, or internal Kubernetes components—pass through the API Server.

### Example Commands

```bash
kubectl get pods                  # List all Pods in the current namespace

kubectl apply -f deployment.yaml  # Create or update resources from a YAML file

kubectl create deployment myapp   # Create a new Deployment named "myapp"

kubectl delete pod pod-1          # Delete the Pod named "pod-1"
```

#### How It Works
```
User
  ↓
kubectl
  ↓  (API Request)
API Server
````
**kubectl** is the command-line tool used to communicate with Kubernetes.
  
When a command is executed:

* kubectl sends a request to the API Server.
* The API Server validates the request.
* The API Server updates or retrieves cluster information.
* Other Kubernetes components act on the request if needed.

---
####Responsibilities
✅ Receives all cluster requests  
✅ Authenticates and authorizes users  
✅ Validates requests  
✅ Updates cluster state  
✅ Communicates with other Control Plane   components  
✅ Acts as the central communication hub of the cluster

* The API Server receives the request.

####Responsibilities

### 📍 Scheduler

> The **Scheduler** is a Control Plane component responsible for deciding which Worker Node should run a newly created Pod.
>
> Think of it as: **Scheduler = Placement Engine of Kubernetes**

> The Scheduler does not create Pods. It simply decides **where a Pod should run**.

The Scheduler continuously watches for Pods that do not yet have a node assigned and selects the most suitable Worker Node for them.

---
#### How It Works

```text
   User
     ↓
kubectl apply -f deployment.yaml
     ↓
API Server
     ↓
Scheduler
     ↓ Select Best Node
Worker Node
```

Example:

```text
Available Nodes

Node-1   CPU Usage: 80%
Node-2   CPU Usage: 30%
Node-3   CPU Usage: 50%
            ↓
      Scheduler Decision
            ↓
    Pod Assigned to Node-2
```

The Scheduler evaluates available Worker Nodes and selects the most suitable one based on resource availability and scheduling rules.

---

#### What Does the Scheduler Consider?

✅ CPU Availability   
✅ Memory Availability  
✅ Resource Requests   
✅ Resource Limits   
✅ Node Labels   
✅ Node Affinity Rules   
✅ Taints and Tolerations   
✅ Pod Distribution Constraints  

---

#### Responsibilities

✅ Selects a Worker Node for newly created Pods  
✅ Distributes workloads across nodes  
✅ Optimizes resource utilization   
✅ Prevents node overloading  
✅ Ensures scheduling rules are followed   

---

#### Why Is the Scheduler Important?

Without the Scheduler:

* New Pods would not know where to run.
* Some nodes could become overloaded.
* Cluster resources would be used inefficiently.

The Scheduler helps balance workloads across the cluster.




### 🎮 Controller Manager

> The **Controller Manager** is a Control Plane component responsible for ensuring that the current state of the cluster matches the desired state.    
>  
> Think of it as: **Controller Manager = State Manager of Kubernetes**  

The Controller Manager continuously monitors the cluster and takes corrective actions whenever the actual state differs from the desired state.

---

#### How It Works

```text

Deployment Says:
Run 3 Pods
       │
       ▼
Current State:
Only 2 Pods Running
       │
       ▼
Controller Manager Detects Difference
       │
       ▼
Creates 1 New Pod
       │
       ▼
Current State = Desired State
(3 Pods Running)
```

---

#### Responsibilities

✅ Maintains the desired state of the cluster  
✅ Monitors cluster resources   
✅ Replaces failed Pods   
✅ Handles Node failures   
✅ Ensures Deployments, ReplicaSets, and Nodes are in the expected state  

---

#### Why Is the Controller Manager Important?

Without the Controller Manager:

* Failed Pods would not be replaced.
* Deployments would not maintain the required number of replicas.
* The cluster could drift away from its desired state.

Example:
* Desired Pods = 3
* Current Pods = 2
* Controller Manager Action: Create 1 New Pod

----

#### Scheduler vs Controller Manager

**Controller Manager**
- Ensures the cluster's **desired state** matches the **current state**.
- Creates, replaces, or removes resources when needed.
- Example: If 3 Pods are desired but only 2 are running, it creates 1 more Pod.

**Scheduler**
- Decides **which Worker Node** should run a Pod.
- Assigns Pods to the most suitable node based on available resources and scheduling rules.
- Example: Chooses whether a Pod should run on Node-1, Node-2, or Node-3.

```text
Controller Manager → "Do we have the correct number of Pods?"

Scheduler → "Which node should run these Pods?"
```

### 🤖 Kubelet

>
> Think of it as: **Kubelet = Node Manager**
>
> The Kubelet is responsible for managing everything on a Worker Node.
>
> It continuously communicates with the **Control Plane (API Server)** and ensures that Pods and Containers are running as expected on the node.
>
> The Kubelet receives instructions from the Control Plane and works with the **Container Runtime** to start, stop, and monitor containers.
>
> In simple terms, the Kubelet acts as the **local representative of Kubernetes on each Worker Node**.

```text
      │                         │
      ▼                         ▼

      │                         │
      ▼                         ▼
```

The Kubelet continuously checks whether the Pods are healthy and running as expected.

## 📄 Pod YAML

> Kubernetes objects are typically defined using **YAML files**.
>
> A YAML file describes the **desired state** of a Kubernetes object.

In this section, we will learn how a Pod is represented using YAML and understand the structure of a Kubernetes manifest.

---
### 🤔 Why Use YAML?

Kubernetes follows a **Declarative Model**.

Instead of telling Kubernetes **how** to perform a task, you describe the **desired state**, and Kubernetes works to achieve it.

For example, instead of manually starting a container, you define the container configuration in a YAML file and let Kubernetes create and manage it.

YAML files are the standard way to describe Kubernetes objects such as: `Pods`, `Services`, `Deployments`, `ConfigMaps`, `Secrets`

---

### ⚡ Imperative vs Declarative

* **Imperative Approach**: You tell Kubernetes exactly what action to perform.

  ```bash
  kubectl run nginx --image=nginx
  ```

* **Declarative Approach**: You define the desired state in a YAML file.   
  👉 Most Kubernetes manifests follow the same structure
  ```Yaml
  apiVersion:v1     → Which API to use
  kind:  Pod        → What object to create
  metadata:         → Object information
   name: nginx-pod
  spec:             → Desired state
  ```
  **Then apply it**:
  ```
  kubectl apply -f pod.yaml
  ```
  
---

### 🏗️ Example Pod YAML

A Pod YAML file is simply a declarative description of how a Pod should look and behave.

`````
apiVersion: v1               # ✅ Specifies which Kubernetes API version should be used.

kind: Pod                    # ✅ Specifies Kubernetes object type (Pod | Service | Deployment |ConfigMap |Secret)

metadata:                    # ✅ Contains information that identifies the object.
  name: nginx-pod               # Unique name of the Pod
  labels:                       # (Optional) Key-value pairs used to group and identify objects
    app: nginx                     # Marks this object as belonging to the "nginx" application
  annotations:                  # (Optional) Additional metadata for tools and users
    owner: admin                   # Example annotation
  namespace: default            # (Optional) Logical grouping of resource

spec:                        # ✅ Defines the desired state of the object.
  containers:                   # List of containers to run inside the Pod
    - name: nginx                  # Container name
      image: nginx                 # Container image to run
  volumes:                      # (Optional) Shared storage for containers
  ports:                        # (Optional) Network ports exposed by containers
  resources:                    # (Optional) CPU and Memory requirements
`````

▶️ **Running the Pod YAML**

* **Create the Pod**:
  ```bash
  kubectl apply -f pod.yaml
  ```

* **Verify the Pod**:
  ```bash
  kubectl get pods
  
  Example Output:

    NAME        READY   STATUS    RESTARTS   AGE
    nginx-pod   1/1     Running   0          10s
  ```

* **View Pod Details**:
  ```bash
  kubectl describe pod nginx-pod
  ```

* **Delete the Pod**:
  ```bash
  kubectl delete pod nginx-pod
  ```

```text
    ↓
```

## 🏷️ Labels & Selectors
> Labels and Selectors are used to **identify**, **group**, and **connect** Kubernetes objects.
>
> Think of it as:
>```
> Label = Identity Tag
>
> Selector = Search Filter
>```

## 🤔 Why Do We Need Labels & Selectors?

Imagine a cluster with hundreds of Pods. How does Kubernetes know which Pods belong to a particular application, environment, or service, and how does it find only the Pods it needs?

This is where **Labels** and **Selectors** come in.

* 🏷️ A **Label** is a simple **key-value pair** attached to a Kubernetes object that acts as an **Identity Tag**.
* 🎯 A **Selector** is a **Search Filter** used to find objects that match specific Labels.

  ```text
  Label    = Identity Tag

  Selector = Search Filter
  ```

## 📋 Example

Suppose a Pod belongs to the **nginx application**, is running in the **production environment**, and is part of the **frontend tier**.

```yaml
metadata:
  labels:                     # 🏷️ Identity Tags
    app: nginx                # Application Name
    env: production           # Environment
    tier: frontend            # Application Layer

spec:
  selector:                   # 🎯 Search Filter
    app: nginx                # Find all Pods where: app = nginx
```

### ✅ How They Work Together

```text
Label  (app = nginx | env = production | tier = frontend)
      ↓
Selector (app = nginx)
       ↓
Match Found ✅
```

The Selector finds all objects whose Labels match the specified criteria.

---

### ✅ Key Points

* ✅ Labels are key-value pairs
* ✅ Labels identify and group resources
* ✅ Selectors find resources using Labels
* ✅ Services, ReplicaSets, and Deployments use Selectors
* ✅ Labels and Selectors work together to connect Kubernetes objects


### 📌 Common Labels

```text
app  → Application Name

env  → Environment (dev, test, prod)

tier → Application Layer (frontend, backend, database)
```

## 🔁 ReplicaSets

> A **ReplicaSet** ensures that a specified number of Pod replicas are always running.
>
> Think of it as: **ReplicaSet = Pod Count Manager**

---

## 🤔 Why Do We Need ReplicaSets?

Suppose your application should always run with **3 Pods**.

```text
Desired State     Current State
✅ Pod-1          ✅ Pod-1    
✅ Pod-2          ✅ Pod-2
✅ Pod-3          ❌ Pod-3   -> Now one Pod crashes
```

**Without a ReplicaSet**: Manual Intervention Required

**With a ReplicaSet**: if Pod Crashes ReplicaSet Detects Failure and Creates New Pod to match the Desired State. This is called **Self-Healing**.

---

## 🏗️ How Does a ReplicaSet Work?

A ReplicaSet uses **Labels** and **Selectors** to identify and manage Pods.

```text         
                                  Pod-1
ReplicaSet  -->  Selector  -->    Pod-2
                (app=nginx)       Pod-3

```

It continuously checks:( Desired Pods = 3, Current Pods = ? ) and  takes corrective action whenever they don't match.

---

## 📋 Example ReplicaSet

```yaml
apiVersion: apps/v1
kind: ReplicaSet

metadata:
  name: nginx-rs

spec:
  replicas: 3        # Always keep 3 Pods running.

  selector:
    matchLabels:
      app: nginx     # Manage Pods having the label: app = nginx

  template:          # Blueprint for creating new Pods whenever needed.
    metadata:
      labels:         
        app: nginx   # Label assigned to created Pods

    spec:
      containers:       
        - name: nginx   # Container name
          image: nginx  # Container image
```

  ```bash
  kubectl apply -f replicaset.yaml
  ```

* **Verify the ReplicaSet**:
  ```bash
  kubectl get rs
  

  Output:

    NAME       DESIRED   CURRENT   READY
    nginx-rs   3         3         3
  ```

* **View Pods**:

  ```bash
  kubectl get pods
  ```

* **Delete a Pod**:

  ```bash
  kubectl delete pod <pod-name>
  ```

* **Watch Kubernetes recreate it automatically**:

  ```bash
  kubectl get pods -w
  ```

## ✅ Key Points

* ✅ ReplicaSet maintains the desired number of Pods
* ✅ ReplicaSet provides self-healing
* ✅ ReplicaSet uses Labels and Selectors
* ✅ ReplicaSet automatically creates replacement Pods
* ✅ ReplicaSet manages Pod replicas

## 🎯 Key Takeaway

```text
Pod = Runs Application

ReplicaSet = Maintains Pod Count
```

A ReplicaSet continuously ensures that the required number of Pod replicas are running.

 ## 🚀 Deployments

> A **Deployment** is a Kubernetes object that manages ReplicaSets and Pods.
>
> Think of it as: **Deployment = Application Lifecycle Manager**

---

### 🤔 Why Do We Need Deployments & How Do They Work?

* Suppose you're running an application with **3 Pods**.

* A ReplicaSet can ensure that 3 Pods are always running.

* However, modern applications also need:
  * ✅ Application Updates
  * ✅ Rollbacks
  * ✅ Scaling
  * ✅ Release Management



* A ReplicaSet is primarily responsible for maintaining the desired number of Pods and cannot easily handle these application lifecycle operations. This is where a **Deployment** comes in.

* A Deployment sits above a ReplicaSet and manages it.

  ```text
  Deployment          A Deployment:
      │                * Creates ReplicaSets
      ▼                * Manages ReplicaSets
  ReplicaSet           * Updates ReplicaSets
      │                * Rolls back ReplicaSets
      ▼                * Scales application Pods
      Pods
  ```

### ✅ What Happens Behind the Scenes?
* **Create Pods**
  * Suppose we create a Deployment (YAML) and run:
    ```
    kubectl apply -f deployment.yaml
    ```
  * The Deployment creates a ReplicaSet.
  * The ReplicaSet creates the required Pods mention in Yaml file.

* **Scale Applications**
  * Suppose our Deployment was originally created with: `replicas: 3`
  * Now we update the Deployment YAML to: `replicas: 5`   
  * and run:
    ```
    kubectl apply -f deployment.yaml
    ```
  * The Deployment instructs the ReplicaSet to create **2 additional Pods** to match the new desired state

* **Update Applications**
  * Suppose our Deployment was originally running: `image: nginx:v1`
  * Now we update the Deployment YAML to: `image: nginx:v2`
  * and run:
    ```
    kubectl apply -f deployment.yaml
    ```
  * The Deployment performs a Rolling Update and gradually replaces the old Pods with new ones.
  * This allows application updates with minimal downtime.
* **Rollback Applications**
  * Suppose the Deployment was originally running: `image: nginx:v1` and
  * After updating to: `image: nginx:v2` an issue is discovered.
  * Kubernetes can roll back the Deployment to the previous version:
    ```
    kubectl rollout undo deployment nginx-deployment
    ```
  * This allows quick recovery when a new release causes problems.
---

### 📋 Example Deployment

```yaml
apiVersion: apps/v1
kind: Deployment

metadata:
  name: nginx-deployment

spec:
  replicas: 3                  # Always keep 3 Pods running

  selector:
    matchLabels:
      app: nginx               # Manage Pods with label app=nginx

  template:                    #  Blueprint for creating new Pods whenever needed.
    metadata:
      labels:
        app: nginx

    spec:
      containers:
        - name: nginx          # Container name
          image: nginx:latest  # Application version to run.
```


  ```bash
  Example Output:

  NAME               READY   UP-TO-DATE   AVAILABLE
  nginx-deployment   3/3     3            3
  ```

* **View ReplicaSets**
  ```bash
  kubectl get rs
  ```

* **View Pods**
  ```bash
  kubectl get pods
  ```

* **Scale Deployment**
  ```bash

  ```

* **Delete Deployment**
  ```bash
  kubectl delete deployment nginx-deployment
  ```

---

### ✅ Key Points

* ✅ Deployments manage ReplicaSets
* ✅ ReplicaSets manage Pods
* ✅ Deployments support scaling
* ✅ Deployments support rolling updates
* ✅ Deployments support rollbacks
* ✅ Deployments are the preferred way to run applications


## 🏢 Namespaces

> A **Namespace** is a logical partition within a Kubernetes cluster.
>
> Think of it as: **Namespace = Virtual Folder for Kubernetes Resources**

---

### 🤔 Why Do We Need Namespaces?

* Imagine multiple teams (`Development`, `Testing`, `Production`) sharing the same Kubernetes cluster.
* **Without Namespaces**: all resources would exist together in the same cluster, making management difficult.
* **With Namespaces**: Resources are logically separated and easier to manage.

    ```text
    Without Namespaces:        With Namespaces:
    -------------------        -------------------

    Cluster                     Cluster
    │                           │
    ├── dev-nginx               ├── dev
    ├── test-nginx              │   ├── nginx
    ├── prod-nginx              │   └── redis
    ├── dev-redis               │
    ├── prod-redis              ├── test
    └── ...                     │   └── nginx
                                └── prod
                                    ├── nginx
                                    └── redi
    ```
* Namespaces provide logical isolation between resources while still sharing the same Kubernetes cluster.

### 🏗️ How Do Namespaces Work?
* Kubernetes stores resources inside a Namespace.
* A resource created inside one Namespace is logically separated from resources in another Namespace.
* The same resource name can exist in different Namespace

  ```
  Creates a Namespace named: dev    |  creates a Namespace named: Prod
  ----------------------------------|---------------------------------
  apiVersion: v1                    |   apiVersion: v1      
  kind: Namespace                   |   kind: Namespace
                                    |
  metadata:                         |   metadata:
    name: nginx-deployment          |     name: nginx-deployment
    namespace: dev                  |     namespace: prod
  ```
* Even though the resource names(`nginx-deployment`) are the same, Kubernetes treats them as separate resources because they belong to different Namespaces.


### 📌 Common Namespaces

* **default**         → Default namespace for user resources
* **kube-system**     → Kubernetes system components
* **kube-public**     → Publicly readable resources
* **kube-node-lease** → Node heartbeat information


### 📋 Using a Namespac

Suppose we want to create a Deployment inside the `dev` Namespace.
```yaml
apiVersion: apps/v1
kind: Deployment

metadata:
  name: nginx-deployment
  namespace: dev

spec:
  replicas: 1

```

Now the Deployment will be created inside the `dev` Namespace.

  ```bash

  ```

* **View Namespaces**

  ```bash
  kubectl get namespaces
  ```

* **View Resources in a Namespace**

  ```bash

  ```

* **Create Resources in a Namespace**

  ```bash
  kubectl apply -f deployment.yaml -n dev
  ```


### ✅ Key Points

* ✅ Namespaces provide logical isolation
* ✅ Multiple teams can share the same cluster

* ✅ Resources can have the same name in different Namespaces

## 🌐 Services

> A **Service** provides a stable way to access and communicate with Pods.
>
> Think of it as: **Service = Stable Network Endpoint**

---

### 🤔 Why Do We Need Services?

* Pods are temporary in nature.
* Suppose a Deployment creates 3 Pods:
  ```text          
                            If a Pod crashes:       
  Pod-1 (10.0.0.1)            Pod-3 ❌
  Pod-2 (10.0.0.2)            New Pod Created ✅
  Pod-3 (10.0.0.3)            Pod-4 (10.0.0.4)
  ```
* Notice that the IP Address changes.
* If applications directly connect to Pod IPs:

  ```text
  Application
      │
      ▼
  10.0.0.3 ❌     communication breaks.
  ```
* This is where a **Service** comes in.
* A Service provides a **single stable endpoint** for accessing Pods.

---
### 🏗️ How Does a Service Work?

* A Service uses **Labels** and **Selectors** to find Pods.
* Suppose our Pods have:
  ```yaml
  labels:
    app: nginx
  ```

* and our Service has:

  ```yaml
  selector:
    app: nginx
  ```
* Result:
  ```text
                                        Pod-1 ✅
  Service -->  Selector (app=nginx) --> Pod-2 ✅
                                        Pod-3 ✅    
  ```
* The Service automatically routes traffic to matching Pods.
* The Service remains the same even if Pods are created, deleted, or replaced.

---

### 📌 Service Types

* **ClusterIP** (Default) :Accessible only inside the cluster.  
  👉 Used for internal communication between applications.
  ```text
  Application A --> Service --> Application B
  ```

* **NodePort**: Exposes the Service through a port on the Kubernetes Node.  
  👉 Commonly used for learning, testing, and troubleshooting.
  ```text
  Browser --> NodeIP:30080 --> Service --> Pods  
  ```

* **LoadBalancer**: Creates a cloud load balancer.  
  👉 Commonly used in cloud environments such as AWS, Azure, and GCP.
  ```text
  Internet --> Load Balancer --> Service --> Pods  
  ```

---
### 📋 Example Service


```yaml
                                   +------------------------+
apiVersion: v1                     |Flow:                   |
kind: Service                      |  Client                |
                                   |    ↓                   |
metadata:                          |  Service (Port 80)     |
  name: nginx-service              |    ↓                   |
                                   |  Pod (TargetPort 80)   |
spec:                              +------------------------+
  type: ClusterIP
  selector:           # The Service looks for Pods having: app = nginx
    app: nginx

  ports:
    - port: 80        # Port exposed by the Service.
      targetPort: 80  # Port on the Pod/container receiving the traffic.
```

▶️ **Running the Service**

* **Create the Service**

  ```bash
  kubectl apply -f service.yaml
  ```

* **View Services**

  ```bash
  kubectl get services
  ```

* **Describe a Service**

  ```bash
  kubectl describe service nginx-service
  ```

* **Delete a Service**

  ```bash
  kubectl delete service nginx-service
  ```

---

### ✅ Key Points

* ✅ Services provide a stable endpoint for Pods
* ✅ Services use Labels and Selectors
* ✅ Pod IPs can change, Service IPs remain stable
* ✅ Services distribute traffic across Pods
* ✅ Services enable communication inside and outside the cluster

---

### 🎯 Key Takeaway

```text
Pod = Temporary

Service = Stable

Service
    ↓
Finds Pods using Labels & Selectors
    ↓
Routes Traffic
```

# 💾 Storage

> Containers are **ephemeral** by nature.
>
> This means if a container is deleted or recreated, any data stored inside the container may be lost.

>
> Kubernetes provides Storage resources to ensure data persists even when Pods are restarted, replaced, or moved to another node.

To solve different storage requirements, Kubernetes provides several storage-related resources, each serving a specific purpose.

---

## Topics
1. Volumes
2. Persistent Volumes (PV)
3. Persistent Volume Claims (PVC)
4. Storage Classes


## 📦 Volumes

> A **Volume** provides storage that can be shared by one or more containers within a Pod.
>
> Think of it as: **Volume = Pod Storage**
>
> ⚠️ Limitation:
>>By default, Volumes such as `emptyDir` are temporary and exist only for the lifetime of the Pod.
>
>> If the Pod is deleted Data will be lost
---

### 🤔 Why Do We Need Volumes?
* Suppose a Pod contains two containers:
  ```text
  Pod                  
  ├── Container-1
  └── Container-2
  ```
* If Container-1 creates a file: `/report.txt`
* how can Container-2 access the same file?
* This is where a **Volume** comes in.
* A Volume provides a **shared storage location** that **all containers within a Pod** can access.
* **Both containers can read and write** data to the same Volume.
  ```text
  Pod                  
  ├── Container-1
  |     ↓
  ├── Volume
  |     ↑
  └── Container-2
  ```
---
### 📋 Example Volume

```yaml
apiVersion: v1
kind: Pod

metadata:
  name: nginx-pod

spec:
  containers:
    - name: nginx
      image: nginx

      volumeMounts:          # Mounts the Volume inside the container.
        - name: app-storage
          mountPath: /data   # The Volume will be accessible at: /data

  volumes:
    - name: app-storage
      emptyDir: {}           # Creates a temporary Volume for the Pod.   
```

### ✅ Key Points
* ✅ Volumes provide storage for Pods
* ✅ Multiple containers can share the same Volume
* ✅ `volumeMounts` attaches storage to a container
* ✅ `emptyDir` creates temporary Pod storage
* ✅ The Volume is deleted when the Pod is deleted



---
## 💽 Persistent Volumes (PV)

> A **Persistent Volume (PV)** is a storage resource in a Kubernetes cluster that exists independently of Pods.
>
> Think of it as: **PV = Long-Term Cluster Storage**

---

### 🤔 Why Do We Need Persistent Volumes?

* In the previous section, we learned that a Volume such as `emptyDir` is temporary.
* If the Pod is deleted, the Volume is also deleted and all data is lost.
* This is where a **Persistent Volume (PV)** comes in.
* A PV allows data to survive even after a Pod is deleted.

---

### 🏗️ How Does a Persistent Volume Work?

* A **Persistent Volume (PV)** provides storage that exists independently of a Pod.
* It represents a storage resource that Kubernetes makes available to applications.
* The underlying storage can come from a **local disk**, **cloud storage**, or a **network storage system**.
* Once created, the PV becomes part of the Kubernetes cluster's storage pool and can be used by applications to store and retrieve data.


---

### 📋 Example Persistent Volume

```yaml
apiVersion: v1
kind: PersistentVolume

metadata:
  name: app-pv

spec:
  capacity:
    storage: 1Gi          # Storage Size

  accessModes:            # Defines how the volume can be accessed.
    - ReadWriteOnce       # Read/Write by one node

  hostPath:
    path: /data/app       # Storage Location
```

---

### ▶️ **Running the Persistent Volume**

* **Create PV**
  ```bash
  kubectl apply -f pv.yaml
  ```

* **View Persistent Volumes**
  ```bash
  kubectl get pv
  ```

* **Describe Persistent Volume**
  ```bash
  kubectl describe pv app-pv
  ```

* **Delete Persistent Volume**
  ```bash
  kubectl delete pv app-pv
  ```

---

### ✅ Key Points

* ✅ PV provides long-term storage
* ✅ PV exists independently of Pods
* ✅ Data survives Pod restarts and recreation
* ✅ PV is a cluster-level resource
* ✅ Pods typically access PVs through PVCs

---



## 📄 Persistent Volume Claims (PVC)

> A **Persistent Volume Claim (PVC)** is used to request storage from a Persistent Volume (PV).
>
> Think of it as: **PVC = Storage Request**

---
### 🤔 Why Do We Need PVCs?

* In the previous section, we learned that a **Persistent Volume (PV)** provides the actual storage.
* However, applications do not directly use a PV.
* Instead, they request storage through a **Persistent Volume Claim (PVC)**.

* Think of it like:

  ```text
  PV  → Actual Storage

  PVC → Request for Storage
  ```

* For example, an application may need:

  ```text
  1Gi Storage
  Read/Write Access
  ```

* A PVC requests these requirements, and Kubernetes finds a matching PV.

---

### 🏗️ How Does a PVC Work?

* A PVC acts as a bridge between an application and a Persistent Volume.
* The application requests storage through a PVC.
* Kubernetes then connects the PVC to a suitable PV.

  ```text
  Application  -->  PVC (Storage Request) --> PV (Actual Storage)  
  ```

---

### 📋 Example PVC

```yaml
apiVersion: v1
kind: PersistentVolumeClaim  # Creates a PVC resource.

metadata:
  name: app-pvc

spec:
  accessModes:
    - ReadWriteOnce          # Requests read/write access from one node.

  resources:
    requests:
      storage: 1Gi           # Requests 1Gi of storage.
```


### 📋 Using a PVC in a Pod

Once a PVC is created, a Pod can use it as a Volume.

```yaml
spec:
  containers:
    - name: nginx
      image: nginx

      volumeMounts:
        - name: app-storage
          mountPath: /data

  volumes:
    - name: app-storage
      persistentVolumeClaim:
        claimName: app-pvc
```

Here:

  ```text
   Pod --> PVC (app-pvc) --> PV
  ```

The Pod stores its data in the Persistent Volume through the PVC.

---

### ▶️ Running the PVC

* **Create PVC**

  ```bash
  kubectl apply -f pvc.yaml
  ```

* **View PVCs**

  ```bash
  kubectl get pvc
  ```

* **Describe PVC**

  ```bash
  kubectl describe pvc app-pvc
  ```

* **Delete PVC**

  ```bash
  kubectl delete pvc app-pvc
  ```

---

### ✅ Key Points

* ✅ PVC is a request for storage
* ✅ PVC connects applications to Persistent Volumes
* ✅ Kubernetes automatically binds a PVC to a matching PV
* ✅ Pods use PVCs instead of directly using PVs
* ✅ PVC simplifies storage management

---

### 🎯 Key Takeaway

```text
PV  = Actual Storage

PVC = Request for Storage

Pod --> PVC --> PV
```

## 🏷️ Storage Classes

> A **StorageClass** defines how Persistent Volumes are automatically created.
>
> Think of it as: **StorageClass = Storage Template**

---

### 🤔 Why Do We Need Storage Classes?

* So far, we manually created a **Persistent Volume (PV)** and then a **Persistent Volume Claim (PVC)**.

* In large Kubernetes environments, manually creating PVs for every application becomes difficult.

* This is where a **StorageClass** comes in.

* A StorageClass allows Kubernetes to automatically create a Persistent Volume whenever a PVC requests storage.

---

### 🏗️ How Does a StorageClass Work?

Instead of manually creating a PV, Kubernetes dynamically provisions storage based on the StorageClass configuration.
```text
Application --> PVC --> StorageClass --> Automatically Creates PV
```

---

### 📋 Example StorageClass

```yaml
apiVersion: storage.k8s.io/v1
kind: StorageClass

metadata:
  name: fast-storage

provisioner: kubernetes.io/no-provisioner
```

---

### 📋 Using a StorageClass in a PVC

```yaml
apiVersion: v1
kind: PersistentVolumeClaim

metadata:
  name: app-pvc

spec:
  storageClassName: fast-storage    # tells Kubernetes which StorageClass should be used to provide the storage.

  accessModes:
    - ReadWriteOnce

  resources:
    requests:
      storage: 1Gi
```

### ▶️ Running the StorageClass

* **Create StorageClass**

  ```bash
  kubectl apply -f storageclass.yaml
  ```

* **View StorageClasses**

  ```bash
  kubectl get storageclass
  ```

* **Describe StorageClass**

  ```bash
  kubectl describe storageclass fast-storage
  ```

* **Delete StorageClass**

  ```bash
  kubectl delete storageclass fast-storage
  ```

---

### ✅ Key Points

* ✅ StorageClass automates PV creation
* ✅ PVC requests storage from a StorageClass
* ✅ Reduces manual PV management
* ✅ Enables dynamic storage provisioning
* ✅ Common in cloud environments

---

### 🎯 Key Takeaway

```text
PV            = Actual Storage

PVC           = Storage Request

StorageClass  = Storage Template

PVC -- > StorageClass --> PV Created Automatically
```

## 🌐 Services (Recap)

> A **Service** provides a stable way to access and communicate with Pods.
>
> Think of it as: **Service = Stable Network Endpoint**
>
> Pods are temporary and their IP addresses can change.
> A Service provides a stable endpoint that routes traffic to the appropriate Pods.

📋 Example Service
```yaml
apiVersion: v1
kind: Service

metadata:
  name: backend-service

spec:
  selector:
    app: backend        # Route traffic to Pods having the label: app=backend

  ports:
    - port: 80          # Service Port used by applications
      targetPort: 8080  # Port on the Pod/container that receives the traffic

  type: ClusterIP       # Accessible only inside the cluster
```

Request Flow:

```text
Frontend Pod --> backend-service:80 --> Backend Pod:8080
```

## 🚪 Ingress
> An **Ingress** manages external access to applications running inside a Kubernetes cluster.
>
> Think of it as: **Ingress = Traffic Manager**

---

### 🤔 Why Do We Need Ingress?

* We previously learned that a **Service** can expose an application.

* Suppose we have multiple applications:
  ```text
  frontend-service
  backend-service
  api-service
  ```

* Without Ingress, each Service may need its own external endpoint.
  ```text
  http://IP1
  http://IP2
  http://IP3
  ```

* Managing multiple endpoints becomes difficult.

* This is where **Ingress** comes in.

* Ingress provides a single entry point and routes traffic to the appropriate Service.

  ```text
  User
   │
   ▼
  Ingress
   ├── /frontend ──► frontend-service
   ├── /backend ───► backend-service
   └── /api ───────► api-service
  ```
---
### 🏗️ How Does Ingress Work?

* Ingress sits in front of Services.
* It receives incoming requests and forwards them to the correct Service based on rules such as:
  * URL Path
  * Hostname (Domain)

  ```text
  User --> Ingress --> Service --> Pods
  ```

---

### 📋 Example Ingress
```yaml
apiVersion: networking.k8s.io/v1
kind: Ingress

metadata:
  name: app-ingress

spec:
  rules:
    - host: myapp.com  # User accesses: https://myapp.com will use this Ingress rule.
      http:
        paths:
          - path: /    # Match requests for the root path (/)
                       # and all sub-paths because Prefix is used.
            pathType: Prefix

            backend:
              service:
                name: frontend-service  # Routes traffic to frontend-service
                port:
                  number: 80    # Send traffic to Service Port 80.
          - path: ....
```

Request Flow:
```
User
  ↓
https://myapp.com
  ↓
Ingress
  ↓
frontend-service:80
  ↓
Pods
```


### 🌐 Path-Based Routing

A single Ingress can route traffic to multiple Services.

* https://myapp.com/frontend  → frontend-service
* https://myapp.com/backend   → backend-service
* https://myapp.com/api       → api-service


Example:

```yaml
apiVersion: networking.k8s.io/v1
kind: Ingress

metadata:
  name: app-ingress

spec:
  rules:
    - host: myapp.com

      http:
        paths:

          - path: /frontend       # Requests to /frontend go to frontend-service.
            pathType: Prefix
            backend:
              service:
                name: frontend-service
                port:
                  number: 80

          - path: /backend        # Requests to /backend go to backend-service.
            pathType: Prefix
            backend:
              service:
                name: backend-service
                port:
                  number: 80

          - path: /api            # Requests to /api go to api-service.
            pathType: Prefix
            backend:
              service:
                name: api-service
                port:
                  number: 80
```

Request Examples:

```text
https://myapp.com/frontend      https://myapp.com/backend      https://myapp.com/api
               ↓                          ↓                           ↓
      frontend-service             backend-service                api-service
               ↓                          ↓                           ↓
             Pods                        Pods                        Pods
```


---

### ▶️ Running the Ingress

* **Create Ingress**

  ```bash
  kubectl apply -f ingress.yaml
  ```

* **View Ingresses**

  ```bash
  kubectl get ingress
  ```

* **Describe Ingress**

  ```bash
  kubectl describe ingress app-ingress
  ```

* **Delete Ingress**

  ```bash
  kubectl delete ingress app-ingress
  ```

### ✅ Key Points

* ✅ Ingress provides a single entry point for applications
* ✅ Routes traffic to Services
* ✅ Supports path-based routing

## 🌍 DNS

> DNS (Domain Name System) enables applications to find and communicate with Services using names instead of IP addresses.
>
> Think of it as: **DNS = Service Discovery**

---

### 🤔 Why Do We Need DNS?

* Suppose a Service has the following IP address:

  ```text
  10.96.15.20
  ```

* Another application needs to communicate with this Service.

* Remembering IP addresses is difficult, and Service IPs may change over time.

* Instead of using IP addresses, Kubernetes allows applications to communicate using Service names.
  ```text
  ❌ http://10.96.15.20

  ✅ http://backend-service
  ```

* This is where **DNS** comes in.

---

### 🏗️ How Does DNS Work?

* Every Service created in Kubernetes automatically gets a DNS name.

* Applications can use the Service name instead of its IP address.

  ```text
  Frontend Pod
        ↓
  backend-service
        ↓
  Service
        ↓
  Backend Pods
  ```

* Kubernetes DNS automatically resolves the Service name (`backend-service`) to the corresponding Service IP.

---

### 📋 Example Service

```yaml
apiVersion: v1
kind: Service

metadata:
  name: backend-service

spec:
  selector:
    app: backend

  ports:
    - port: 80
      targetPort: 8080
```

* Once this Service is created, Kubernetes automatically creates a DNS entry.

* `backend-service` becomes a valid DNS name inside the cluster

* Applications can access the Service using: http://backend-service instead of: `http://<service-ip>`

---

### 🌐 DNS Naming Convention

* Basic Service Name:

  ```text
  http://backend-service
  ```

* Fully Qualified Domain Name (FQDN):

  ```text
  http://backend-service.default.svc.cluster.local
  ```

  Where:
  * **backend-service** → Service Name
  * **default**         → Namespace
  * **svc**             → Service
  * **cluster.local**   → Cluster Domain

---
### ▶️ Running DNS Verification

* Create the Service

  ```bash
  kubectl apply -f service.yaml
  ```

* Verify DNS from another Pod

  ```bash
  kubectl exec -it <pod-name> -- nslookup backend-service
  ```

* Example Output

  ```text
  Name: backend-service
  Address: 10.96.15.20
  ```

---

### ✅ Key Points

* ✅ DNS provides Service discovery
* ✅ Services automatically receive DNS names
* ✅ Applications use names instead of IP addresses
* ✅ DNS simplifies Pod-to-Service communication
* ✅ Kubernetes manages DNS automatically

---

### 🎯 Key Takeaway

```text
Frontend Pod
      ↓
backend-service
      ↓
DNS Lookup
      ↓
Service
      ↓
Backend Pods

DNS = Service Discovery
```

## 🔒 Network Policies (Overview)

> A **Network Policy** controls which Pods can communicate with each other.
>
> Think of it as: **NetworkPolicy = Firewall for Pods**
>
> By default, Pods can communicate with other Pods in the cluster.
>
> Network Policies allow only authorized traffic.

### 📋 Example Network Policy

```yaml
apiVersion: networking.k8s.io/v1
kind: NetworkPolicy

metadata:
  name: backend-policy

spec:
  podSelector:
    matchLabels:
      app: backend      # Apply this policy to Pods having the label: app=backend

  policyTypes:
    - Ingress           # Control incoming traffic

  ingress:
    - from:
        - podSelector:
            matchLabels:
              app: frontend   # Allow traffic only from Pods having app=frontend
```

Traffic Flow:

```text
Frontend Pod
      ↓
Backend Pod

Random Pod ❌
      ↓
Backend Pod
```

## 👥 RBAC (Role-Based Access Control)

> **RBAC** controls what users and applications are allowed to do in a Kubernetes cluster.
>
> Think of it as: **RBAC = Permission Management**

---

### 🤔 Why Do We Need RBAC?

* Suppose a Kubernetes cluster is shared by multiple users: **`Developer`**, **`Tester`**,
**`Administrator`**

* Without RBAC: Everyone can do everything (Open access) ❌
  * Examples: **`Delete Deployments`** , **`View Secrets`**, **`Modify Services`**, **`Create Namespaces`**

* This is risky and can lead to accidental or unauthorized changes.

* This is where **RBAC** comes in. RBAC allows us to define:
  * Who can access?
  * What actions can they perform?
  * Which resources can they access?

---

### 🏗️ How Does RBAC Work?

RBAC is built using four components:

* 👤 **User / ServiceAccount**: The identity requesting access.  
  Examples:
  ```text
  john
  alice
  fastapi-service-account
  ```
* 📋 **Role** : Defines permissions within a specific Namespace.  
  Example:
  ```text
  Can View Pods
  Can View Services
  Cannot Delete Deployments
  ```
* 🌍 **ClusterRole** : Defines permissions across the entire cluster.
  Example:
  ```text
  Can View All Nodes
  Can View All Namespaces
  ```

* 🔗 **RoleBinding / ClusterRoleBinding**: Connects a User or ServiceAccount to a Role or ClusterRole.
  ```text
  Developer
      ↓
  Read Only Role
  ```
Together, these components determine who can access which resources and what actions they can perform.
```text
User / ServiceAccount
        ↓
RoleBinding / ClusterRoleBinding
        ↓
Role / ClusterRole
        ↓
    Permissions
```

### 📋 Example Role

```yaml
apiVersion: rbac.authorization.k8s.io/v1
kind: Role

metadata:
  name: pod-reader
  namespace: dev      # Applies only to the dev namespace

rules:
  - apiGroups: [""]
    resources: ["pods"]    # Resource type
    verbs: ["get", "list"] # Allowed actions
```

---

### 📋 Example RoleBinding

```yaml
apiVersion: rbac.authorization.k8s.io/v1
kind: RoleBinding

metadata:
  name: pod-reader-binding
  namespace: dev

subjects:
  - kind: User
    name: john         # User receiving permissions

roleRef:
  kind: Role
  name: pod-reader     # Role being assigned
  apiGroup: rbac.authorization.k8s.io
```

This grants:

```text
User: john
      ↓
RoleBinding: pod-reader-binding
      ↓
Role: pod-reader
      ↓
Can View Pods in dev Namespace
```

---
### ▶️ Running RBAC

* **Create Role**

  ```bash
  kubectl apply -f role.yaml
  ```

* **Create RoleBinding**

  ```bash
  kubectl apply -f rolebinding.yaml
  ```

* **View Roles**

  ```bash
  kubectl get roles
  ```

* **View RoleBindings**

  ```bash
  kubectl get rolebindings
  ```

---
### ✅ Key Points

* ✅ RBAC manages permissions
* ✅ Role defines permissions
* ✅ RoleBinding assigns permissions
* ✅ Roles are Namespace-specific
* ✅ ClusterRoles are cluster-wide
* ✅ Follows the principle of least privilege

---

### 🎯 Key Takeaway

```text

 ↓
User / ServiceAccount

Gets Permissions Through
 ↓
RoleBinding / ClusterRoleBinding

Which References
 ↓
Role / ClusterRole

RBAC = Permission Management
```

## 🪪 Service Accounts

> A **ServiceAccount** provides an identity for Pods running inside a Kubernetes cluster.
>
> Think of it as: **ServiceAccount = Pod Identity**

---

### 🤔 Why Do We Need Service Accounts?

* In the previous section, we learned that **RBAC** controls permissions.

* The next question is: **``Who uses those permissions?``**

  Examples:


* This is where a **ServiceAccount** comes in.

---

### 🏗️ How Do Service Accounts Work?

  ```text
                          For example:
  ```

## 🔒 Network Policies

> A **Network Policy** controls how Pods communicate with each other and with external resources.
>
> Think of it as: **NetworkPolicy = Firewall for Pods**

---

### 🤔 Why Do We Need Network Policies?

* In the Networking module, we briefly introduced Network Policies as a way to secure communications.

* By default, Pods can communicate with other Pods in the cluster.

* In many situations, this is not desirable.

* For example, we may want:

  ```text
  Frontend Pod ✅ → Backend Pod

  Backend Pod ✅ → Database Pod

  Random Pod ❌ → Database Pod
  ```

* This is where **Network Policies** come in.

* Network Policies allow us to define:

  * Which Pods can communicate?
  * Which traffic is allowed?
  * Which traffic is blocked?

---

### 🏗️ How Do Network Policies Work?

* Network Policies use **Labels** and **Selectors** to identify Pods.
* They then define rules that allow or deny traffic.

  ```text
  Frontend Pod  → Backend Pod ✅

  Random Pod  → Backend Pod ❌
  ```

* Only the traffic specified in the policy is allowed.

---

### 📋 Example Network Policy

```yaml
apiVersion: networking.k8s.io/v1
kind: NetworkPolicy

metadata:
  name: backend-policy

spec:
  podSelector:
    matchLabels:
      app: backend      # Apply this policy to Pods
                        # having the label: app=backend

  policyTypes:
    - Ingress           # Control incoming traffic

  ingress:
    - from:
        - podSelector:
            matchLabels:
              app: frontend   # Allow traffic only from
                              # Pods having app=frontend

      ports:
        - protocol: TCP
          port: 80            # Allow traffic on Port 80
```

---

```text
  Random Pod  → Backend Pod ❌
```

## ✅ Jobs

> A **Job** is used to run a task until it successfully completes.
>
> Think of it as: **Job = Run Once and Finish**

---

### 🤔 Why Do We Need Jobs?

* Deployments are designed for long-running applications.

  ```text
  Web Applications
  APIs
  Microservices
  ```

* However, some tasks only need to run once.

  Examples:

  ```text
  Database Backup

  Data Processing

  Report Generation

  File Migration
  ```

* Running these tasks as a Deployment is not appropriate because Deployments try to keep Pods running continuously.

* This is where a **Job** comes in.

* A Job creates a Pod, runs the task, and stops once the task completes successfully.

---

### 🏗️ How Does a Job Work?

```text
Job --> Pod --> Task Runs --> Task Completes
                                 ↓
                              Pod Exits
                                 ↓
                              Job Complete ✅
```

If the Pod fails before finishing:

```text
Job --> Pod Fails ❌
          ↓
        New Pod Created
          ↓
        Task Retries
```

The Job continues until the task completes successfully.

---

### 📋 Example Job

```yaml
apiVersion: batch/v1
kind: Job

metadata:
  name: hello-job

spec:
  template:
    spec:
      containers:
        - name: hello

          image: busybox

          command:
            - echo
            - "Hello Kubernetes Job"  # Task to execute

      restartPolicy: Never            # Do not restart container
```

---

### ✅ What Happens?

```text
Job Created
      ↓
Pod Started
      ↓
echo "Hello Kubernetes Job"
      ↓
Task Completed
      ↓
Job Succeeded ✅
```

## ⏰ CronJobs

> A **CronJob** creates Jobs on a schedule.
>
> Think of it as: **CronJob = Scheduled Job**

---

### 🤔 Why Do We Need CronJobs?

* In the previous section, we learned that a **Job** runs a task once and then completes.

* However, some tasks need to run repeatedly.

  Examples:

  ```text
  Daily Database Backup

  Hourly Report Generation

  Weekly Cleanup Task

  Nightly Data Processing
  ```

* Manually creating Jobs every time is not practical.

* This is where a **CronJob** comes in.

* A CronJob automatically creates and runs Jobs according to a schedule.

---

### 🏗️ How Does a CronJob Work?

* A CronJob contains a schedule.
* At the scheduled time, Kubernetes creates a Job.
* The Job runs and completes.

  ```text
  CronJob
      ↓
  Scheduled Time
      ↓
  Job Created
      ↓
  Pod Runs
      ↓
  Task Complete ✅
  ```

---

### 📋 Example CronJob


```yaml

apiVersion: batch/v1
kind: CronJob

metadata:
  name: backup-job

spec:
  schedule: "0 0 * * *"       # Run every day at m*dnight

  jobTemplate:
    spec:
 *    template:
        spec:
      *   containers:
            - name:*backup
              image: busybo*

              command:
         *      - echo
                - "Running Backup"   # Task executed by The CronJob

          restartPolic*: Never       # Do not restart after completion
```

What happens?

```text
Midnight
    ↓
CronJob Trigg*red
    ↓
Job Created
    ↓
Pod Cr*ated
    ↓
"Running Backup"
    ↓
task Complete ✅
```


---

### ✅ Understanding the Schedule

```text
* * * * *
│ │ │ │ │
│ │ │ │ └── Day of Week
│ │ │ └──── Month
│ │ └────── Day of Month
│ └──────── Hour
└────────── Minute
```

Examples:

```text
0 * * * *      → Every hour

0 0 * * *      → Every day at midnight

0 0 * * 0      → Every Sunday

*/5 * * * *    → Every 5 minutes
```

  ```text
  Web Applications

  APIs

  Microservices
  ```

* Pods created by a Deployment are interchangeable.

  ```text
  Pod-1 = Pod-2 = Pod-3
  ```

* However, some applications need:

  ```text
  Stable Pod Names

  Persistent Storage

  Ordered Deployment

  Ordered Scaling
  ```

* Examples:

  ```text
  MySQL

  PostgreSQL

  MongoDB

  Kafka
  ```

* This is where a **StatefulSet** comes in.

---

### 🏗️ How Does a StatefulSet Work?

```text
Deployment
    ↓
Pod Names Change

app-x7d8f
app-a9k2m
app-b4p6q
```

StatefulSets provide stable identities:

```text

mysql-1
mysql-2
```

Even if a Pod is recreated:

```text
mysql-1 Deleted
      ↓

```

The Pod keeps the same identity.

---

### 📋 Example StatefulSet

```yaml

  replicas: 3

  selector:
    matchLabels:
      app: mysql

  template:
    metadata:
      labels:
        app: mysql

    spec:
      containers:
        - name: mysql
          image: mysql:8
```

---

### ✅ What Happens?

```text
StatefulSet
      ↓

mysql-0
mysql-1
mysql-2
```

Each Pod gets:

```text
Unique Identity

Stable Network Name

Persistent Storage
```

---

### 🔄 Ordered Operations

StatefulSets create Pods in order:

```text
mysql-0
   ↓
mysql-1
   ↓
mysql-2
```

Scaling down also happens in reverse order:

```text
mysql-2
   ↓
mysql-1
   ↓
mysql-0
```

## 🧮 Resource Requests

> A **Resource Request** specifies the minimum CPU and Memory required by a Pod.
>
> Think of it as: **ResourceRequest = Guaranteed Resources**

---

### 🤔 Why Do We Need Resource Requests?

* Multiple Pods share the same Node resources.

* Kubernetes needs to know how much CPU and Memory a Pod requires before scheduling it.

* Resource Requests help Kubernetes place Pods on Nodes with sufficient resources.

---

### 🏗️ How Do Resource Requests Work?

```text
Pod
 ↓
Requests Resources
 ↓
Scheduler
 ↓
Suitable Node
```

---

### 📋 Example Resource Requests

```yaml
apiVersion: v1
kind: Pod

metadata:
  name: app-pod

spec:
  containers:
    - name: app
      image: nginx

      resources:
        requests:
          cpu: "250m"      # Request 0.25 CPU
          memory: "256Mi"  # Request 256 MB Memory
```

---

### ✅ Key Points

* ✅ Defines minimum required resources
* ✅ Used by the Kubernetes Scheduler
* ✅ Helps prevent resource contention
* ✅ CPU is measured in millicores (m)
* ✅ Memory is measured in Mi/Gi

---

### 🎯 Key Takeaway

```text
Resource Requests
        ↓
Minimum CPU & Memory
        ↓
Used for Scheduling

ResourceRequest = Guaranteed Resources
```

#### 🚀 **Phase 1**: Build the FastAPI Application

>In this phase, we will create a simple FastAPI application and prepare it for containerization.
>
>---
>
>##### 📋 **Step 1.1**: Create FastAPI Application
>> Create the application's main entry point.
>>
>>**app/main.py**
>>```python
>>from fastapi import FastAPI
>>from fastapi.responses import JSONResponse
>>import os
>>
>>app = FastAPI()
>>
>>@app.get("/")
>>def root():
>>    return {
>>        "application": "FastAPI on EKS",
>>        "environment": os.getenv("APP_ENV"),
>>        "version": os.getenv("APP_VERSION")
>>    }
>>
>>@app.get("/health")
>>def health():
>>    return {"status": "healthy"}
>>
>>@app.get("/config")
>>def config():
>>    return {
>>        "env": os.getenv("APP_ENV"),
>>        "version": os.getenv("APP_VERSION")
>>    }
>>```
>
>---
>
>##### 📋 **Step 1.2**: Define Python Dependencies
>> Create a file containing the required Python packages.
>>
>>**app/requirements.txt**
>>```
>>fastapi
>>uvicorn
>>```
>
>---
>
>##### 📋 **Step 1.3**: Create Dockerfile
>>Create a Dockerfile to package the application into a container image.
>>
>>**app/Dockerfile**
>>```text
>>FROM python:3.12-slim
>>
>>WORKDIR /app
>>
>>COPY requirements.txt .
>>
>>RUN pip install -r requirements.txt
>>
>>COPY . .
>>
>>CMD ["uvicorn","main:app","--host","0.0.0.0","--port","8000"]
>>```

>>
>>    spec:
>>
>>      serviceAccountName: fastapi-sa
>>
>>      containers:
>>
>>        - name: fastapi
>>
>>          image: 123456789012.dkr.ecr.us-east-1.amazonaws.com/fastapi-app:v1
>>
>>          ports:
>>            - containerPort: 8000
>>
>>          envFrom:
>>            - configMapRef:
>>                name: fastapi-config
>>
>>            - secretRef:
>>                name: fastapi-secret
>>
>>          volumeMounts:
>>            - mountPath: /data
>>              name: storage
>>
>>          resources:
>>            requests:
>>              cpu: 250m
>>              memory: 256Mi
>>
>>            limits:
>>              cpu: 500m
>>              memory: 512Mi
>>
>>          readinessProbe:
>>            httpGet:
>>              path: /health
>>              port: 8000
>>
>>            initialDelaySeconds: 5
>>
>>          livenessProbe:
>>            httpGet:
>>              path: /health
>>              port: 8000
