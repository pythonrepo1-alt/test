# 🧠 Agentic AI

Agentic AI is a way of building AI systems that don’t just give answers, but can **think, make decisions, and take actions step by step to solve a problem**.

Instead of only generating text, these systems act like **agents**. They can use tools (APIs, functions), remember information, and work towards completing a goal.

```
  LLM (🧠 Brain) → Agent (🤖 Thinks + Acts) → Agentic AI (🏗 System of Agents)
```
---

## 👉 What is an Agent?

An agent is a system that can **think, decide, and act** to complete a task.

It takes input, reasons using an LLM, decides what to do, and produces an output.

👉 **Agent = Think + Act**

---

## 🔧 Agent Components & Flow

An agent is built using key components and works in a step-by-step flow.

- **Components** = what the agent has  
- **Flow** = how the agent works  

Together → they define how an agent behaves  


```
     Agent Flow                             Agent Components
    ------------                           ------------------

    User Input                              ┌──────────────┐
         ↓                                  │     LLM      │  🧠 (thinking)
    Think (LLM reasoning)                   └──────┬───────┘
         ↓                                         │
    Decide Action                        ┌─────────┼──────────┐
         ↓                               ↓         ↓          ↓
    Act (use tool or respond)          Tools     Memory     Planner
         ↓                              🛠         🧾         🧩
      Output
```

- **LLM** → Brain (reasoning, decision making)  
- **Tools** → Actions (APIs, functions, search, calculator)  
- **Memory** → Stores past information  
- **Planner** → Breaks tasks into steps  

---

## 🔁 Agent Loop (ReAct Pattern)

Instead of solving everything in one step, agents **work iteratively (step by step)**.

Agents work in a loop:
  ```
  Thought → Action → Observation → Repeat
  ```
1. **Thought** → Decide what to do  
2. **Action** → Use tool or respond  
3. **Observation** → Get result  
4. Repeat until task is done  


---
## 🌐 Other Important Terms

### 🤖 Single Agent

 - One system does all tasks
  ```
     [ One Agent ]
          ↓
    Handles everything
  ```
---

### 👥 Multi-Agent System

- Multiple agents working together  
- Each has a specific role  
- 👉 Think: **Team of agents**

  ```
  [ Research Agent ] → [ Writer Agent ] → [ Reviewer Agent ]
  ```

---

### 🔥 Deep Agent

- Handles complex, multi-step tasks  
- Uses planning, memory, and tools  
- 👉 Think: **Advanced agent**

  ```
  Goal → Plan → Step1 → Step2 → Step3 → Output
  ↑
  Memory + Tools
  ```

---

### ⚡ Autonomous Agent

- Works with minimal human input  
- 👉 Think: **Self-operating agent**
  ```
    Start Task → Runs on its own → Final Result
  ```
---

## 📌 What This Means

- Agents combine **thinking + doing**

- They can handle **multi-step tasks**

## 🎯 Final Summary

# ⚙️ Simple Rule-Based Agent (No LLM)

This is a basic implementation of an agent that can:
- Think (decide what to do)
- Take action (use a tool or respond)
- Store memory  

⚠️ **Note:**  
- This is a **deterministic agent** (no real LLM involved).  
- The "thinking" is rule-based (hardcoded logic), not actual AI reasoning.

---


````
class SimpleAgent:
    def __init__(self):
        self.memory = []

    def think(self, user_input):
        """Decide what action to take (Deterministic reasoning: rule-based, no LLM)"""
        user_input = user_input.lower()

        if "add" in user_input:
            return "use_calculator"
        return "respond"

    def act(self, action, user_input):
        """Perform action based on decision (Simulated tools/actions)"""
        if action == "use_calculator":
            numbers = [int(x) for x in user_input.split() if x.isdigit()]
            return sum(numbers)
        return f"I understand: {user_input}"

    def run(self, user_input):
        print(f"\nUser Input: {user_input}")

        # Step 1: Think
        action = self.think(user_input)
        print(f"Thought: I should {action}")

        # Step 2: Act
        result = self.act(action, user_input)
        print(f"Action Result: {result}")

        # Step 3: Store memory
        self.memory.append((user_input, result))

        return result


# Create agent
agent = SimpleAgent()

# Run examples
agent.run("hello")
agent.run("add 10 and 20")
````

# 🤖 LLM-Based Agent

An LLM-based agent uses a language model to **think and decide what to do**, instead of relying on fixed rules.

- 🧠 Decision making is done by the LLM  
- The LLM can return structured outputs (e.g., JSON) describing which action to take and with what inputs
- 🛠 Actions are executed using tools (functions or APIs)  
- 🔁 Behavior is dynamic (not predefined by rules)


---

### 🔹 How It Works
```
User Input → LLM (Reasoning) → Decide Action → Execute Action → Output
```
### 🔹 Core Components

- 🧠 LLM → decides what action to take  
- 🛠 Tools → perform the action  
- 🧾 Memory → stores past interactions  
- 🔁 Loop → Think → Act → Output  

___
#### ✅ 1. Basic LLM Agent (Simple)

- LLM decides what action to take  
- Action is executed manually in code  
- No structured tool definition  

👉 Example:

- **User Input**  : "add 10 and 20"

- **LLM Output** : use_calculator

- **System**: Executes calculator function

---


✅ 1. Tool Layer (Actions)

````
# ✅ Tool: Calculator
def use_calculator(numbers):
    return sum(numbers)
````

✅ 2. LLM Layer (Thinking / Decision)

````
from openai import OpenAI
import os

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def think(user_input):
    """LLM decides what action to take"""

    prompt = f"""
You are an AI agent.
Decide what action to take.

Available actions:
- use_calculator
- respond

User input: {user_input}

Only return the action name.
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content.strip()
````

✅ 3. Orchestrator (Main Agent Flow)

````
def run_agent(user_input):
    print("\nUser Input:", user_input)

    # Step 1: Think
    action = think(user_input)
    print("LLM Decision:", action)

    # Step 2: Act (manual execution)
    if "use_calculator" in action:
        numbers = [int(x) for x in user_input.split() if x.isdigit()]
        result = use_calculator(numbers)
    else:
        result = f"I understand: {user_input}"

    print("Final Output:", result)
    return result
````

✅ Run Example

````
run_agent("add 10 and 20")
run_agent("hello")
````

### ✅ 2. LLM Agent with Tools (Structured)

- Tools are explicitly defined using schemas  
- LLM returns structured tool calls  
- System automatically executes tools based on structured output

👉 Example:

* **Tool Definition**:
  - use_calculator(numbers: list)

* **User Input**: "add 10 and 20"

* **LLM Output (tool call)**:
  ```
    {
      "name": "use_calculator",
      "arguments": {
        "numbers": [10, 20]
      }
    }
  ```
* **System**:
  - Executes tool using arguments
  - Returns result to LLM
``

💡 This is the pattern used in real-world agent systems

🛠 Tools definition

````
# ✅ Define tools (what actions the agent can perform)
tools = [
    {
        "type": "function",
        "function": {
            "name": "use_calculator",
            "description": "Add a list of numbers",
            "parameters": {
                "type": "object",
                "properties": {
                    "numbers": {
                        "type": "array",
                        "items": {"type": "integer"}
                    }
                },
                "required": ["numbers"]
            }
        }
    }
]


# ✅ Tool implementation
def use_calculator(numbers):
    return sum(numbers)
````

 ⚙️ Tool execution

````
import json

# ✅ Handle tool calls → executes tools based on LLM decision
def handle_tool_calls(tool_calls):
    results = []

    # ✅ Loop through all tool calls returned by the LLM
    for tool_call in tool_calls:

        # Extract tool name + arguments
        name = tool_call.function.name
        args = json.loads(tool_call.function.arguments)  # JSON string → dict

        # ✅ Execute the correct tool
        if name == "use_calculator":
            output = use_calculator(**args)
        else:
            output = "Unknown tool"

        # ✅ Format tool result → must be sent back to LLM
        results.append({
            "role": "tool",
            "tool_call_id": tool_call.id,
            "content": str(output)
        })

    # ✅ Return all tool results to be added back into messages
    return results
````

🔁 Agent loop + Memory

* 💡 The agent runs in a loop (Think → Act → Observe → Repeat), allowing it to solve tasks step by step using tools.

* 💡 The agent maintains memory by storing past interactions and passing them back to the LLM, enabling context-aware conversations.


````
import os
import openai


openai.api_key = os.getenv("OPENAI_API_KEY")

# ✅ System prompt → defines behavior of the agent
system_prompt = "You are a helpful AI assistant."


# ✅ Main chat agent (loop version)
def chat(user_input, history=None):

    # ✅ ✅ Initialize memory (conversation history) if not provided
    if history is None:
        history = []

    # ✅ Prepare messages (includes past memory)
    messages = (
        [{"role": "system", "content": system_prompt}]
        + history
        + [{"role": "user", "content": user_input}]
    )

    # ✅ Safety limit → prevents infinite loops
    max_steps = 5

    # ✅ 🔁 Agent Loop → Think → Act → Observe → Repeat
    for step in range(max_steps):

        # ✅ Step 1: Call LLM (reasoning)
        response = openai.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            tools=tools
        )

        choice = response.choices[0]
        message = choice.message
        finish_reason = choice.finish_reason

        # ✅ Step 2: If LLM wants to call a tool
        if finish_reason == "tool_calls":

            tool_calls = message.tool_calls

            # Debug: print tool calls
            for tool_call in tool_calls:
                print("\nTool:", tool_call.function.name)
                print("Args:", tool_call.function.arguments)

            # ✅ Step 3: Execute tools
            results = handle_tool_calls(tool_calls)

            # ✅ Step 4: Add LLM decision + tool results to messages
            messages.append(message)
            messages.extend(results)

        else:
            # ✅ Step 5: Final Output → LLM is done
            messages.append(message)

            print("\nFinal Output:", message.content)


            # ✅ 🧠 Memory: store conversation for future context
            history.append({"role": "user", "content": user_input})
            history.append({"role": "assistant", "content": message.content})

            return message.content

    # ✅ Fallback: If loop exceeds max steps
    return "Max steps reached"

````

▶️ Run Example

````
chat("add 10 and 20")
chat("explain the result of adding 10 and 20")
chat("hello!")
````

# ֎ **OpenAI Agents SDK**

The OpenAI Agents SDK provides a lightweight yet powerful framework for building multi-agent workflows. It is **provider-agnostic**, supporting OpenAI APIs as well as **100+ other LLMs**.

---

### **Core Primitives**

Instead of implementing all components manually, the SDK provides a small set of core primitives:

* **Agents** : which are LLMs equipped with instructions and tools
* **Agents as tools / Handoffs** : which allow agents to delegate to other agents for specific tasks
* **Guardrails**: which enable validation of agent inputs and outputs

---

### **Additional Capabilities**

- 📊 Built-in **tracing** for debugging and visualization  
- 📈 Support for **evaluation and optimization** of agent workflows  

---

### **Responses API (Low-Level) vs Agents SDK**

The SDK uses the Responses API by default for OpenAI models, but it adds a higher-level runtime around model calls.

✅ **Use the Responses API (Low-Level) when:**

- you want full control over loops, tools, and memory  
- your workflow is simple or short-lived  
- you prefer handling orchestration manually


✅ **Use the Agents SDK when:**
- you want automatic handling of multi-step execution  
- your agent uses tools, guardrails, or multiple steps  
- you prefer a structured workflow with less manual code

💡 You do not need to choose one globally. Many applications use the SDK for managed workflows and call the Responses API directly for lower-level paths.

## ✅ Example 6: Handoff
💡 Agents can delegate tasks to other agents automatically

```
                  User Input
                      ↓
             Main Agent (decides)
                      ↓
            [ decides to delegate ]
                      ↓
           ┌───────────────────────┐
           ↓                       ↓
       Research Agent        Writer Agent
               (only one runs)
           ↓                       ↓
           └───── Final Output ────┘
```
💡 Only one agent is selected because the goal is to delegate the **entire task** to the most appropriate agent

✅ Agent 1: Researcher Agent

💡 Adds guidance to help the main agent choose this agent for research-heavy tasks

````
from agents import Agent

# ✅ Agent 1: Research agent generates content (with guidance)
research_agent = Agent(
    name="researcher",
    instructions="Provide detailed factual information.",
    handoff_description="Use this agent for deep research and detailed explanations",
    model="gpt-4o-mini"

)
````

  ✅ Agent 2: Writer Agent

````
from agents import Agent, Runner, handoff

# ✅ Agent 2: Writer agent improves explanation (customized via handoff())
writer_agent = Agent(
    name="writer",
    instructions="Rewrite and simplify content clearly.",
    model="gpt-4o-mini"

)
````

✅ Main Agent (Handoff Logic)


````
triage_agent = Agent(
    name="triage-agent",
    instructions="""
    Decide how to handle the request:
    - Use researcher for detailed explanations
    - Use writer for simplification
    """,
    handoffs=[
        research_agent,  # ✅ default handoff (guided by handoff_description)
        handoff(         # ✅ customized handoff
            writer_agent,
            tool_name="simplifier",
            tool_description="Use this to simplify or rewrite content"
        )
    ],
    model="gpt-4o-mini"
)
````

💡 The agents are exposed as tools, and the triage agent selects one based on the task

💡 Tracing helps visualize how the main agent selects and delegates to another agent

💡 The purpose of handoff is not to combine multiple agents, but to choose the right agent for the task

* ❌ Not → Do this, then that, then combine

👉 For multi-step workflows involving multiple agents, use tools or manual orchestration instead

## ✅ Example 7: Guardrails
💡 Guardrails enable you to do **checks** and **validations** of **user input** and **agent output**.

👉 Example: Block unsafe or unnecessary requests before calling expensive models

---

### Types of Guardrails

- **Input Guardrails**  
  → run on user input before the agent executes  
  → Input guardrails run **only on the first agent**

- **Output Guardrails**  
  → run on the final agent output  
  → Output guardrails run **only on the final agent**  

- **Tool Guardrails**  
  → run before and after function-tool execution  
  → Tool guardrails run **on every tool call**.

💡 Use tool guardrails when you need validation for each tool execution in multi-agent workflows

---

### Execution Flow

####  🔹**Input Guardrail Execution Flow**

1. Receive user input  
2. Run guardrail function  
3. Check result:

   - ✅ Valid → agent executes  
   - ❌ Invalid → execution is stopped (tripwire triggered)


#### 🔹 **Output Guardrail Execution Flow**

1. Agent produces output  
2. Guardrail validates result  
3. If invalid → execution is blocked or error is raised

#### 🔹 **Tool Guardrail**

1. Agent decides to call a tool  
2. **Input tool guardrail** validates tool input  
3. If:
   - ✅ valid → tool executes  
   - ❌ invalid → **execution is blocked (tripwire)**  

4. Tool produces output  
5. **Output tool guardrail** validates tool output  
6. If invalid → **output is blocked / modified / error raised**

---

#### Execution modes

💡 Input guardrails support two execution modes:

- **Parallel (default, `run_in_parallel=True`)**  
  → guardrail runs **alongside the agent**  
  → faster response time ✅  
  → agent may already start execution before validation completes  

- **Blocking**(`run_in_parallel=False`)  
  → guardrail runs **before the agent executes**  
  → agent runs only if validation passes ✅  
  → prevents unnecessary token usage and tool execution  

---

### Tripwire

💡 A **tripwire** is triggered when validation fails:

- execution stops immediately ❌  
- agent does not continue  
- an error is raised

---

### Tool Guardrails (Important Notes)

💡 Tool guardrails validate tool execution **before and after the tool runs**

- Apply only to tools created using `function_tool` ✅  

- ❌ Do NOT apply to:
  - handoffs (separate pipeline)  
  - hosted / built-in tools  
  - `Agent.as_tool()`  


💡 Tool guardrails behavior:

- **Before execution (input guardrail)**  
  → can block the tool call or trigger a tripwire  

- **After execution (output guardrail)**  
  → can modify output or raise an error  

---

### 🔹 Key Insight

💡 Guardrails = **"validate → stop if unsafe"**

---

### ✅ Full Guardrail Flow
```
User Input  
   ↓  
Input Guardrail ✅ / ❌  
   ↓  
Agent Execution ✅  
   ↓  
Tool Guardrail (if tools used) ✅ / ❌  
   ↓  
Output Guardrail ✅ / ❌  
   ↓  
Final Output 🚀 or Blocked 🚫
```


✅ Input Guardrail

````
from agents import input_guardrail, GuardrailFunctionOutput

@input_guardrail
def validate_input(ctx, user_input: str):
    if "hack" in user_input.lower():
        return GuardrailFunctionOutput(
            output_info="Unsafe input detected",
            tripwire_triggered=True
        )

    return GuardrailFunctionOutput(
        output_info="Safe input",
        tripwire_triggered=False
    )
````

✅ Output Guardrail

````
from pydantic import BaseModel
from agents import (
    Agent,
    GuardrailFunctionOutput,
    RunContextWrapper,
    Runner,
    output_guardrail,
)

class MessageOutput(BaseModel):
    response: str

class MathOutput(BaseModel):
    reasoning: str
    is_math: bool

# ✅ Guardrail Agent
guardrail_agent = Agent(
    name="Guardrail check",
    instructions="""
    Determine whether the content contains mathematical expressions,
    calculations, or numeric reasoning.
    """,
    output_type=MathOutput,
    model="gpt-4o-mini"
)


@output_guardrail
async def output_math_guardrail(
                          ctx: RunContextWrapper,
                          agent: Agent,
                          output: MessageOutput
                        ) -> GuardrailFunctionOutput:

    result = await Runner.run(guardrail_agent, output.response, context=ctx.context)

    return GuardrailFunctionOutput(
        output_info=result.final_output,
        tripwire_triggered=result.final_output.is_math,
    )
````

✅ Tool guardrails -- (Unified)

- Unified → single function handling input + output  




````
from agents import Agent, function_tool, tool_guardrail, GuardrailFunctionOutput

# ✅ Tool
@function_tool
def calculator(a: int, b: int) -> int:
    return a + b


# ✅ Tool Guardrail (Input + Output validation)
@tool_guardrail
def calculator_guardrail(ctx, tool_call):

    # ✅ Before execution (input check)
    if tool_call.name == "calculator":
        if tool_call.args.get("a", 0) > 1000:
            return GuardrailFunctionOutput(
                output_info="Input too large",
                tripwire_triggered=True   # ❌ block tool
            )

    # ✅ After execution (output check)
    if tool_call.output is not None:
        if tool_call.output > 10000:
            return GuardrailFunctionOutput(
                output_info="Output exceeds limit",
                tripwire_triggered=True   # ❌ block output
            )

    return GuardrailFunctionOutput(tripwire_triggered=False)


# ✅ Agent using tool + guardrail
agent = Agent(
    name="tool-agent",
    instructions="Use the calculator to solve math problems.",
    tools=[calculator],
    tool_guardrails=[calculator_guardrail],
    model="gpt-4o-mini"
)
````

✅ Tool guardrails -- (Split)
- Split → separate input and output guardrails (more modular and recommended)

````
import json
from agents import (
    Agent,
    Runner,
    ToolGuardrailFunctionOutput,
    function_tool,
    tool_input_guardrail,
    tool_output_guardrail,
)

# ✅ Input Tool Guardrail (runs BEFORE tool execution)
@tool_input_guardrail
def block_large_numbers(data):

    # 🔹 tool_arguments is a JSON string → convert to dict
    args = json.loads(data.context.tool_arguments or "{}")

    # 🔹 validate input (before tool runs)
    if args.get("a", 0) > 1000:
        return ToolGuardrailFunctionOutput.reject_content(
            "Input too large. Reduce the values."
        )

    # ✅ allow execution
    return ToolGuardrailFunctionOutput.allow()


# ✅ Output Tool Guardrail (runs AFTER tool execution)
@tool_output_guardrail
def limit_output(data):

    # 🔹 get tool output
    output = data.output

    # 🔹 validate output after execution
    if output and output > 10000:
        return ToolGuardrailFunctionOutput.reject_content(
            "Output exceeds allowed limit."
        )

    # ✅ allow result
    return ToolGuardrailFunctionOutput.allow()


# ✅ Tool with guardrails attached
@function_tool(
    tool_input_guardrails=[block_large_numbers],   # ✅ before execution
    tool_output_guardrails=[limit_output],         # ✅ after execution
)
def calculator(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b


# ✅ Agent using the tool
agent = Agent(
    name="calculator-agent",
    instructions="Use the calculator to solve math problems.",
    tools=[calculator],
    model="gpt-4o-mini"
)
````

## ✅ Example 8: Streaming

💡 So far, we’ve only seen final outputs.  
Streaming allows us to observe how the agent executes **step-by-step in real time**.


💡 This is useful for:
- showing end-user progress updates  
- displaying partial responses  
- debugging agent behavior  

---

### 🔹 How to Enable Streaming

- Use `Runner.run_streamed()` → returns a **`RunResultStreaming`** object  
- Call `stream.stream_events()` → provides an **async stream of `StreamEvent` objects**

---

### 🔹 Consuming the Stream

- Iterate over `stream.stream_events()` using an async loop  
- Continue consuming events **until the iterator finishes**  

---

### 🔹 Important Behavior

- A streaming run is **not complete** until the async iterator ends  
- Even after the last visible output token, additional processing may still occur  

---

### 🔹 Final State

- Once the loop exits, `stream.is_complete` reflects the **final run state**

---

### 🔹 What Events Include

💡 Events in the stream may include:

- incremental output (text tokens)  
- tool calls  
- handoff (delegation) decisions  

👉 This provides full visibility into how the agent reasons and executes tasks

````
import asyncio
from agents import Runner


# ✅ Streaming Execution
async def main():
    # Start streaming
    stream = await Runner.run_streamed(
        triage_agent,
        "Explain Agentic AI and simplify it"
    )

    print("🔄 Streaming Output:\n")

    # Consume events
    async for event in stream.stream_events():
        # Print incremental text output
        if hasattr(event, "delta") and event.delta:
            print(event.delta, end="", flush=True)

    # Final completion state
    print("\n\n✅ Completed:", stream.is_complete)

# ✅ Run program
asyncio.run(main())
````

💡 This code streams the agent’s response in real time:

- `run_streamed()` starts execution and returns a streaming result  
- `stream_events()` provides a live stream of events  
- Each `event.delta` prints incremental output  
- The async loop continues until execution completes  


💡 Events represent everything happening during execution:
- text generation  
- tool calls  
- handoff decisions  

👉 This makes streaming useful for debugging and UI development

👉 **Agent = Model + Harness**

- The **model (LLM)** → performs reasoning 🧠

💡 The harness includes everything around the agent loop:

- **prompt** → defines behavior  
- **tools** → enable actions

👉 So:
```
Agent = LLM + (Prompt + Tools + Logic + Control)
```

## 🔹 Agent Behavior (Core Loop)

💡 In LangChain, an agent operates in a loop:
> ```
> Think → Call Tool → Observe → Repeat
>```

- Think → model decides next step
- Act → call tools (optional)
- Observe → It observes the result  
- Repeat → It continues until the task is complete


👉 This is the same **ReAct pattern** used in agent systems.

💡 In LangChain, the agent state consists of a sequence of **messages** that represent the ongoing conversation.

💡 Each invocation adds new messages to the state, and the agent
processes them using its **loop (Think → Act → Observe → Repeat)**.

```

💡 **context** → (external runtime data)
* Used to pass external data for a single run
* Accessible by tools and middleware
* Used for **user-specific data**, **API keys**, **feature flags**, **metadata for tools**

## ✅ Memory (Short-Term)

💡 Agents need memory to carry context across interactions.

👉 Without memory, each request is treated independently.  
👉 With memory, the agent can continue a conversation.

---

### 🔹 What is Short-Term Memory?

💡 Short-term memory allows the agent to **retain previous messages within a single conversation (thread)**.

- Maintains conversation history  
- Used during reasoning  
- **Limited to a single session**

---

### 🔹 How It Works

- **messages** → conversation history  
- **checkpointer** → persists messages  
- **thread_id** → identifies the conversation ✅  

---

### 🔹 Note

- 💡 Short-term memory is limited to a **single conversation (thread)**  
- 💡 Long-term memory enables the agent to retain information **across multiple conversations** (not covered here)
---

🔹 Step 1: Define Tool & Structured Output

````
from pydantic import BaseModel
from langchain.tools import tool

class Answer(BaseModel):
    result: int
    explanation: str

@tool
def add(a: int, b: int) -> int:
    """Adds two numbers"""
    return a + b
````

🔹 Step 3: Create the Agent

````
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver


agent = create_agent(
    model="openai:gpt-4o-mini",
    tools=[add],
    system_prompt="You are a helpful math assistant.",
    response_format=Answer,
    name="math_agent",
    checkpointer=InMemorySaver()         # enables memory ✅
)
````

🔹 Use a Conversation Thread

````
config = {
    "configurable": {"thread_id": "1"}
}
````

🔹 Interact Across Turns

````
# First interaction
response1 = agent.invoke(
    {"messages": [{"role": "user", "content": "My name is Bob"}]},
    config
)

print("Response 1:", response1["messages"][-1].content)
# Response 1: Nice to meet you, Bob!

# Second interaction (same thread)
response2 = agent.invoke(
    {"messages": [{"role": "user", "content": "What's my name?"}]},
    config
)

print("Response 2:", response2["messages"][-1].content)
# Response 2: You are Bob



# --- New thread (fresh memory) ---

new_config = {
    "configurable": {"thread_id": "2"}
}

response3 = agent.invoke(
    {"messages": [{"role": "user", "content": "What's my name?"}]},
    new_config
)

print("Response 3 (new thread):", response3["messages"][-1].content)
# Expected: I don't know your name
````

## ✅ Streaming

💡 Streaming allows you to observe agent execution **step-by-step in real time**

🔹 **Key Idea**

* **invoke()** → returns final output
* **stream()** → returns intermediate steps ✅



💡 For more details, refer to the official [Streaming documentation](https://docs.langchain.com/oss/python/langchain/streaming)



````
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model="openai:gpt-4o-mini",
    tools=[],
    checkpointer=InMemorySaver()   # memory enabled ✅
)


config = {"configurable": {"thread_id": "1"}}

# First interaction (store memory)
agent.invoke(
    {"messages": [{"role": "user", "content": "My name is Bob"}]},
    config
)
````

 **Supported stream modes**

 💡 stream_mode controls what type of data is streamed during execution.

* `stream_mode="updates"`
  * 👉 Streams state updates after each agent step.
  * 👉 If multiple updates are made in the same step (e.g., multiple nodes are run), those updates are streamed separately.
  * 👉 You get: *model steps* , *tool execution*, *final response*
  * ✔ Best for understanding agent workflow ✅
  * ✔ Useful for debugging ✅

* `stream_mode="messages"`
  * 👉 Streams LLM tokens with metadata
  * 👉 Streams tuples of (token, metadata) from any graph nodes where an LLM is invoked.
  * ✔ Best for chat UI ✅
  * ✔ Real-time typing experience ✅

* `stream_mode="custom"`
  * 👉 Streams user-defined data from tools or middleware
  * 👉Streams custom data from inside your graph nodes using the stream writer.
  * 👉 You can stream: *progress updates*, *logs*, *custom signals*
  * ✔ Useful for advanced workflows ✅

* `stream_mode="values"` (Simple)

  * 👉 Returns **messages (full state) at each step**
  * Access via: `event["messages"]`
  * Shows conversation as it evolves  
  * ✔ Best for simple examples ✅  


````
# Stream response (uses memory)
for event in agent.stream(
    {"messages": [{"role": "user", "content": "What's my name?"}]},
    config,
    stream_mode="updates"
):
    print(event)
````

## ✅ **Event Streaming**

💡 Event streaming provides a **structured way to observe agent execution in real time**.

---

👉 `stream()` → returns **mixed events** (messages, tool calls, updates)  
👉 Requires **manual parsing** ❌  

<br>

👉 `stream_events()` → returns **structured outputs**  (messages, tool calls, state updates)  
👉 Everything is **separate and easy to use ✅**

---
💡 Key Insight:  
* `stream()` → mixed  
* `stream_events()` → structured ✅


💡 For more details, refer to the official [Event Streaming documentation](https://docs.langchain.com/oss/python/langchain/event-streaming)

````
stream = agent.stream_events(
    {"messages": [{"role": "user", "content": "Hello"}]},
    version="v3"
)

# Stream model output
for message in stream.messages:
    for delta in message.text:
        print(delta, end="", flush=True)

# Final result
final_output = stream.output
````

### 🔹Custom Middleware

💡 Build custom middleware by implementing hooks that run at specific points in the agent execution flow.

💡 Middleware provides two types of hooks:

* ✅ **Node-style hooks** : Run at specific steps in the agent flow:

  - `before_agent` → before agent starts  
  - `before_model` → before model call  
  - `after_model` → after model response  
  - `after_agent` → after agent finishes  

  - ✔ Best for: *logging*, *validation*, *simple updates*  

* ✅ **Wrap-style hooks**: run around each call, giving you control over execution

  - `wrap_model_call` → Around each model call
  - `wrap_tool_call` → 	Around each tool call
  - ✔ Best for: *retries*, *fallbacks*, *caching*, * full execution control*  

---

🔹 Key Insight

💡 Node hooks = step-based control  
💡 Wrap hooks = execution control

👉 Together, they give full control over the agent lifecycle.

#### 🛠 Example: Node-style Middleware

````
from langchain.agents.middleware import before_model, after_model, AgentState
from langchain.messages import AIMessage
from langgraph.runtime import Runtime
from typing import Any

@before_model(can_jump_to=["end"]) # ✅ allows early exit
def check_message_limit(state: AgentState, runtime: Runtime) -> dict[str, Any] | None:
    if len(state["messages"]) >= 50:
        return {
            "messages": [AIMessage(content="Conversation limit reached.")],
            "jump_to": "end"  # ✅ stops execution early
        }
    return None

@after_model
def log_response(state: AgentState, runtime: Runtime) -> dict[str, Any] | None:
    print(f"Model returned: {state['messages'][-1].content}")  # ✅ logging
    return None
````

🔹 What This Does

* ✅ Stops execution when message limit is reached
* ✅ Logs model output after each response
* ✅ Uses jump_to="end" to exit the agent early (controls execution flow)

🔹 **Agent jumps**

💡 To exit early from middleware, return a dictionary with `jump_to`:

Available jump targets:
* `'end'`: Jump to the end of the agent execution (or the first after_agent hook)
* `'tools'`: Jump to the tools node
* `'model'`: Jump to the model node (or the first before_model hook)


#### 🛠 Example: Wrap-style Middleware (Retry)

````
from langchain.agents.middleware import wrap_model_call
from typing import Callable

@wrap_model_call
def retry_model(request, handler: Callable):
    for attempt in range(3):
        try:
            return handler(request)
        except Exception:
            if attempt == 2:
                raise
````

🔹 **Execution Order**

💡 When multiple middleware are used, execution follows specific rules:

```
agent = create_agent(
    model="gpt-5.4",
    middleware=[middleware1, middleware2, middleware3],
    tools=[...],
)
```

* `before_*` hooks: First to last
* `after_*` hooks: Last to first (reverse)
* `wrap_*` hooks: Nested (first middleware wraps all others)

- system prompt

---
🔹**system prompt**: The system prompt defines how the model should behave and respond.  


````
from langchain.agents import create_agent
from langchain.agents.middleware import dynamic_prompt, ModelRequest

@dynamic_prompt
def state_aware_prompt(request: ModelRequest) -> str:
    # request.messages is a shortcut for request.state["messages"]
    message_count = len(request.messages)

    base = "You are a helpful assistant."

    if message_count > 10:
        base += "\nThis is a long conversation - be extra concise."

    return base

agent = create_agent(
    model="openai:gpt-4o-mini",
    tools=[],
    middleware=[state_aware_prompt]
)
````

🔹**messages (conversation)**  

* 💡 Messages represent the conversation history passed to the model.  
* 👉 They provide context from past interactions to guide current responses.



````
from langchain.agents import create_agent
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from typing import Callable

@wrap_model_call
def add_instruction(request: ModelRequest, handler: Callable[[ModelRequest], ModelResponse]):
    # Append extra context to messages
    messages = [
        *request.messages,
        {"role": "user", "content": "Always respond in a concise way."}
    ]

    # Override messages for this model call
    request = request.override(messages=messages)

    return handler(request)

agent = create_agent(
    model="openai:gpt-4o-mini",
    tools=[],
    middleware=[add_instruction]
)
````

🔹 What This Does
- ✅ Uses existing conversation history (request.messages)
- ✅ Adds additional instruction dynamically
- ✅ Modifies what the model sees for that specific call

🔹 **model selection**
- 💡 Different models have different strengths, costs, and context windows.
- 💡 Select the right model for the task at hand, which might change during an agent run.

````
# Example: Dynamic Model Selection
from langchain.agents import create_agent
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from langchain.chat_models import init_chat_model
from typing import Callable

# Initialize models
simple_model = init_chat_model("gpt-4o-mini")
advanced_model = init_chat_model("gpt-5.4")

@wrap_model_call
def select_model(request: ModelRequest, handler: Callable[[ModelRequest], ModelResponse]):
    # Choose model based on conversation length
    if len(request.messages) > 10:
        model = advanced_model   # ✅ better reasoning for complex tasks
    else:
        model = simple_model     # ✅ faster and cheaper for simple tasks

    request = request.override(model=model)
    return handler(request)

agent = create_agent(
    model="gpt-4o-mini",  # default
    tools=[],
    middleware=[select_model]
)
````

🔹 **response format**
- 💡 Response format defines the **structure of the model’s final output**.  
- 👉 It ensures the output is consistent, validated, and usable in downstream systems.


````
from langchain.agents import create_agent
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from pydantic import BaseModel, Field
from typing import Callable

class SimpleResponse(BaseModel):
    """Simple response for early conversation."""
    answer: str = Field(description="A brief answer")

class DetailedResponse(BaseModel):
    """Detailed response for established conversation."""
    answer: str = Field(description="A detailed answer")
    reasoning: str = Field(description="Explanation of reasoning")
    confidence: float = Field(description="Confidence score 0-1")

@wrap_model_call
def state_based_output(
    request: ModelRequest,
    handler: Callable[[ModelRequest], ModelResponse]
) -> ModelResponse:
    """Select output format based on State."""
    # request.messages is a shortcut for request.state["messages"]
    message_count = len(request.messages)

    if message_count < 3:
        # Early conversation - use simple format
        request = request.override(response_format=SimpleResponse)
    else:
        # Established conversation - use detailed format
        request = request.override(response_format=DetailedResponse)

    return handler(request)

agent = create_agent(
    model="gpt-5.4",
    tools=[...],
    middleware=[state_based_output]
)
````

###🔹Custom Guardrails

💡 For more sophisticated guardrails, you can create custom middleware that runs before or after the agent executes.

>**Before agent guardrails** (Input Filter)
>
>💡Use “before agent” hooks to validate requests once at the start of each invocation.



````
from langchain.agents import create_agent
from langchain.agents.middleware import before_agent

@before_agent(can_jump_to=["end"])
def block_harmful_requests(state, runtime):
    content = state["messages"][0]["content"].lower()

    if "hack" in content:
        return {
            "messages": [{
                "role": "assistant",
                "content": "I cannot help with that request."
            }],
            "jump_to": "end"
        }

agent = create_agent(
    model="openai:gpt-4o-mini",
    tools=[],
    middleware=[block_harmful_requests]
)
````

>**After agent guardrails** (Output Validation)
>
> 💡 Use “after agent” hooks to validate final outputs once before returning to the user
>
> 💡 This is useful for model-based safety checks, quality validation, or final compliance scans on the complete agent response.

````
from langchain.agents.middleware import after_agent, AgentState, hook_config
from langgraph.runtime import Runtime
from langchain.messages import AIMessage
from langchain.chat_models import init_chat_model
from typing import Any

safety_model = init_chat_model("gpt-5.4-mini")

@after_agent(can_jump_to=["end"])
def safety_guardrail(state: AgentState, runtime: Runtime) -> dict[str, Any] | None:
    """Model-based guardrail: Use an LLM to evaluate response safety."""
    # Get the final AI response
    if not state["messages"]:
        return None

    last_message = state["messages"][-1]
    if not isinstance(last_message, AIMessage):
        return None

    # Use a model to evaluate safety
    safety_prompt = f"""Evaluate if this response is safe and appropriate.
    Respond with only 'SAFE' or 'UNSAFE'.

    Response: {last_message.content}"""

    result = safety_model.invoke([{"role": "user", "content": safety_prompt}])

    if "UNSAFE" in result.content:
        last_message.content = "I cannot provide that response. Please rephrase your request."

    return None

# Use the safety guardrail
from langchain.agents import create_agent

agent = create_agent(
    model="gpt-5.4",
    tools=[search_tool, calculator_tool],
    middleware=[safety_guardrail],
)

result = agent.invoke({
    "messages": [{"role": "user", "content": "How do I make explosives?"}]
})
````

### 🔹 Combining Multiple Guardrails

💡 You can stack multiple guardrails by adding them to the middleware array.

💡 They execute in order, allowing you to build layered protection:

👉 Each layer handles a different risk:

- input filtering → blocks harmful requests  
- PII detection → protects sensitive data  
- human approval → prevents risky actions  
- output validation → ensures safe responses  


````
from langchain.agents import create_agent
from langchain.agents.middleware import PIIMiddleware, HumanInTheLoopMiddleware

agent = create_agent(
    model="gpt-5.4",
    tools=[search_tool, send_email_tool],
    middleware=[
        # Layer 1: Deterministic input filter (before agent)
        ContentFilterMiddleware(banned_keywords=["hack", "exploit"]),

        # Layer 2: PII protection (before and after model)
        PIIMiddleware("email", strategy="redact", apply_to_input=True),
        PIIMiddleware("email", strategy="redact", apply_to_output=True),

        # Layer 3: Human approval for sensitive tools
        HumanInTheLoopMiddleware(interrupt_on={"send_email": True}),

        # Layer 4: Model-based safety check (after agent)
        SafetyGuardrailMiddleware(),
    ],
)
````


💡 Guardrails work best when applied in layers across the entire agent lifecycle


## ✅ Human-in-the-loop (HITL)
 💡 Human-in-the-loop middleware allows agents to **pause execution and require human approval** before taking certain actions.

 👉 Adds a human safety layer for critical decisions such as:
 - writing to a file  
 - executing SQL queries  
 - sending emails  
 - modifying databases  
 - any high-risk action

---
### 🔹 How It Works

💡 Implemented using middleware:

- agent proposes a tool call  
- middleware checks policy (`interrupt_on`)  
- execution pauses (interrupt)  
- human reviews and decides  
- agent resumes execution
---

 ### 🔹 Human Decisions

👉 A human can respond in 4 ways:

- ✅ **approve** → execute as-is  
- ✏️ **edit** → modify before execution  
- ❌ **reject** → block execution  
- 💬 **respond** → return human reply as tool result  

### 🔹 Configuring interrupts

````
from langchain.agents import create_agent
from langchain.agents.middleware import HumanInTheLoopMiddleware
from langgraph.checkpoint.memory import InMemorySaver


agent = create_agent(
    model="gpt-5.4",
    tools=[write_file, execute_sql, read_data],
    middleware=[
        HumanInTheLoopMiddleware(
            interrupt_on={
                "write_file": True,  # All decisions (approve, edit, reject, respond) allowed
                "execute_sql": {"allowed_decisions": ["approve", "reject"]},  # No editing allowed
                "read_data": False, # Safe operation, no approval needed
            },
            # Prefix for interrupt messages - combined with tool name and args to form the full message
            # e.g., "Tool execution pending approval: execute_sql with query='DELETE FROM...'"
            # Individual tools can override this by specifying a "description" in their interrupt config
            description_prefix="Tool execution pending approval",
        ),
    ],
    # Human-in-the-loop requires checkpointing to handle interrupts.
    # In production, use a persistent checkpointer like AsyncPostgresSaver.
    checkpointer=InMemorySaver(),
)
````

###🔹 Conditional Approval
💡 By default, every tool call listed in `interrupt_on` pauses for review.

💡 But Not all actions need approval — only risky ones.

👉 Example:

* ✅ allow read-only queries (SELECT)
* ❌ require approval for write queries (DELETE, UPDATE)

💡 To pause only some calls, add a `when` predicate to a tool’s `InterruptOnConfig`

````
from langchain.agents import create_agent
from langchain.agents.middleware import HumanInTheLoopMiddleware, ToolCallRequest
from langgraph.checkpoint.memory import InMemorySaver


def writes_outside_workspace(request: ToolCallRequest) -> bool:
    """Pause writes to paths outside the workspace directory."""
    path = request.tool_call["args"].get("path", "")
    return not path.startswith("/workspace/")


def is_write_query(request: ToolCallRequest) -> bool:
    """Pause SQL that isn't a read-only SELECT."""
    query = request.tool_call["args"].get("query", "")
    return not query.lstrip().upper().startswith("SELECT")


agent = create_agent(
    model="gpt-5.4",
    tools=[write_file, execute_sql, read_data],
    middleware=[
        HumanInTheLoopMiddleware(
            interrupt_on={
                "write_file": {
                    "allowed_decisions": ["approve", "edit", "reject"],
                    "when": writes_outside_workspace,
                },
                "execute_sql": {
                    "allowed_decisions": ["approve", "reject"],
                    "when": is_write_query,
                },
            },
        ),
    ],
    checkpointer=InMemorySaver(),
)
````

* When the `when` predicate returns `False`,  the call runs without interrupting.
* When it returns `True`, or when you omit when, the call pauses as usual.
* Calls that evaluate to `False` are never added to the interrupt batch, so a reviewer only sees the actions that need a decision.

### 🔹 Responding to Interrupts

💡 When an interrupt occurs, the agent pauses and waits for a **human decision** to continue execution.

👉 When a tool call requires approval:

- agent pauses execution  
- an interrupt is returned  
- human reviews the action  
- a decision is provided  
- agent resumes execution  


````
from langgraph.types import Command

# Human-in-the-loop leverages LangGraph's persistence layer.
# You must provide a thread ID to associate the execution with a conversation thread,
# so the conversation can be paused and resumed (as is needed for human review).
config = {"configurable": {"thread_id": "some_id"}}
# Run the graph until the interrupt is hit.
result = agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "Delete old records from the database",
            }
        ]
    },
    config=config,
    version="v2",
)


# Resume execution with human decision
agent.invoke(
    Command(
        resume={
            "decisions": [{"type": "approve"}]  # ✅ approve / reject / edit / respond
        }
    ),
    config={"configurable": {"thread_id": "1"}} # Same thread ID to resume the paused conversation
)
````


**Multiple decisions**

When multiple actions are under review, provide a decision for each action in the same order as they appear in the interrupt:
```
{
    "decisions": [
        {"type": "approve"},
        {
            "type": "edit",
            "edited_action": {
                "name": "tool_name",
                "args": {"param": "new_value"}
            }
        },
        {
            "type": "reject",
            "message": "This action is not allowed"
        }
    ]
}
```

### 🔹 Streaming with human-in-the-loop
* You can use `stream()` instead of `invoke()` to **get real-time updates** while the agent runs and handles interrupts.

* Use `stream_mode=['updates', 'messages']` with `version="v2"` to stream both agent progress and LLM tokens in the unified v2 format.

````
from langgraph.types import Command

config = {"configurable": {"thread_id": "1"}}

# Stream until interrupt
for event in agent.stream(
    {"messages": [{"role": "user", "content": "Delete old records"}]},
    config,
    stream_mode=["updates", "messages"],
    version="v2"
):
    print(event)

# Resume after approval
for event in agent.stream(
    Command(resume={"decisions": [{"type": "approve"}]}),
    config,
    stream_mode=["updates", "messages"],
    version="v2"
):
    print(event)

````

## ✅ Multi-Agent Systems

💡 Multi-agent systems use **multiple agents working together** to solve complex tasks.

👉 Instead of one large agent, tasks are distributed across smaller, focused agents.

### 🔹 Why Multi-Agent

💡 Multi-agent systems help with:

- 🧠 **Context management** → avoid overloading a single prompt  
- 👥 **Distributed development** → different teams build different agents  
- ⚡ **Parallelization** → multiple agents work at the same time.

👉 Multi-agent patterns are particularly valuable when a single agent has too many tools and makes poor decisions about which to use

### 🔹 Common Multi‑Agent Patterns

💡 Here are the main patterns for building multi-agent systems, each suited to different use cases:

* ✅ Subagents
  * A main agent coordinates subagents as tools.
  * All routing passes through the main agent.
  * which decides when and how to invoke each subagent.
* ✅ Handoffs
  * Behavior changes dynamically based on state.
  * Tool calls update a state variable that triggers routing or configuration changes,
  * switching agents or adjusting the current agent’s tools and prompt.
* ✅ Skills
  * Specialized prompts and knowledge loaded on-demand.
  * A single agent stays in control while loading context from skills as needed.

* ✅ Router
  * A routing step classifies input and directs it to one or more specialized agents.
  * Results are synthesized into a combined response.

* ✅ Custom workflow

  * Build fully custom orchestration using LangGraph
  * mixing deterministic logic and agentic behavior.
  * Embed other patterns as nodes in your workflow.

### 📊 Choosing a Pattern

Use this table to match your requirements to the right pattern:

| Pattern    | Distributed Development | Parallelization | Multi-hop | Direct User Interaction |
|------------|------------------------|-----------------|-----------|--------------------------|
| Subagents  | ⭐⭐⭐⭐⭐                 | ⭐⭐⭐⭐⭐        | ⭐⭐⭐⭐⭐  | ⭐                       |
| Handoffs   | –                      | –               | ⭐⭐⭐⭐⭐  | ⭐⭐⭐⭐⭐               |
| Skills     | ⭐⭐⭐⭐⭐                 | ⭐⭐⭐          | ⭐⭐⭐⭐⭐  | ⭐⭐⭐⭐⭐               |
| Router     | ⭐⭐⭐                   | ⭐⭐⭐⭐⭐        | –         | ⭐⭐⭐                    |

#### 🔍 Metric Definitions

- **Distributed development**: Can different teams maintain components independently?  
- **Parallelization**: Can multiple agents execute concurrently?  
- **Multi-hop**: Does the pattern support calling multiple subagents in sequence?  
- **Direct user interaction**: Can subagents converse directly with the user?




### 🔹 Subagents

💡 A central main agent (the **Supervisor**) coordinates specialized worker subagents by calling them as tools.

💡 **Core Concept**: Subagents are stateless. They do not remember past conversations. The main supervisor handles all global memory, while subagents work in completely clean, isolated context windows to prevent token bloat.

👉 **Key Characteristics**:
* **Central Control**: All routing passes through the main agent.

* **No Direct User Chat**: Subagents return results back to the supervisor, not directly to the end-user.

* **provides context isolation**: each subagent invocation works in a clean context window, preventing context bloat in the main conversation.

#### 🛠 Example 1: Basic Implementation
The core mechanism wraps a subagent as a tool that the main agent can call

````
from langchain.tools import tool
from langchain.agents import create_agent

# Create a subagent
subagent = create_agent(model="google_genai:gemini-3.5-flash",
                        tools=[...],
                        system_prompt="You are an expert researcher. Return detailed findings.")

# Wrap it as a tool
@tool("research", description="Research a topic and return findings")
def call_research_agent(query: str):
    result = subagent.invoke({"messages": [{"role": "user", "content": query}]})
    return result["messages"][-1].content

# Main agent with subagent as a tool
main_agent = create_agent(model="google_genai:gemini-3.5-flash",
                          tools=[call_research_agent],
                          system_prompt="You are a supervisor. Coordinate worker agents to fulfill the request.")
````


#### 🛠 Example 2: Single dispatch tool
image.png

💡 Instead of giving your main agent a long list of separate sub-agent tools, you give it just one smart routing tool. This tool can talk to any temporary sub-agent on demand.

**Why it helps**:
* **Easier updates** → You can add new sub-agents anytime without changing the main agent's tools.
* **Team friendly** → Different teams can build and manage separate sub-agents independently.
* **Clean memory** → Complex tasks happen in separate windows, so the main agent's memory doesn't get cluttered with unnecessary details.

👉 **The Trade-off**: You lose some ability to fine-tune how data is shared between agents, but you gain a system that is incredibly simple to build, scale, and maintain.

👉 **When to use this**: Use this when your system is growing fast and you want a clean, scalable way to add new specialized agents without breaking your main agent.

````
from langchain.tools import tool
from langchain.agents import create_agent

# Sub-agents developed by different teams
research_agent = create_agent(model="gpt-5.4", prompt="You are a research specialist...")

writer_agent = create_agent(model="gpt-5.4", prompt="You are a writing specialist...")

# Registry of available sub-agents
SUBAGENTS = {
    "research": research_agent,
    "writer": writer_agent,
}

@tool
def task(
    agent_name: str,
    description: str
) -> str:
    """Launch an ephemeral subagent for a task.

    Available agents:
    - research: Research and fact-finding
    - writer: Content creation and editing
    """
    agent = SUBAGENTS[agent_name]
    result = agent.invoke({
        "messages": [
            {"role": "user", "content": description}
        ]
    })
    return result["messages"][-1].content

# Main coordinator agent
main_agent = create_agent(
    model="gpt-5.4",
    tools=[task],
    system_prompt=(
        "You coordinate specialized sub-agents. "
        "Available: research (fact-finding), "
        "writer (content creation). "
        "Use the task tool to delegate work."
    ),
)
````

📊 **Subagent Design Decisions**

💡 When implementing the subagents pattern, you need to configure how data and execution flow between your supervisor and your subagents.

|Decision|Options|Best For|
|---|---|---|
|1. Execution Mode|Sync (Blocking) vs. Async (Background)|Choosing between instant answers or long-running tasks.|
|2. Tool Pattern|Tool per Agent vs. Single Dispatch Tool|Choosing between custom tool definitions or dynamic scalability.|
|3. Subagent Specs|System Prompt vs. Enum Constraint vs. Tool Discovery|Controlling how the supervisor finds and picks the right agent.|
4. Subagent Inputs|Query Only vs. Full Context |Deciding if a subagent needs just a task string or past chat history.|
5. Subagent Outputs|Result Only vs. Full Conversation History |Deciding if the supervisor gets just the answer or the whole thought process.|

### 🔹 Handoffs

💡 In the handoffs pattern, different specialized agents take turns talking **directly to the user**. Instead of a central boss agent routing everything, the agents themselves use tools to pass the conversation back and forth.

👉 **Core Concept**: Think of this like a customer support phone system. You start by talking to a `Flight Agent`. When you say you want to book a room, that agent hits a transfer button on their phone, passes your call and notes over to the `Hotel Agent`, and hangs up.

**Key Characteristics**:
* **Direct Interaction**: The active agent speaks directly to the user. There is no middleman or supervisor filtering the messages.
* **Shared Memory**: The new agent gets the entire chat history, so the user never has to repeat themselves after a transfer.
* **One at a Time**: This pattern works in a straight line. Agents cannot work in parallel or call multiple sub-agents at the exact same time.

👉 **When to use this**: Use this when your system has distinct departments and the user needs to finish one task before moving to the next (e.g., verifying an account with a `Security Agent` before talking to a `Billing Agent`).

🛠 **The Core Mechanism**: The Handoff ToolTo make a handoff happen, a tool must return a `Command` object to update the system state.

⚠️ **Important Rule**: When an agent runs a handoff tool, it must include a **ToolMessage** matching the **tool_call_id**. If you forget this, the LLM's chat history breaks, and the application will crash.

````
from langchain.tools import tool
from langchain.messages import ToolMessage
from langgraph.types import Command

@tool
def transfer_to_specialist(runtime) -> Command:
    """Transfer the user to the specialist agent step."""
    return Command(
        update={
            # 1. Safely complete the model's tool request cycle ✅
            "messages": [
                ToolMessage(
                    content="Transferred to specialist",
                    tool_call_id=runtime.tool_call_id
                )
            ],
            # 2. Change the current step to trigger new rules
            "current_step": "specialist"
        }
    )
````

#### 🛠 Example 1: Single Agent with Middleware

* You build just one single agent, but it dynamically changes its behavior
* Middleware intercepts each model call and dynamically adjusts the system prompt and available tools.
* Tools update the state variable to trigger transitions:

````
from langchain.agents import AgentState, create_agent
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from langchain.tools import tool, ToolRuntime
from langchain.messages import ToolMessage
from langgraph.types import Command
from typing import Callable

# Step 1: Define the state schema to track the active step
class SupportState(AgentState):
    """Track which step is currently active."""
    current_step: str = "triage"
    warranty_status: str | None = None

# Step 2: Create a tool that updates the current_step via a Command
@tool
def record_warranty_status(
    status: str,
    runtime: ToolRuntime[None, SupportState]
) -> Command:
    """Record warranty status and transition to next step."""
    return Command(update={
        "messages": [
            ToolMessage(
                content=f"Warranty status recorded: {status}",
                tool_call_id=runtime.tool_call_id
            )
        ],
        "warranty_status": status,
        # Transition to next step
        "current_step": "specialist"
    })

# Step 3: Middleware swaps instructions and tools based on the current_step
@wrap_model_call
def apply_step_config(
    request: ModelRequest,
    handler: Callable[[ModelRequest], ModelResponse]
) -> ModelResponse:
    """Configure agent behavior based on current_step."""
    step = request.state.get("current_step", "triage")

    # Map steps to their configurations
    configs = {
        "triage": {
            "prompt": "Collect warranty information...",
            "tools": [record_warranty_status]
        },
        "specialist": {
            "prompt": "Provide solutions based on warranty: {warranty_status}",
            "tools": [provide_solution, escalate]
        }
    }

    config = configs[step]
    request = request.override(
        system_prompt=config["prompt"].format(**request.state),
        tools=config["tools"]
    )
    return handler(request)

# Step 4. Create agent with middleware
agent = create_agent(
    model,
    tools=[record_warranty_status, provide_solution, escalate],
    state_schema=SupportState,
    middleware=[apply_step_config],
    checkpointer=InMemorySaver()  # Persist state across turns  #
)
````

#### 🛠 Example 2: Multiple Agent Subgraphs
* Instead of changing a single agent's instructions, you create completely separate, distinct agents.
* Each agent lives as its own node inside a graph framework (like LangGraph).

* When a handoff tool runs, its **Command** payload tells the framework to stop running the current agent node and routing execution to a completely different agent node.

````
from typing import Literal

from langchain.agents import AgentState, create_agent
from langchain.messages import AIMessage, ToolMessage
from langchain.tools import tool, ToolRuntime
from langgraph.graph import StateGraph, START, END
from langgraph.types import Command
from typing_extensions import NotRequired


# 1. Define state with active_agent tracker
class MultiAgentState(AgentState):
    active_agent: NotRequired[str]


# 2. Create handoff tools
@tool
def transfer_to_sales(
    runtime: ToolRuntime,
) -> Command:
    """Transfer to the sales agent."""
    last_ai_message = next(
        msg for msg in reversed(runtime.state["messages"]) if isinstance(msg, AIMessage)
    )
    transfer_message = ToolMessage(
        content="Transferred to sales agent from support agent",
        tool_call_id=runtime.tool_call_id,
    )
    return Command(
        goto="sales_agent",
        update={
            "active_agent": "sales_agent",
            "messages": [last_ai_message, transfer_message],
        },
        graph=Command.PARENT,
    )


@tool
def transfer_to_support(
    runtime: ToolRuntime,
) -> Command:
    """Transfer to the support agent."""
    last_ai_message = next(
        msg for msg in reversed(runtime.state["messages"]) if isinstance(msg, AIMessage)
    )
    transfer_message = ToolMessage(
        content="Transferred to support agent from sales agent",
        tool_call_id=runtime.tool_call_id,
    )
    return Command(
        goto="support_agent",
        update={
            "active_agent": "support_agent",
            "messages": [last_ai_message, transfer_message],
        },
        graph=Command.PARENT,
    )


# 3. Create agents with handoff tools
sales_agent = create_agent(
    model="google_genai:gemini-3.5-flash",
    tools=[transfer_to_support],
    system_prompt="You are a sales agent. Help with sales inquiries. If asked about technical issues or support, transfer to the support agent.",
)

support_agent = create_agent(
    model="google_genai:gemini-3.5-flash",
    tools=[transfer_to_sales],
    system_prompt="You are a support agent. Help with technical issues. If asked about pricing or purchasing, transfer to the sales agent.",
)


# 4. Create agent nodes that invoke the agents
def call_sales_agent(state: MultiAgentState) -> Command:
    """Node that calls the sales agent."""
    response = sales_agent.invoke(state)
    return response


def call_support_agent(state: MultiAgentState) -> Command:
    """Node that calls the support agent."""
    response = support_agent.invoke(state)
    return response


# 5. Create router that checks if we should end or continue
def route_after_agent(
    state: MultiAgentState,
) -> Literal["sales_agent", "support_agent", "__end__"]:
    """Route based on active_agent, or END if the agent finished without handoff."""
    messages = state.get("messages", [])

    # Check the last message - if it's an AIMessage without tool calls, we're done
    if messages:
        last_msg = messages[-1]
        if isinstance(last_msg, AIMessage) and not last_msg.tool_calls:
            return "__end__"

    # Otherwise route to the active agent
    active = state.get("active_agent", "sales_agent")
    return active if active else "sales_agent"


def route_initial(
    state: MultiAgentState,
) -> Literal["sales_agent", "support_agent"]:
    """Route to the active agent based on state, default to sales agent."""
    return state.get("active_agent") or "sales_agent"


# 6. Build the graph
builder = StateGraph(MultiAgentState)
builder.add_node("sales_agent", call_sales_agent)
builder.add_node("support_agent", call_support_agent)

# Start with conditional routing based on initial active_agent
builder.add_conditional_edges(START, route_initial, ["sales_agent", "support_agent"])

# After each agent, check if we should end or route to another agent
builder.add_conditional_edges(
    "sales_agent", route_after_agent, ["sales_agent", "support_agent", END]
)
builder.add_conditional_edges(
    "support_agent", route_after_agent, ["sales_agent", "support_agent", END]
)

graph = builder.compile()
result = graph.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "Hi, I'm having trouble with my account login. Can you help?",
            }
        ]
    }
)

for msg in result["messages"]:
    msg.pretty_print()
````

**Implementation considerations**

As you design your multi-agent system, consider:
* **Context filtering strategy**: Will each agent receive full conversation history, filtered portions, or summaries? Different agents may need different context depending on their role.
* **Tool semantics**: Clarify whether handoff tools only update routing state or also perform side effects. For example, should transfer_to_sales() also create a support ticket, or should that be a separate action?
* **Token efficiency**: Balance context completeness against token costs. Summarization and selective context passing become more important as conversations grow longer.

###🔹 Skills (On-Demand Context Loading)

💡 In the **skills** architecture, a single agent stays in complete control of the conversation, but it **dynamically loads** specialized prompts, guidelines, and domain knowledge on-demand.

👉 **Core Concept**: Instead of stuffing instructions for 50 different domains into one giant system prompt (which triggers reasoning issues and explodes token costs), the agent uses a tool to fetch specific instructions only when a task requires them. Think of it as on-demand documentation loading.

```
                 ┌─────────────────┐
                 │      User       │
                 └────────┬────────┘
                          │
                          ▼
                 ┌─────────────────┐
                 │  Single Agent   │◀─── Loading Skills
                 │ (Stays Active)  │     On-Demand
                 └────────┬────────┘
                          │
           ┌──────────────┼──────────────┐
           ▼              ▼              ▼
     ┌───────────┐  ┌───────────┐  ┌───────────┐
     │ SQL Skill │  │ Git Skill │  │ CRM Skill │
     └───────────┘  └───────────┘  └───────────┘
```
**Key Characteristics**:

* **Single Agent Execution**: The active agent node never changes, meaning conversation history and global state stay unified in one spot.

* **Prompt-Driven Specialization**: Skills are lightweight modules composed of specialized instructions, rather than heavy, standalone sub-agents.

* **Distributed Maintenance**: Different product squads can write, test, and optimize separate skill instructions independently.

👉 **When to use this**: Use this pattern when you want a single general-purpose assistant that needs to specialize in dozens of different disciplines (e.g., a software engineering helper that needs distinct rules for Python, SQL, and Git on-demand).

#### 🛠 Example 1: Loading Skills with Middleware

This pattern uses a custom tool to save skill instructions into the agent's state, combined with middleware that dynamically updates the system prompt on the next turn.

````
from typing import Callable
from langchain.agents import AgentState, create_agent
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from langchain.tools import tool, ToolRuntime

# Step 1: Define agent state to hold the active skill context
class SkillState(AgentState):
    active_skill_content: str | None = None

# Mock Database of Skills Instructions
SKILLS_DB = {
    "sql": "You are a SQL Expert. Always wrap table names in brackets and format queries cleanly.",
    "charts": "You are a Data Visualizer. Use matplotlib to generate clean, readable graphs."
}

# Step 2: Create a tool to fetch skill context on-demand
@tool
def load_skill(skill_name: str, runtime: ToolRuntime[None, SkillState]) -> str:
    """Loads a specialized skill on-demand when the user switches tasks."""
    skill_content = SKILLS_DB.get(skill_name)
    if skill_content:
        # Save the skill text directly into the agent state
        runtime.state["active_skill_content"] = skill_content
        return f"Successfully loaded the {skill_name} skill."
    return "Skill not found."

# Step 3: Middleware appends the loaded skill to the base system prompt
@wrap_model_call
def inject_skill_prompt(
    request: ModelRequest,
    handler: Callable[[ModelRequest], ModelResponse]
) -> ModelResponse:
    """Combines base instructions with on-demand skill content before calling the LLM."""
    active_skill = request.state.get("active_skill_content")
    base_prompt = "You are a helpful assistant."

    if active_skill:
        base_prompt = f"{base_prompt}\n\n[Active Skill Instructions]:\n{active_skill}"

    request = request.override(system_prompt=base_prompt)
    return handler(request)

# Step 4: Assemble the complete system
agent = create_agent(
    model="google_genai:gemini-3.5-flash",
    tools=[load_skill],
    state_schema=SkillState,
    middleware=[inject_skill_prompt]
)
````

###🔹 Router
💡 In the router architecture, a routing step classifies input and directs it to specialized agents.

👉 **Core Concept**: A router acts like an automated mail sorting system. It evaluates an incoming request, figures out where it belongs, dispatches it to the correct specialized agent, and returns the result

```
                        ┌───────────────┐
                        │     User      │
                        └───────┬───────┘
                                │
                                ▼
                        ┌────────────────┐
                        │    Router      │
                        │(Classification)│
                        └───────┬────────┘
                                │
                 ┌──────────────┼──────────────┐
                 ▼              ▼              ▼
          ┌────────────┐ ┌────────────┐ ┌────────────┐
          │ Slack Agent│ │Notion Agent│ │GitHub Agent│
          └────────────┘ └────────────┘ └────────────┘
```

**Key Characteristics**:
* **Stateless Preprocessing**: Classifies and forwards tasks instantly without holding an ongoing conversation loop.
* **Parallel Fan-out**: Queries multiple specialized backend agents or data sources at the exact same time.
* **Router vs. Supervisor**: A router is a single dispatch step with no ongoing context; a supervisor is a full agent that coordinates multi-turn dialogue.

👉 **When to use this**: Use the router pattern when you have distinct verticals (separate knowledge domains that each require their own agent), need to query multiple sources in parallel, and want to synthesize results into a combined response.
