# Introduction

**FastAPI** is a modern, high-performance Python web framework for building **RESTful APIs**. It is designed for speed, simplicity, and developer productivity, leveraging Python’s type hints for automatic validation and documentation.

<br>

## ✅ FastAPI is commonly used for

- Building scalable backend services  
- Creating machine learning / AI APIs  
- Developing microservices and data-driven applications  
- Prototyping APIs quickly for frontend/mobile teams  
- Creating high-performance async services


<br>

## 🚀 Key Features

- ⚡ High performance — comparable to Node.js and Go  
- ✅ Automatic validation using Python type hints  
- 📄 Auto-generated API docs (Swagger UI & OpenAPI)  
- 🔁 Async support using `async` / `await`  
- 🧩 Easy integration with modern tools like `uv`, `Uvicorn`, and `Pydantic`  

<br>

## 🌐 REST API Basics

A **REST API (Representational State Transfer API)** allows different systems to communicate over HTTP using standard methods and URLs.

<br>

### 🔑 **6 Core Principles of REST APIs**

- **Stateless** — Each request contains all required information; the server does not store client state  
- **Client–Server Architecture** — The client and server are independent, enabling separate development and scaling  
- **Resource-Based** — Everything is treated as a resource identified by URLs (e.g., `/users`, `/orders`)  
- **Uniform Interface** — Standard HTTP methods (GET, POST, PUT, DELETE) are used consistently  
- **Representation** — Resources are transferred in a standard format, typically JSON  
- **Cacheable** — Responses can be cached to improve performance and reduce server load  

<br>

### 🔁 **How REST APIs Work**

* A REST API follows a request–response model, where a client sends a request to a server, and the server processes it and returns a response.

* This flow represents the complete lifecycle of a request from client to server and back in a RESTful system

  ```
    Client (Browser / Mobile App)
               ↓
     HTTP Request (GET / POST)
               ↓
    FastAPI Application (Uvicorn)
               ↓
    Validation & Business Logic
               ↓
    Database / External Services
               ↓
          JSON Response
               ↓
      Client (Consumes Data)
  ```

# CRUD Operations

**CRUD** stands for:

- **Create** → Add new data  
- **Read** → Retrieve existing data  
- **Update** → Modify existing data  
- **Delete** → Remove data  

These operations map directly to HTTP methods:

| CRUD Operation | HTTP Method | Purpose | Example Endpoint |
|---------------|------------|------|------------------|
| Create        | POST       | Add data |/users           |
| Read          | GET        | Fetch data |/users/{id}      |
| Update        | PUT/PATCH  | Modify data |/users/{id}      |
| Delete        | DELETE     | Remove data |/users/{id}      |

---

##  📌 Basic HTTP Methods

`````python
from fastapi import FastAPI

app = FastAPI()


# GET
@app.get("/api-endpoint1")
async def read_item():
    return {'message': 'Hello Eric!'}


# POST
@app.post("/api-endpoint2")
async def create_item():
    return {'message': 'Data created successfully'}


# PUT
@app.put("/api-endpoint3")
async def update_item():
    return {'message': 'Data updated successfully'}


# DELETE
@app.delete("/api-endpoint4")
async def delete_item():
    return {'message': 'Data deleted successfully'}
`````

Run the server:
uv run uvicorn main:app --reload

## 📌 Path Parameter
Path parameters allow you to pass dynamic values in the URL.

* Query parameters are optional
* Defined using = (default values)
* Passed using ?key=value
* Multiple params use &


## ⚠️ Important Rule

👉 **Always define static paths before dynamic paths**


`````python
#static paths
@app.get("/items/special")
async def get_special_item():
    return {"message": "This is a special item"}

# dynamic paths
@app.get("/items/{item_id}")
async def get_item(item_id: int):
    return {"item_id": item_id}
`````

🧠 **Key Idea**

* `{item_id}` is a dynamic value
* Value is taken from the URL
* FastAPI automatically converts it to the correct data type


## 📌 Query Parameters

Query parameters are used to send optional data in the URL.

They come after `?` in the URL.


`````python
@app.get("/items")
async def get_items(limit: int = 10):
    return {"limit": limit}
`````

✅ How to use
* Open in browser: http://127.0.0.1:8000/items?limit=5
* Swagger UI: http://127.0.0.1:8000/docs

```
Output:
    {

       "limit": 5
    }
```

 **Multiple Query Parameters**

 * How to use : http://127.0.0.1:8000/products?skip=2&limit=5

`````python
@app.get("/products")
async def get_products(skip: int = 0, limit: int = 10):
    return {"skip": skip, "limit": limit}
`````

**Path + Query Parameters Together**

You can use **both path and query parameters in the same API**.


`````python
@app.get("/items/{item_id}")
async def get_item(item_id: int, limit: int = 10):
    return {
        "item_id": item_id,
        "limit": limit
    }
`````

✅ **How to use**
* Open in browser:
http://127.0.0.1:8000/items/5?limit=2
* Swagger UI:
http://127.0.0.1:8000/docs

<br>

🧠 **Key Idea**
* item_id → Path parameter (required → identifies resource)
* limit → Query parameter (optional → modifies result)

> “Path parameters identify the resource, query parameters control how we fetch it.”

## 📌 Request Body
* Request body is used to send data to the API (mainly for POST and PUT).

* You can send **multiple values in the request body** using `Body()

* Data is usually sent in **JSON format**.

`````python

from fastapi import FastAPI, Body

app = FastAPI()

data_store = {}


# ⚠️ READ (GET) — Body not recommended / Method Limitation

# - GET requests are used to **retrieve data**
# - ❌ They should NOT use request body
# - ✅ Use path or query parameters instead



# ✅ CREATE (POST)
@app.post("/items")
async def create_item(item: dict = Body()):
    data_store["item"] = item
    return {
        "message": "Item created",
        "data": item
    }



# ✅ UPDATE (PUT)
@app.put("/items")
async def update_item(item: dict = Body()):
    data_store["item"] = item
    return {
        "message": "Item updated",
        "data": item
    }



# ✅ DELETE (DELETE)
@app.delete("/items")
async def delete_item(confirm: dict = Body(...)):
    deleted = data_store.pop("item", None)
    return {
        "message": "Item deleted",
        "confirmation": confirm,
        "deleted_data": deleted
    }


#Multiple Request Body Parameters
@app.post("/items/multiple")
async def create_item(
    name: str = Body(),
    price: float = Body(),
    in_stock: bool = Body()
):
    return {
        "name": name,
        "price": price,
        "in_stock": in_stock
    }
`````

**Combination Example**

Now let’s combine everything we’ve learned into one real API.

`````python
from fastapi import FastAPI, Body

app = FastAPI()


@app.post("/users/{user_id}/orders/{order_id}")
async def create_order(
    user_id: int,                     # ✅ Path parameter 1
    order_id: int,                    # ✅ Path parameter 2

    status: str = "new",              # ✅ Query parameter 1
    priority: bool = False,           # ✅ Query parameter 2

    item_name: str = Body(...),       # ✅ Body field 1
    price: float = Body(...),         # ✅ Body field 2
    quantity: int = Body(...)         # ✅ Body field 3
):
    return {
        "user_id": user_id,
        "order_id": order_id,
        "status": status,
        "priority": priority,
        "item": {
            "name": item_name,
            "price": price,
            "quantity": quantity
        }
    }
`````

✅ **How to use**
* URL (Path + Query) : http://127.0.0.1:8000/users/1/orders/101?status=confirmed&priority=true
* Request Body:
  ```
    {
      "item_name": "Laptop",
      "price": 50000,
      "quantity": 2
    }
  ```
* Output:

  ```
    {
      "user_id": 1,
      "order_id": 101,
      "status": "confirmed",
      "priority": true,
      "item": {
        "name": "Laptop",
        "price": 50000,
        "quantity": 2
        }
    }
  ```

>“Path identifies who/what, query modifies how, and body contains actual data.”

>⚠️ When using multiple body fields, Body() is required — otherwise FastAPI treats them as query parameters




# Data Validation

FastAPI uses **Pydantic** to define structured request bodies and validate input data automatically

“Validation happens in three places: Body, Path, and Query parameters.”

<br>

**Why Data Validation is Important**

- Ensures only **correct and valid data** is accepted  
- Prevents **invalid or unsafe data** from entering your API  
- Reduces manual validation code  
- Improves API reliability and consistency  

<br>

**How FastAPI Handles Validation**

- Converts incoming JSON → Python objects  
- Validates data using Pydantic rules  
- Automatically rejects invalid data  
- Returns detailed error messages  





## 📌 **Body Validation (Pydantic)**

### ✅ **Model Definition**

Defines the structure and validation rules for incoming request data.

🧠 **Key Understanding**

* **BaseModel** → defines structure of request data
* **Field()** → adds validation rules
* **Optional** → makes a field optional
* **model_config** → improves API documentation (Swagger UI)

`````python
from pydantic import BaseModel, Field
from typing import Optional

class BookRequest(BaseModel):

    # Field rules:
    id: Optional[int] = Field(description='ID is not needed on create', default=None) # optional field
    title: str = Field(min_length=3)                       # at least 3 characters
    author: str = Field(min_length=1)                      # cannot be empty
    description: str = Field(min_length=1, max_length=100) # value between 1–100 characters
    rating: int = Field(gt=0, lt=6)                        # values between 1 and 5
    published_date: int = Field(gt=1999, lt=2031)          # Value year between 2000 and 2030


    # Adds example request automatically in Swagger UI
    model_config = {
        "json_schema_extra": {
            "example": {
                "title": "A new book",
                "author": "codingwithroby",
                "description": "A new description of a book",
                "rating": 5,
                "published_date": 2029
            }
        }
    }
`````

### ✅ **API Example**

Uses the Pydantic model to validate incoming request data.


**What happens in this API**

- Receives JSON input from the client  
- Converts JSON → `BookRequest` object  
- Validates data before executing the function  
- Returns structured response  


`````python

from fastapi import FastAPI

app = FastAPI()

@app.post("/books")
async def create_book(book: BookRequest):
    return {
        "message": "Book created successfully",
        "data": book
    }
`````

✅ **How to Test**

* Open Swagger UI: http://127.0.0.1:8000/docs

✅ **Example Input**
* Input:

  ```
      {
        "title": "A new book",
        "author": "codingwithroby",
        "description": "A new description of a book",
        "rating": 5,
        "published_date": 2029
      }
  ```


💡 **What happens if input is invalid?**

- FastAPI automatically returns an error response  
- The function will NOT run
- 👉 Example:
  - If `rating = 10`, API will reject the request  

> “Pydantic validates input before it reaches your API logic.”

## 📌 Path Parameter Validation

Path parameters can also have type validation and constraints.

**What this does**

* int → ensures value is a number
* gt=0 → value must be greater than 0
* description → shown in Swagger UI



`````python
from fastapi import FastAPI, Path

app = FastAPI()

@app.get("/books/{book_id}")
async def get_book(
    book_id: int = Path(..., gt=0, description="Book ID must be greater than 0")
):
    return {"book_id": book_id}
`````

## 📌 Query Parameter Validation

<br>

**What this does**
* Default value = 10
* gt=0 → must be greater than 0

## 📌 Path + Query + Body Validation Together

You can apply validation to **path parameters, query parameters, and request body at the same time**.



`````python
from fastapi import FastAPI, Path, Query
from pydantic import BaseModel, Field
from typing import Optional

app = FastAPI()


class BookRequest(BaseModel):
    title: str = Field(min_length=3)
    price: float = Field(gt=0)


@app.post("/users/{user_id}/books/{book_id}")
async def create_book(
    user_id: int = Path(..., gt=0, description="User ID must be > 0"),
    book_id: int = Path(..., gt=0, description="Book ID must be > 0"),
    discount: float = Query(0, ge=0, le=50, description="Discount between 0–50%"),
    book: BookRequest = ...
):
    return {
        "user_id": user_id,
        "book_id": book_id,
        "discount": discount,
        "book": book
    }
`````

✅ **What is happening here**

* Path Validation → user_id, book_id must be > 0
* Query Validation → discount must be between 0 and 50
* Body Validation → book follows Pydantic rules

<br>

✅ **How to use**
* URL (Path + Query): http://127.0.0.1:8000/users/1/books/10?discount=20

* Request Body:
  ```
    {
      "title": "FastAPI Guide",
      "price": 500
    }
  ```
* Invalid Cases:
  * user_id = -1 → invalid path
  * discount = 100 → invalid query
  * "price": -10 → invalid body
  * 👉 FastAPI will reject the request automatically

<br>

🧠 **Key Understanding**

* Validation works independently but together
* FastAPI validates:
  * Path → first
  * Query → next
  * Body → last
* If any validation fails → API does not execute

>“FastAPI validates input at every level — URL, query, and request body.”

# Error Handling & Status Codes

APIs must handle errors properly and use appropriate status codes to communicate results clearly

**Types of Errors in FastAPI**

* **Validation Errors** → handled automatically by FastAPI
* **Custom Errors** → handled manually using `HTTPException`



## 📌 **Automatic Validation Errors**

FastAPI automatically validates input using Pydantic.


**What happens when validation fails?**

- Request is rejected ❌  
- API function does NOT execute  
- Returns error response with details ✅  
- Status code → **422 (Unprocessable Entity)**  

<br>

**Example Error Response**

```json
{
  "detail": [
    {
      "loc": ["body", "title"],
      "msg": "String should have at least 3 characters",
      "type": "string_too_short"
    }
  ]
}
```

<br>

🧠 **Understanding the Error**

* loc → where the error occurred (body, path, query)
* msg → description of the error
* type → type of validation failure


`````python
from fastapi import FastAPI
from pydantic import BaseModel, Field

app = FastAPI()


class BookRequest(BaseModel):
    title: str = Field(min_length=3)
    rating: int = Field(gt=0, lt=6)


@app.post("/books")
async def create_book(book: BookRequest):
    return {
        "message": "Book created successfully",
        "data": book
    }
`````

**How to Use**
* Valid Request:
  ```
  {
    "title": "FastAPI Guide",
    "rating": 5
  }
  ```
* Invalid Request:
  ```
  {
    "title": "AI",

  }
  ```
* Automatic Error Response
  ```
  {
      "detail": [
        {
          "loc": ["body", "title"],
          "msg": "String should have at least 3 characters"
        },
        {
          "loc": ["body", "rating"],
          "msg": "Input should be less than 6"
        }
      ]
    }
  ```

🧠 **What is happening?**
* Input does not meet validation rules
* FastAPI does not call the function
* Automatically returns error response

## 📌 Custom Error Handling

Use `HTTPException` to handle business logic errors

`````python
from fastapi import FastAPI, HTTPException

app = FastAPI()

@app.get("/books/{book_id}")
async def get_book(book_id: int):

    if book_id < 1:
        raise HTTPException(
            status_code=400,
            detail="Book ID must be greater than 0"
        )

    return {"book_id": book_id}
`````

**What this does**
* Raises custom error
* Stops function execution
* Returns error response immediately

**Error Response**
  ```
    {
      "detail": "Book ID must be greater than 0"
    }
  ```

## 📌 Status Codes

Status codes indicate whether a request was successful or failed.

<br>

**Status Code Categories**

HTTP status codes are grouped into categories based on their meaning.

| Category       | Code Range | Meaning                               | Common Examples |
|----------------|-----------|----------------------------------------|-----------------|
| Informational  | 1xx       | Request received, processing continues | 100 Continue |
| Success        | 2xx       | Request completed successfully ✅       | 200 OK <br> 201 Created <br> 204 No Content |
| Redirection    | 3xx       | Further action needed (redirect)       | 301 Moved Permanently <br> 302 Found |
| Client Error   | 4xx       | Error in client request ❌              | 400 Bad Request <br> 401 Unauthorized <br> 403 Forbidden <br> 404 Not Found <br> 422 Validation Error |
| Server Error   | 5xx       | Error on server side ❌                 | 500 Internal Server Error <br> 503 Service Unavailable |

<br>


💡 **Important Note**

- ✅ 2xx → Success responses  
- ✅ 4xx → Client-side errors  
- ✅ 5xx → Server-side errors



---

**Success Responses (Status Codes)**

`````python
from fastapi import FastAPI, status

app = FastAPI()

# ✅ CREATE (201 Created)
@app.post("/books", status_code=status.HTTP_201_CREATED)
async def create_book():
    return {"message": "Book created successfully"}


# ✅ UPDATE (200 OK)
@app.put("/books/{book_id}", status_code=status.HTTP_200_OK)
async def update_book(book_id: int):
    return {
        "message": "Book updated successfully",
        "book_id": book_id
    }


# ✅ DELETE (204 No Content)
@app.delete("/books/{book_id}", status_code=status.HTTP_204_NO_CONTENT)
async def delete_book(book_id: int):
    return
`````

---
**Custom Error with Status Code**

`````python
from fastapi import HTTPException

# ❌ Invalid Input (400)
@app.get("/books/{book_id}")
async def get_book_invalid(book_id: int):

    if book_id < 1:
        raise HTTPException(
            status_code=status.HTTP_400_BAD_REQUEST,
            detail="Invalid book ID"
        )

    return {"book_id": book_id}


# ❌ Not Found (404)
@app.get("/books/notfound/{book_id}")
async def get_book_not_found(book_id: int):

    if book_id != 1:
        raise HTTPException(
            status_code=status.HTTP_404_NOT_FOUND,
            detail="Book not found"
        )

    return {"book_id": book_id}
`````

---
**Success + Error Together (Real API Behavior)**

`````python
from fastapi import HTTPException, status

@app.get("/books/check/{book_id}", status_code=status.HTTP_200_OK)
async def get_book(book_id: int):

    # ❌ Error case
    if book_id < 1:
        raise HTTPException(
            status_code=status.HTTP_400_BAD_REQUEST,
            detail="Book ID must be greater than 0"
        )

    # ✅ Success case
    return {
        "message": "Book fetched successfully",
        "book_id": book_id
    }
`````

🧠 **Key Understandingr**

- Success responses use `status_code`
- Errors are raised using `HTTPException`
- Same API can return success or error based on input
- `status_code` → tells **what happened** (success or failure)  
- `HTTPException` → explains **why it failed**  


---


⚠️ **Important Notes**

- `status_code` is used for successful responses  
- Errors should use `HTTPException`  
- 204 No Content → should NOT return any body  
- FastAPI automatically returns **422** for validation errors

# Response Model

A **response model** is used to control the structure of data returned by an API.

👉 It ensures that the API always returns **clean, consistent, and validated output**.

**API Flow**

Request → Validation → Error Handling → Response ✅

**Why Response Models are Important**

- Ensures consistent API responses  
- Removes unwanted fields from output  
- Validates outgoing data  
- Improves API documentation (Swagger UI)  


## 📌 Basic Example

`````python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()


class BookResponse(BaseModel):
    title: str
    author: str
    rating: int


@app.get("/books", response_model=BookResponse)
async def get_book():
    return {
        "title": "FastAPI Guide",
        "author": "Eric",
        "rating": 5
    }
`````

✅ **Output**

```
{
  "title": "FastAPI Guide",
  "author": "Eric",
  "rating": 5
}
```

**What happens here**
* response_model=BookResponse → controls response structure
* FastAPI:
  * Validates response ✅
  * Filters extra fields ✅
  * Formats data automatically ✅


## 📌 Example (Extra Field Removed)


`````python
@app.get("/books", response_model=BookResponse)
async def get_book():
    return {
        "title": "FastAPI Guide",
        "author": "Eric",
        "rating": 5,
        "extra_field": "This will be removed"
    }
`````

✅ **Output (Filtered)**
  ```
  {
    "title": "FastAPI Guide",
    "author": "Eric",
    "rating": 5
  }
  ```

## 📌 Using Response Model with POST

`````python
from fastapi import status

@app.post("/books", response_model=BookResponse, status_code=status.HTTP_201_CREATED)
async def create_book():
    return {
        "title": "New Book",
        "author": "John",
        "rating": 4
    }
`````


## 📌 Returning a List of Items

`````python
from typing import List

@app.get("/books/all", response_model=List[BookResponse])
async def get_books():
    return [
        {"title": "Book 1", "author": "A", "rating": 5},
        {"title": "Book 2", "author": "B", "rating": 4}
    ]
`````

🧠 **Key Understanding**

* response_model → controls output structure
* Supports:
  * Single object
  * List of objects

* Automatically filters unwanted data
* Keeps API responses consistent

<br>

💡 **Important Notes**

* Response model is optional but recommended
* Used with Pydantic models
* Improves API clarity and safety
* Helps frontend developers understand API output

>“Request model defines input, response model defines output.”

# Dependency Injection

Dependency Injection allows you to **reuse logic across multiple APIs** without repeating code.

👉 It helps keep your API **clean, modular, and maintainable**.



## ✅ Why use Dependencies?

- Avoid repeating code  
- Share common logic across APIs  
- Improve code readability  
- Useful for authentication, validation, database access  


## 📌 Basic Example

`````python

# Without dependencies, you may repeat the same logic in multiple APIs:
@app.get("/items")
async def get_items(q: str = None):
    return {"q": q}

@app.get("/products")
async def get_products(q: str = None):
    return {"q": q}
`````

### ✅ CREATE

**Insert Data**

👉 Used to **add new records** to the database.

👉 Typically used with **POST requests**




`````python
import models
from fastapi import HTTPException
from pydantic import BaseModel, Field

# ✅ Request Model (Data Validation)
class UserRequest(BaseModel):
    email: str = Field(min_length=5)
    username: str = Field(min_length=3)
    first_name: str = Field(min_length=1)
    last_name: str = Field(min_length=1)
    password: str = Field(min_length=6)
    role: str


@app.post("/users", status_code=status.HTTP_201_CREATED)
async def create_user(user: UserRequest, db: db_dependency):

    # ✅ Check if email already exists
    existing_user = db.query(models.Users).filter(
        models.Users.email == user.email
    ).first()

    if existing_user:
        raise HTTPException(
            status_code=status.HTTP_400_BAD_REQUEST,
            detail="Email already registered"
        )

    # ✅ Create new user
    new_user = models.Users(
        email=user.email,
        username=user.username,
        first_name=user.first_name,
        last_name=user.last_name,
        hashed_password=user.password,
        role=user.role
    )

    db.add(new_user)      # Add new record
    db.commit()           # Save to database
    db.refresh(new_user)  # Refresh to get updated fields (like auto-generated ID)



    return {
              "message": "User created successfully",
              "user_id": new_user.id,
              "username": new_user.username
            }
`````

## 📌 Password Hashing
👉 Never store plain passwords ❌

👉 Always store hashed passwords ✅

`````python
from passlib.context import CryptContext

pwd_context = CryptContext(schemes=["bcrypt"], deprecated="auto")


def hash_password(password: str):
    return pwd_context.hash(password)


def verify_password(plain_password: str, hashed_password: str):
    return pwd_context.verify(plain_password, hashed_password)
`````

## 📌 Update Create User (Hash Password)

`````python
import models
from typing import Annotated
from sqlalchemy.orm import Session
from fastapi import FastAPI, Depends, HTTPException, status
from pydantic import BaseModel, Field


# ✅ DB Dependency
db_dependency = Annotated[Session, Depends(get_db)]


# ✅ Request Model (Data Validation)
class UserRequest(BaseModel):
    email: str = Field(min_length=5)
    username: str = Field(min_length=3)
    first_name: str = Field(min_length=1)
    last_name: str = Field(min_length=1)
    password: str = Field(min_length=6)
    role: str


# ✅ Create User API with Password Hashing
@app.post("/users", status_code=status.HTTP_201_CREATED)
async def create_user(user: UserRequest, db: db_dependency):

    # ✅ Check if email already exists
    existing_user = db.query(models.Users).filter(
        models.Users.email == user.email
    ).first()

    if existing_user:
        raise HTTPException(
            status_code=status.HTTP_400_BAD_REQUEST,
            detail="Email already registered"
        )

    # ✅ Create new user
    new_user = models.Users(
        email=user.email,
        username=user.username,
        first_name=user.first_name,
        last_name=user.last_name,
        hashed_password=hash_password(user.password), # ✅ hashed password stored
        role=user.role
    )

    db.add(new_user)      # Add new record
    db.commit()           # Save to database
    db.refresh(new_user)  # Refresh to get updated fields (like auto-generated ID)



    return {
              "message": "User created successfully",
              "user_id": new_user.id,
              "username": new_user.username
            }
`````


🧠 **What is happening?**

1. Input is validated using Pydantic (`UserRequest`)
2. Database is checked for duplicate email
3. Password is hashed using bcrypt
4. User is saved to the database
5. Auto-generated ID is retrieved using `db.refresh()`
6. API returns user details

## 📌 Understanding JWT (JSON Web Token)

JWT (JSON Web Token) is a compact and secure way to transmit information between a client and a server.

👉 It is widely used for authentication and authorization in web applications.

👉 “JWT is a signed token that carries user identity and allows stateless and secure authentication.”


<br>

---

✅ **Structure of JWT**

* A JWT consists of three parts:
  ```
    HEADER.PAYLOAD.SIGNATURE
  ```
* Example Token:
  ```
    eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.
    eyJzdWIiOiJ0ZXN0dXNlciIsImV4cCI6MTY5OTk5OTk5OX0.
    abc123signature
  ```
* *Header*: 👉 Contains metadata about the token
  ```
    {
      "alg": "HS256",  # algorithm used (e.g., HS256)
      "typ": "JWT"     # token type
    }
  ```
*  *Payload*: 👉 Contains user data (called claims)

    ```
      {
        "sub": "testuser",   # subject (username / user id/ role )
        "exp": 1699999999    # expiration time
      }
    ```
* *Signature*: 👉 Ensures token integrity (not tampered).
  * Created using: `Encoded(Header) + Encoded(Payload) + SECRET_KEY`
  * If token is modified, signature becomes invalid ❌


<br>

---

🔐 **Important Properties**

* JWT is signed, not encrypted
* Data is encoded (base64) → readable but protected
* Token has expiration (exp)
* Stateless → no session storage required

<br>

---
🔁 **How JWT Works**
1. User logs in
2. Server verifies credentials
3. Server generates token
4. Client stores token
5. Client sends token in request
6. Server validates token
7. Access granted ✅

<br>

⚠️ **What NOT to store in JWT**
* ❌ Passwords
* ❌ Sensitive personal data
*👉 Because payload can be decoded easily

<br>

✅ **What SHOULD be stored**
* ✔ username
* ✔ user_id
* ✔ role (for authorization)

## 📌 Login API with JWT Token

👉 This API verifies user credentials and generates a JWT token.


`````python
import models
from jose import jwt
from typing import Annotated
from pydantic import BaseModel
from datetime import datetime, timedelta
from fastapi import HTTPException, status, Depends
from fastapi.security import OAuth2PasswordRequestForm


# ✅ JWT Configuration
SECRET_KEY = "mysecretkey"
ALGORITHM = "HS256"
ACCESS_TOKEN_EXPIRE_MINUTES = 30


# ✅ Response Model
class TokenResponse(BaseModel):
    access_token: str
    token_type: str


# ✅ Create JWT Token
def create_access_token(data: dict):
    to_encode = data.copy()

    expire = datetime.utcnow() + timedelta(minutes=ACCESS_TOKEN_EXPIRE_MINUTES)
    to_encode.update({"exp": expire})

    return jwt.encode(to_encode, SECRET_KEY, algorithm=ALGORITHM)


# ✅ Login API
@app.post("/login",response_model=TokenResponse)
async def login(form_data:Annotated[OAuth2PasswordRequestForm , Depends()], db: db_dependency):

    # ✅ Step 1: Check if user exists
    user = db.query(models.Users).filter(models.Users.username == form_data.username).first()

    if not user:
        raise HTTPException( status_code=401, detail="User not found")


    # ✅ Step 2: Verify password
    if not verify_password(form_data.password, user.hashed_password):
        raise HTTPException( status_code=401, detail="Invalid password")

    # ✅ Step 3: Create JWT token
    access_token = create_access_token(data={"sub": user.username})

    # ✅ Step 4: Return token
    return { "access_token": access_token, "token_type": "bearer" }
`````

🧠 **What is happening?**

* User sends username & password
* API checks if user exists
* Password is verified using hashing
* JWT token is generated
* Token is returned to client

✅ **Example Response**

```
   {  
      "access_token":"eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9...",  
      "token_type": "bearer"
    }
```

## 📌 Verify Token
👉 Every protected API must validate the token.


`````python
from jose import jwt, JWTError
from fastapi.security import OAuth2PasswordBearer
from fastapi import Depends, HTTPException, status


oauth2_scheme = OAuth2PasswordBearer(tokenUrl="login")


def get_current_user(token: Annotated[str, Depends(oauth2_scheme)]):

    try:
        # ✅ Step 1: Decode token
        payload = jwt.decode(token, SECRET_KEY, algorithms=[ALGORITHM])

        # ✅ Step 2: Extract username
        username = payload.get("sub")

        if username is None:
            raise HTTPException(
                status_code=status.HTTP_401_UNAUTHORIZED,
                detail="Invalid token"
            )

        return username  # ✅ Valid user

    except JWTError:
        # ✅ Step 3: Handle invalid/expired token
        raise HTTPException(
            status_code=status.HTTP_401_UNAUTHORIZED,
            detail="Token is invalid or expired"
        )
`````

## 📌 Protect API with Token

After validating the token, we can use it to **protect APIs** so that only authenticated users can access them.

👉 A protected API requires a valid JWT token in the request.

✅ **How it works**

* Depends(get_current_user) → validates the token ✅
* Extracts username from token ✅
* Allows access only if token is valid ✅


`````python
@app.get("/secure-data")
async def secure_data(current_user: Annotated[str, Depends(get_current_user)]):
    return {
        "message": f"Hello {current_user}, this is protected data"
    }
`````

`````python
from fastapi import HTTPException, status

@app.delete("/users/{user_id}", status_code=status.HTTP_200_OK)
async def delete_user(
    user_id: int,
    current_user: Annotated[str, Depends(get_current_user)],  # ✅ Protected route
    db: db_dependency
):

    # ✅ Step 1: Get user from DB
    user = db.query(models.Users).filter(models.Users.id == user_id).first()

    # ✅ Step 2: Check if user exists
    if not user:
        raise HTTPException(
            status_code=status.HTTP_404_NOT_FOUND,
            detail="User not found"
        )

    # ✅ Step 3: Delete user
    db.delete(user)
    db.commit()

    return {"message": "User deleted successfully"}
`````

🔁 **Full Flow**
* User logs in → gets token
* User sends request with token
* oauth2_scheme extracts token
* get_current_user validates token
* If valid → API executes


🧠 **Key Understanding**
* Protected APIs require a valid token
* Depends(get_current_user) secures endpoints
* Token must be sent in Authorization header
* Authentication happens before API logic executes

>“Protected APIs use token validation to ensure that only authenticated users can access them.”
