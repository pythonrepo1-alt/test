# Introduction

**FastAPI** is a modern, high-performance Python web framework for building **RESTful APIs**. It is designed for speed, simplicity, and developer productivity, leveraging Python’s type hints for automatic validation and documentation.

<br>

## ✅ FastAPI is commonly used for

- Building scalable backend services  
- Creating machine learning / AI APIs  
- Developing microservices and data-driven applications  
- Prototyping APIs quickly for frontend/mobile teams  
- Creating high-performance async services


<br>

## 🚀 Key Features

- ⚡ High performance — comparable to Node.js and Go  
- ✅ Automatic validation using Python type hints  
- 📄 Auto-generated API docs (Swagger UI & OpenAPI)  
- 🔁 Async support using `async` / `await`  
- 🧩 Easy integration with modern tools like `uv`, `Uvicorn`, and `Pydantic`  

<br>

## 🌐 REST API Basics

A **REST API (Representational State Transfer API)** allows different systems to communicate over HTTP using standard methods and URLs.

<br>

### 🔑 **6 Core Principles of REST APIs**

- **Stateless** — Each request contains all required information; the server does not store client state  
- **Client–Server Architecture** — The client and server are independent, enabling separate development and scaling  
- **Resource-Based** — Everything is treated as a resource identified by URLs (e.g., `/users`, `/orders`)  
- **Uniform Interface** — Standard HTTP methods (GET, POST, PUT, DELETE) are used consistently  
- **Representation** — Resources are transferred in a standard format, typically JSON  
- **Cacheable** — Responses can be cached to improve performance and reduce server load  

<br>

### 🔁 **How REST APIs Work**

* A REST API follows a request–response model, where a client sends a request to a server, and the server processes it and returns a response.

* This flow represents the complete lifecycle of a request from client to server and back in a RESTful system

  ```
    Client (Browser / Mobile App)
               ↓
     HTTP Request (GET / POST)
               ↓
    FastAPI Application (Uvicorn)
               ↓
    Validation & Business Logic
               ↓
    Database / External Services
               ↓
          JSON Response
               ↓
      Client (Consumes Data)
  ```


#   Installation & Setup
## 🧱 One-Time Setup
Run these once to initialize your project and dependencies.
* **`Step 1`**: Initialize the project

          uv init my-project
          cd my-project

     This initializes a new project and creates a `pyproject.toml` file.

<br>

* **`Step 2`**: Edit pyproject.toml to include dependencies like

          [project]
          name = "my-project"
          version = "0.1.0"
          description = "Sample FastAPI project using uv"
          requires-python = ">=3.12"
          dependencies = [
                          "rich==14.0.0",
                          "fastapi==0.136.1",
                          "uvicorn==0.30.0"
                        ]
    What this does:

    * Defines project metadata
    * Declares required Python version
    * Lists all dependencies with fixed versions (best for reproducibility)

<br>

* **`Step 3`**: Create a virtual environment

          uv venv

    This creates a `.venv` directory inside your project.

    💡**Note**: You do not need to manually activate the virtual environment. using `uv run <command>` uv automatically executes the command inside the `.venv` environment.

<br>

* **`Step 4`**: Install dependencies and generate lock file

          uv sync

    What happens here:
    * Reads dependencies from `pyproject.toml`
    * Resolves all transitive dependencies
    * Creates a `uv.lock` file 🔒
    * Installs all packages into `.venv`

<br>

* **`Step 5`**: Export to requirements.txt (optional)

          uv pip freeze > requirements.txt

    Use this when:
    * Working with legacy tools
    * Sharing dependencies with systems that require requirements.txt

# CRUD Operations

**CRUD** stands for:

- **Create** → Add new data  
- **Read** → Retrieve existing data  
- **Update** → Modify existing data  
- **Delete** → Remove data  

These operations map directly to HTTP methods:

| CRUD Operation | HTTP Method | Purpose | Example Endpoint |
|---------------|------------|------|------------------|
| Create        | POST       | Add data |/users           |
| Read          | GET        | Fetch data |/users/{id}      |
| Update        | PUT/PATCH  | Modify data |/users/{id}      |
| Delete        | DELETE     | Remove data |/users/{id}      |

---

##  📌 Basic HTTP Methods

````python
from fastapi import FastAPI

app = FastAPI()


# GET
@app.get("/api-endpoint1")
async def read_item():
    return {'message': 'Hello Eric!'}


# POST
@app.post("/api-endpoint2")
async def create_item():
    return {'message': 'Data created successfully'}


# PUT
@app.put("/api-endpoint3")
async def update_item():
    return {'message': 'Data updated successfully'}


# DELETE
@app.delete("/api-endpoint4")
async def delete_item():
    return {'message': 'Data deleted successfully'}
````

Run the server:
uv run uvicorn main:app --reload

✅ **View API**

- Direct endpoint (GET only):
  http://127.0.0.1:8000/api-endpoint

- Swagger UI (for all methods: GET, POST, PUT, DELETE):
  http://127.0.0.1:8000/docs

## 📌 Path Parameter
Path parameters allow you to pass dynamic values in the URL.

* Query parameters are optional
* Defined using = (default values)
* Passed using ?key=value
* Multiple params use &


## ⚠️ Important Rule

👉 **Always define static paths before dynamic paths**


````python
#static paths
@app.get("/items/special")
async def get_special_item():
    return {"message": "This is a special item"}

# dynamic paths
@app.get("/items/{item_id}")
async def get_item(item_id: int):
    return {"item_id": item_id}
````

🧠 **Key Idea**

* `{item_id}` is a dynamic value
* Value is taken from the URL
* FastAPI automatically converts it to the correct data type


## 📌 Query Parameters

Query parameters are used to send optional data in the URL.

They come after `?` in the URL.


````python
@app.get("/items")
async def get_items(limit: int = 10):
    return {"limit": limit}
````

✅ How to use
* Open in browser: http://127.0.0.1:8000/items?limit=5
* Swagger UI: http://127.0.0.1:8000/docs

```
Output:
    {

       "limit": 5
    }
```

 **Multiple Query Parameters**

 * How to use : http://127.0.0.1:8000/products?skip=2&limit=5

````python
@app.get("/products")
async def get_products(skip: int = 0, limit: int = 10):
    return {"skip": skip, "limit": limit}
````

**Path + Query Parameters Together**

You can use **both path and query parameters in the same API**.


````python
@app.get("/items/{item_id}")
async def get_item(item_id: int, limit: int = 10):
    return {
        "item_id": item_id,
        "limit": limit
    }
````

✅ **How to use**
* Open in browser:
http://127.0.0.1:8000/items/5?limit=2
* Swagger UI:
http://127.0.0.1:8000/docs

<br>

🧠 **Key Idea**
* item_id → Path parameter (required → identifies resource)
* limit → Query parameter (optional → modifies result)

> “Path parameters identify the resource, query parameters control how we fetch it.”

## 📌 Request Body
* Request body is used to send data to the API (mainly for POST and PUT).

* You can send **multiple values in the request body** using `Body()

* Data is usually sent in **JSON format**.

````python

from fastapi import FastAPI, Body

app = FastAPI()

data_store = {}


# ⚠️ READ (GET) — Body not recommended / Method Limitation

# - GET requests are used to **retrieve data**
# - ❌ They should NOT use request body
# - ✅ Use path or query parameters instead



# ✅ CREATE (POST)
@app.post("/items")
async def create_item(item: dict = Body()):
    data_store["item"] = item
    return {
        "message": "Item created",
        "data": item
    }



# ✅ UPDATE (PUT)
@app.put("/items")
async def update_item(item: dict = Body()):
    data_store["item"] = item
    return {
        "message": "Item updated",
        "data": item
    }



# ✅ DELETE (DELETE)
@app.delete("/items")
async def delete_item(confirm: dict = Body(...)):
    deleted = data_store.pop("item", None)
    return {
        "message": "Item deleted",
        "confirmation": confirm,
        "deleted_data": deleted
    }


#Multiple Request Body Parameters
@app.post("/items/multiple")
async def create_item(
    name: str = Body(),
    price: float = Body(),
    in_stock: bool = Body()
):
    return {
        "name": name,
        "price": price,
        "in_stock": in_stock
    }
````

**Combination Example**

````python
from fastapi import FastAPI, Body

app = FastAPI()


@app.post("/users/{user_id}/orders/{order_id}")
async def create_order(
    user_id: int,                     # ✅ Path parameter 1
    order_id: int,                    # ✅ Path parameter 2

    status: str = "new",              # ✅ Query parameter 1
    priority: bool = False,           # ✅ Query parameter 2

    item_name: str = Body(...),       # ✅ Body field 1
    price: float = Body(...),         # ✅ Body field 2
    quantity: int = Body(...)         # ✅ Body field 3
):
    return {
        "user_id": user_id,
        "order_id": order_id,
        "status": status,
        "priority": priority,
        "item": {
            "name": item_name,
            "price": price,
            "quantity": quantity
        }
    }
````

✅ **How to use**
* URL (Path + Query) : http://127.0.0.1:8000/users/1/orders/101?status=confirmed&priority=true
* Request Body:
  ```
    {
      "item_name": "Laptop",
      "price": 50000,
      "quantity": 2
    }
  ```
* Output:

  ```
    {
      "user_id": 1,
      "order_id": 101,
      "status": "confirmed",
      "priority": true,
      "item": {
        "name": "Laptop",
        "price": 50000,
        "quantity": 2
        }
    }
  ```

>“Path identifies who/what, query modifies how, and body contains actual data.”

>⚠️ When using multiple body fields, Body() is required — otherwise FastAPI treats them as query parameters
