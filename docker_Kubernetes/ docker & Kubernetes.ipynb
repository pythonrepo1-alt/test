# 🐳 Docker
👉 Docker is a tool for creating and managing containers

## 📘 Introduction

>**Docker** is a containerization technology that allows you to package an application along with its dependencies so that it can run consistently across different environments.

👉 In simple terms:
> **"It works on my machine" → Docker makes it work everywhere.**

---


## 📦 What is a Container?

A **container** is a *running instance* of an application that includes:

- A package of code + all required dependencies  
  *(e.g., Python app + Python runtime + libraries)*


- The **same container always produces the same result**, regardless of:
  - where it runs  
  - who runs it

- Modern operating systems **natively support containers**

- Docker simplifies: *Creating*,*Running*, *Managing* of such containers

- When you create a container (via `docker run` ), a thin read-write layer is added on top of the Image.
- Containers run in isolation, i.e. they don't share any application state or written data.
---

## 🧱 Key Concepts

These are the core building blocks of Docker:

* **Image**:
  - Static blueprint used to create containers.
  - Immutable (cannot be changed after creation).
  - 👉 Example: `nginx` image → creates nginx container



* **Dockerfile**:
  - A text file containing **set of instructions** used to build a Docker image  
  - Defines: *Base image (e.g., Python, Node)*, *Application code*, *Dependencies and setup steps*


* **Volume**
  - Provides **persistent storage** for containers  
  - Ensures data is **not lost** when containers stop or are removed  
  - 👉 Common use cases: *Databases*, *Logs*, *File sharing between containers*
  
* **Network**:

  - Enables **communication between containers**.
  - Each container gets its own IP inside the network   
  - Supports: *Container-to-container communication*, *Port mapping (host ↔ container)*.
  - 👉 Example: Backend container talks to database container via network

## 🚀 Why Docker?

- ✅ Eliminates environment mismatch issues
- ✅ Lightweight compared to virtual machines
- ✅ Fast startup time
- ✅ Easy to deploy and share applications
- ✅ Consistent development → testing → production

## 🏗️ Docker Architecture

- **Docker Client** → Runs docker commands (CLI)
- **Docker Daemon** → Manages containers and images
- **Docker Registry** → Stores images (Docker Hub)

👉 Flow:
User → Docker CLI → Docker Daemon → Container runs

## 🖥️ Virtual Machines vs 🐳 Docker Containers
```
             🖥️ Virtual Machines                                                     🐳 Docker Container

       VM 1            VM 2            VM 3
        ↓               ↓               ↓                                   Container      Container       Container
+--------------+----------------+---------------+                              ↓               ↓               ↓                                           
|     APP A    |      APP B     |     APP C     |                       +--------------+----------------+---------------+                                         
|--------------|----------------|---------------|                       |     APP A    |      APP B     |     APP C     |
|   Libraries  |   Libraries,   |   Libraries,  |                       |--------------|----------------|---------------|
| Dependencies |  Dependencies  |  Dependencies |                       |  Libraries,  |   Libraries,   |  Libraries    |   
|--------------|----------------|---------------|                       | Dependencies |  Dependencies  | Dependencies  |
|  Virtual OS  |   Virtual OS   |  Virtual OS   |                       +--------------+----------------+---------------+                                   
| (e.g. Linux) |  (e.g. Linux)  | (e.g. Linux)  |                       +-----------------------------------------------+
+--------------+----------------+---------------+                       |                 Docker Engine                 |
+-----------------------------------------------+                       +-----------------------------------------------+           
|                  Hypervisor        	        |                       |   OS Built-in / Emulated Container Support 	|  
+-----------------------------------------------+                       +-----------------------------------------------+           
|             Your Operating System   	        |                       |              Your Operating System  	        |  
+-----------------------------------------------+                       +-----------------------------------------------+           
|                  Hardware  	                |                       |                   Hardware   	                |  
+-----------------------------------------------+                       +-----------------------------------------------+
```


| Feature            | Virtual Machine (VM)                          | Docker (Container)                     |
|------------------|----------------------------------------------|----------------------------------------|
| Architecture     | Runs on Hypervisor + Guest OS                 | Runs on Host OS (shares kernel)        |
| OS Requirement   | Each VM has its own full OS                   | No separate OS (lightweight)           |
| Size             | Heavy (GBs)                                  | Lightweight (MBs)                      |
| Startup Time     | Slow (minutes)                               | Fast (seconds)                         |
| Performance      | Slower (more overhead)                       | Faster (less overhead)                 |
| Resource Usage   | High (CPU, RAM)                              | Low                                    |
| Portability      | Moderate                                     | Very high                              |
| Isolation        | Strong (separate OS)                         | Process-level isolation                |
|Environment | Separate, environment-specific | Easily shared & reproducible |
|Storage | Redundant duplication (each VM has OS) | Minimal duplication |
|Distribution | Harder to share and rebuild | Easy to share and rebuild |
| Encapsulation | Whole machine | Application + dependencies only |
| Example          | VMware, VirtualBox                           | Docker                                 |

<br>

👉 **VM** = Application + Full OS (Heavy)

👉 **Docker** = Application + Dependencies (Lightweight)

✅ Docker provides better performance, scalability, and resource efficiency.

## 🧰 Docker Commands

📍 **Run in:** Terminal / Docker Desktop / Playground  



### 🔹 **Basic Commands**
  ```
      docker --version                 # Check installed version
      docker info                      # View system-wide Docker info
      
      docker --help                    # Show all Docker commands
      docker <command> --help          # Show help for a specific command (e.g., run, pull, build)
  ```

---
### 🔹 **Image Commands**
  ```    
      docker pull <image_name>:<tag>   # Download image from Docker Hub (e.g., python:3.10)
      docker push <image_name>         # Upload image to registry
      
      docker build .                   # Build image from Dockerfile in current directory (.)

      docker images                    # List all local images

      docker rmi <image_id>            # Remove an image

      docker search <image_name>       # Search images on Docker Hub (e.g., python)
  ```
  👉 . = current directory (where Dockerfile is located)
    

---
### 🔹 **Container Commands**
  ```
      docker run <image_name>                 # Create and start a container
      docker run --name <name> <image_name>   # Create & start container with a custom name
      
      docker start <container_id>             # Start an existing stopped container
      docker restart <container_id>           # Restart a container
      
      docker ps                               # List running containers
      docker ps -a                            # List all containers (including stopped)
      docker ps -q                            # Show only container IDs
      docker ps -f "status=running"           # Filter containers

      docker stop <container_id>              # Stop a container
      docker kill <container_id>              # Force stop container immediately
      docker rm <container_id>                # Remove a container
  ```
  👉 `<image_name>` = Docker image (e.g., nginx, python:3.10)

  👉 `<name>` = User-defined container name (used for easy identification)



✅ **Exec Into Container**
  ```
      docker exec -it <container_id> /bin/bash   # Access container shell (interactive terminal)
  ```
      
  👉 `-i` = interactive input (keeps input open)  
  👉 `-t` = terminal access (allocates a terminal)  
  👉 `/bin/bash` = open shell inside container  
  ⚠️ If `bash` is not available:
    `docker exec -it <container_id> /bin/sh`


---
### 🔹 **Debugging Commands**
  ```
      docker logs <container_id>        # View container logs (useful for debugging)
      docker logs -f <container_id>     # Follow logs in real-time
      docker inspect <container_id>     # Detailed container information
  ```

---
### 🔹 **Cleanup Commands**
  ```
      docker container prune            # Remove all stopped containers
      docker image prune                # Remove all dangling images (untagged images)
      docker image prune -a             # Remove all unused images
  ```



---
### 🔹 **Command Options**

👉 Options (also called *flags*) are used to modify how Docker commands behave  

👉 They are added along with commands (e.g., `docker run`, `docker exec`)


  ```
      -d        # Run container in background (detached mode)
      -it       # Run in interactive mode (terminal access)
      -p        # Map ports (host:container)
      --name    # Assign a name to container
      --rm      # Automatically remove container after stopping
  ```
💡 How Options Are Used
  ```
      docker run -d -p 8080:80 --name myapp --rm <image_name>
  ```
👉 Explanation:
  *  **`--name`** → Assign a custom NAME to the container (used for easy identification, stopping, removing).
  
  * **`-d`** → Run the container in **detached** mode (output not visible; terminal does not wait for it to stop)

  * **`-it`** → Run the container in **"interactive"** mode (allows input via terminal; can stop with `CTRL + C`)


  * `--rm` → Automatically remove the container when it's stopped
  * `-p` → Map ports between host and container (host_port:container_port)
    * Example: `docker run -p 8080:80 nginx`  
→ Access app at localhost:8080


  👉 Multiple options can be combined with a command

---
###👉 Important Rules


* ✅ docker exec → works only when container is running  
* ✅ docker start → works only on stopped containers  
* ✅ docker rm → requires container to be stopped    
* ❌ Cannot exec into a stopped container  
* ❌ Cannot start a removed container  


---
###✅ Quick Flow (Very Important)

```
Pull / Build → Run → Exec → Stop ⇄ Start → Remove
```

👉 Pull / Build → Get or create an image  
👉 Run           → Create and start a container  
👉 Exec          → Interact with the running container  
👉 Stop          → Stop the container  
👉 Start         → Restart the container  
👉 Remove        → Delete the container

##🧱 Dockerfile (Simple App)

👉 This example runs a simple Python application inside a container and prints output in the terminal

📁 Project Structure
```
simple-app/
├── Dockerfile
├── requirements.txt     (optional)
├── .dockerignore       # prevents unnecessary files from entering the image
└── src/
    └── app.py
```

👉 Prevents unnecessary and sensitive files (e.g., .git, __pycache__) from being copied into the image  
👉 Keeps image small, fast, and secure


### 🔹 Simple python App

`````
# src/app.py
print("Hello from Docker!")
`````

###🔹 Dockerfile

`````
# 1. Use official Python Base Image (Python installed)
FROM python:3.10

# 2. Set working directory inside container
WORKDIR /app

# 3. Copy files from local machine to container
# First "." → current folder on your system (where Dockerfile exists)
# Second "." → destination inside container (WORKDIR → /app)
COPY . .

# 4. Install dependencies (if requirements.txt exists)
RUN pip install -r requirements.txt

# 5. Command to run when container starts. It executes: python app.py inside container
CMD ["python", "src/app.py"]
`````

### 🔹 Run Commands
📍 **Run in**: Terminal / Docker Desktop / Playground
```
docker build -t myapp:1.0     # Build image (name: myapp, tag: 1.0)
```

```
docker run --rm myapp:1.0          # Create and run container from the image and auto-remove
````

###🔹 Access Application

```
 Output will be shown in terminal:
 Hello from Docker!
```

## 🚀 Dockerfile (FastAPI App)

👉 This example runs a FastAPI application inside a container and makes it accessible in the browser

📁 Project Structure
  ```
    fastapi-app/
    ├── Dockerfile
    ├── requirements.txt
    ├── .dockerignore         # prevents unnecessary files from entering the image
    └── src/
        └── main.py

  ```
  
👉 Prevents unnecessary and sensitive files (e.g., .git, __pycache__) from being copied into the image  
👉 Keeps image small, fast, and secure


### 🔹 **Simple FastAPI App**

`````
# src/main.py
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello from FastAPI inside Docker!"}
`````

###🔹 Dockerfile

`````
# Use official Python base image
FROM python:3.10

# Set working directory
WORKDIR /app

# Copy dependencies first (improves build caching)
COPY requirements.txt .

# Install dependencies
RUN pip install -r requirements.txt

# Copy source code
COPY src/ .

# Expose application port (for documentation only, actual access via -p  )
EXPOSE 8000

# Run FastAPI application
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
`````

### 🔹 Run Commands
📍 **Run in**: Terminal / Docker Desktop / Playground
```
  docker build -t fastapi-app:1.0               # Build image (name: fastapi-app, tag: 1.0)
```
```
  docker run --name fastapi-container -p 8000:8000 --rm fastapi-app:1.0   
```
* Create and run container from the image
  * **--name** → container name
  * **`-p 8000:8000`** → Map container port(8000) → host port (8000) so app is accessible in browser
  * **`--rm`**         → auto-remove container after it stops
  * **`fastapi-app:1.0`** → image used to create container

### 🔹 Access Application
```
  http://localhost:8000
  http://localhost:8000/docs
```
