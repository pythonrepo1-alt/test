# 🐳 Docker
👉 Docker is a tool for creating and managing containers

## 📘 Introduction

>**Docker** is a containerization technology that allows you to package an application along with its dependencies so that it can run consistently across different environments.

👉 In simple terms:
> **"It works on my machine" → Docker makes it work everywhere.**

---


## 📦 What is a Container?

A **container** is a *running instance* of an application that includes:

- A package of code + all required dependencies  
  *(e.g., Python app + Python runtime + libraries)*


- The **same container always produces the same result**, regardless of:
  - where it runs  
  - who runs it

- Modern operating systems **natively support containers**

- Docker simplifies: *Creating*,*Running*, *Managing* of such containers

- When you create a container (via `docker run` ), a thin read-write layer is added on top of the Image.
- Containers run in isolation, i.e. they don't share any application state or written data.
---

## 🧱 Key Concepts

These are the core building blocks of Docker:

* **Image**:
  - Static blueprint used to create containers.
  - Immutable (cannot be changed after creation).
  - Images are built in layers — each Dockerfile instruction creates a layer,
  - which enables caching for faster builds and allows layers to be reused
  - 👉 Example: `nginx` image → creates nginx container



* **Dockerfile**:
  - A text file containing **set of instructions** used to build a Docker image  
  - Defines: *Base image (e.g., Python, Node)*, *Application code*, *Dependencies and setup steps*


* **Volume**
  - Provides **persistent storage** for containers  
  - Ensures data is **not lost** when containers stop or are removed  
  - 👉 Common use cases: *Databases*, *Logs*, *File sharing between containers*
  
* **Network**:

  - Enables **communication between containers**.
  - Each container gets its own IP inside the network   
  - Supports: *Container-to-container communication*, *Port mapping (host ↔ container)*.
  - 👉 Example: Backend container talks to database container via network

- ✅ Eliminates environment mismatch issues
- ✅ Lightweight compared to virtual machines
- ✅ Fast startup time

| Example          | VMware, VirtualBox                           | Docker                                 |

  ```    
      docker pull <image_name>:<tag>   # Download image from Docker Hub (e.g., python:3.10)

      docker build .                   # Build image from Dockerfile in current directory (.)

      docker search <image_name>       # Search images on Docker Hub (e.g., python)
  ```
  👉 . = current directory (where Dockerfile is located)
    

---
### 🔹 **Container Commands**
  ```
      docker run <image_name>                 # Create and start a container
      docker run --name <name> <image_name>   # Create & start container with a custom name
      
      docker start <container_id>             # Start an existing stopped container
      docker restart <container_id>           # Restart a container
      
      docker ps                               # List running containers
      docker ps -a                            # List all containers (including stopped)
      docker ps -q                            # Show only container IDs
      docker ps -f "status=running"           # Filter containers

      docker stop <container_id>              # Stop a container
      docker kill <container_id>              # Force stop container immediately
      docker rm <container_id>                # Remove a container
  ```
  👉 `<image_name>` = Docker image (e.g., nginx, python:3.10)

  👉 `<name>` = User-defined container name (used for easy identification)

  ```
      
  👉 `-i` = interactive input (keeps input open)  
  👉 `-t` = terminal access (allocates a terminal)  
  👉 `/bin/bash` = open shell inside container  
  ⚠️ If `bash` is not available:
    `docker exec -it <container_id> /bin/sh`

### 🔹 **Command Options**

👉 Options (also called *flags*) are used to modify how Docker commands behave  

👉 They are added along with commands (e.g., `docker run`, `docker exec`)


  ```
      -d        # Run container in background (detached mode)
      -it       # Run in interactive mode (terminal access)
      -p        # Map ports (host:container)
      --name    # Assign a name to container
      --rm      # Automatically remove container after stopping
  ```
💡 How Options Are Used
  ```
      docker run -d -p 8080:80 --name myapp --rm <image_name>
  ```
👉 Explanation:
  *  **`--name`** → Assign a custom NAME to the container (used for easy identification, stopping, removing).
  
  * **`-d`** → Run the container in **detached** mode (output not visible; terminal does not wait for it to stop)

  * **`-it`** → Run the container in **"interactive"** mode (allows input via terminal; can stop with `CTRL + C`)


  * `--rm` → Automatically remove the container when it's stopped
  * `-p` → Map ports between host and container (host_port:container_port)
    * Example: `docker run -p 8080:80 nginx`  
→ Access app at localhost:8080


  👉 Multiple options can be combined with a command

---
###👉 Important Rules


* ✅ docker exec → works only when container is running  
* ✅ docker start → works only on stopped containers  
* ✅ docker rm → requires container to be stopped    
* ❌ Cannot exec into a stopped container  
* ❌ Cannot start a removed container  


---
###✅ Quick Flow (Very Important)

```
Pull / Build → Run → Exec → Stop ⇄ Start → Remove
```

👉 Pull / Build → Get or create an image  
👉 Run           → Create and start a container  
👉 Exec          → Interact with the running container  
👉 Stop          → Stop the container  
👉 Start         → Restart the container  
👉 Remove        → Delete the container

##🧱 Dockerfile (Simple App)

👉 This example runs a simple Python application inside a container and prints output in the terminal

📁 Project Structure
```
simple-app/
├── Dockerfile
├── requirements.txt     (optional)
├── .dockerignore       # prevents unnecessary files from entering the image
└── src/
    └── app.py
```

👉 Prevents unnecessary and sensitive files (e.g., .git, __pycache__) from being copied into the image  
👉 Keeps image small, fast, and secure


### 🔹 Simple python App

`````
# src/app.py
print("Hello from Docker!")
`````

###🔹 Dockerfile

`````
# 1. Use official Python Base Image (Python installed)
FROM python:3.10

# 2. Set working directory inside container
WORKDIR /app

# 3. Copy dependencies first (improves build caching)
COPY requirements.txt .

# 4. Install dependencies (if requirements.txt exists)
RUN pip install -r requirements.txt

# ❌ Avoid:
# COPY . . (copies unnecessary files → larger, slower builds)

# 5. Copy application source code
# Left → src/ (local project folder)
# Right → ./src/ (inside container → /app/src)
# Result: local/src → container:/app/src
COPY src/ ./src/


# 6. Command to run when container starts. It executes: python app.py inside container
CMD ["python", "src/app.py"]
`````

### 🔹 Run Commands
📍 **Run in**: Terminal / Docker Desktop / Playground
```
docker build -t myapp:1.0          # Build image (name: myapp, tag: 1.0)
```

```
docker run --rm myapp:1.0          # Create and run container from the image and auto-remove
````

###🔹 Access Application

```
 Output will be shown in terminal:
 Hello from Docker!
```

## 📦 Storage in Docker

Docker provides ways to **manage and persist data outside containers**.

---

### ⚠️ Why Storage is Needed ?

👉 Containers are **ephemeral (temporary)**

* **Problem 1**: Container is stopped and removed → all data written in the container is lost ❌  

  ✅ **Solution**: `Volumes`  
  👉 Store data outside the container so it persists even after removal  



👉 Containers are also **isolated from the host filesystem**

* **Problem 2**: Changes in local files → NOT reflected inside container ❌  

  ✅ **Solution**: `Bind Mounts`  
  👉 Link host files with the container for real-time updates  

---

### ✅ Storage Options

```
                     📦 Storage in Docker
                                │
            ┌───────────────────┴───────────────────┐
            │                                       │
 📦 Volumes (Docker-managed)            🔗 Bind Mounts (Host-managed)
            │                                        │
   ┌────────┴───────┐                         ┌──────┴──────┐
   │                │                         │             │
🔹 Anonymous   🔹 Named            🔹 Direct Mapping (Host ↔ Container)
   │                  │                          │
   │                  │                          │
❌ No name           ✅ Has name               ✅ Uses host path
⚠️ Temporary         ✅ Persistent             ✅ Live sync (real-time)
❌ Hard reuse        ✅ Reusable               ✅ Full control
❌ Location unknown  ✅ Production use         ❌ Not for production
```

### 🧠 Key Insight

👉 Volumes solve → **data persistence** ✅  
👉 Bind Mounts solve → **live file syncing**✅

---
### 📦 Volumes (Docker-managed Storage)

👉 Volumes are used to **persist data outside the container lifecycle**

👉 Volumes are folders (and files) managed by Docker on the host machine which are connected to folders / files inside of a container.

👉 Even if container is removed → data remains ✅  

---

#### ✅ Types of Volumes
>Docker provides two types of volumes:
>1. **Anonymous Volume**
>>```
>>docker run -v /some/path/in/container myapp
>>```
>2. **Named Volume**    
>>```
>>docker run -v mydata:/some/path/in/container myapp
>>```


👉 Key difference:
* Anonymous Volume → no name, created automatically (temporary)
* Named Volume → user-defined name, reusable (persistent)

---
#### 🔹 **Anonymous Volume**
👉 Anonymous volumes are created **automatically by Docker (no name)**

<br>

👉 Created when you use:

  ```
    docker run -v /some/path/in/container myapp
  ```

👉 Since no name is provided, Docker assigns a random volume ID

---

👉 Behavior:  
* ✅ Data is stored outside the container
* ⚠️ Removed automatically when container is removed **using --rm**
* ⚠️ Hard to reuse (no name to reference)

---

👉 Key Idea:  
  ```
    Anonymous volume = temporary storage created automatically by Docker
  ```

---

#####🐳 Example with Dockerfile

📁 Project Structure
```
app/
├── Dockerfile
└── app.py
```
---

🧾 app.py

```
with open("/data/file.txt", "a") as f:
    f.write("Hello from anonymous volume\n")

print("Data written!")
```
---

🐳 Dockerfile
```
FROM python:3.10

WORKDIR /app

COPY app.py .

# Declare mount point (optional → creates anonymous volume if none provided)
VOLUME ["/data"]

CMD ["python", "app.py"]
````
---

▶️ Run Container

```
docker build -t myapp .
docker run --rm myapp
```

🔍 What Happens
* 1️⃣ Docker sees VOLUME ["/data"]
* 2️⃣ No volume provided at runtime
* 3️⃣ Docker automatically creates an **anonymous volume**
* 4️⃣ It mounts it to /data
  ```
  Anonymous volume (random ID) ↔ /data (container)
  ```
* 5️⃣ App writes data → stored in volume ✅
* 6️⃣ Container stops → volume removed (because --rm) ❌

---

🔍 Verify Anonymous Volume  

* **List volumes**:
  ```
  docker volume ls


  👉 Example output:

      DRIVER    VOLUME NAME
      local     3f8c9a7d9b8c6e2a...
  ```
  👉 That random name = anonymous volume ✅

* **Inspect volume**:

  ```
  docker volume inspect <volume_id>

  👉 Example output:
      Mountpoint: /var/lib/docker/volumes/<volume_id>/_data
  ```
  👉 This is where data is stored on the host ✅

---

🧠 Flow Summary
* Dockerfile → defines mount point (/data)
* docker run → no volume given → Docker creates anonymous volume

---

✅ Final Takeaway   
* 👉 If you don’t explicitly define a volume, Docker will create an anonymous one automatically   
* 👉 Useful for temporary or internal storage, but not ideal for long-term or production use ❌



---
#### 🔹 Named Volume (Recommended ✅)

👉 Named volumes are created **with a specific name (user-defined)**  
👉 Persists until manually deleted (NOT
removed automatically)  
👉 Best for databases, logs, uploads

```
docker run -v mydata:/some/path/in/container myapp
```

👉 mydata → volume name   
👉 /some/path/in/container → container path

---

🔹 Behavior
* ✅ Data is stored outside the container   
* ✅ Persistent → survives container removal  
* ✅ Reusable across multiple containers  
* ❌ NOT removed automatically  

---
🔹 Key Idea
```
Named volume = persistent storage controlled by the user
```

### 🔗 Bind Mounts (host-managed Storage)

👉 Bind mounts are similar to volumes, but here **you (the developer) define the exact path on the host machine**

👉 Bind mounts connect a **host directory directly to a container path**
```
Host directory ↔ Container directory
```
---

### 🔹 Created when you use:
* 📍 Run in: Terminal / Docker Desktop / Playground
  ```
    docker run -v /absolute/path/on/host:/some/path/in/container myapp
  ```

  👉 Left →  host path (must be absolute)   
  👉 Right → container path
  

---

### 🔹 Behavior

* ✅ Real-time file sync (host ↔ container, changes reflect instantly)
* ✅ Changes on host → instantly reflected inside container
* ✅ Changes inside container → visible on host
* ✅ Full control over files (you know exact location)
* ❌ Depends on host filesystem
* ❌ Can break container isolation
* ❌ Not ideal for production

---

#### 🔹 **Key Idea**   

👉 During development, you frequently modify application files.   
👉 Without bind mounts, every change would require rebuilding and restarting the container ❌


👉 Bind mounts eliminate this overhead by syncing files in real time —  
👉 changes made on the host are instantly reflected inside the running container ✅

---

Container

  ```
* Create a **Named Volume** (named

    docker volume create VOL_NAME
  ```

  ```
    docker volume rm VOL_NAME
  ```
*  **Remove all unused Volumes** (i.e. not connected to a currently running or stopped container)
    ```
      docker volume prune
    ```

## 🌐 Docker Networking

👉 Multi-container applications are very common in real-world systems

👉 But by default, Docker containers are **isolated**, so they cannot communicate ❌

👉 Docker networking allows **containers to communicate with each other and with the outside world** ✅

👉 In real applications, Communication happens in three directions:

- Container ↔ Internet (world wide web)
- Container ↔ Host machine
- Container ↔ Container



---

### 🌐 Container ↔ Internet

👉 Containers can access the internet **by default**  
👉 This means a container can:

- Fetch APIs (e.g., call external services)  
- Download packages (pip, npm, apt, etc.)  
- Send requests to website


---
### 🔹 Example

👉 The container makes a request to the internet ✅
```
docker run python:3.10 curl https://api.github.com
```
---
### 🔹 **How It Works**

  👉 Docker connects containers to a default **bridge network**  
  👉 This network provides:

  - Internet access ✅
  - Outbound connections ✅


---

### 🔹 Key Idea
👉 Containers can access the internet out of the box, just like applications running on your system

👉 No extra setup is needed — containers can directly communicate with the internet ✅

---
### 🖥️ Container ↔ Host Machine

* Sometimes a container needs to communicate with services running on the host machine  
(e.g., a database or API running locally)

* However, by default `localhost` inside a container refers to the container itself ❌  

* This means a container **cannot access services running on the host using localhost** ❌  

* So this will NOT work inside a container:

  ```
  import requests
  
  requests.get("http://localhost:3000")
  ```
  👉 Because it tries to connect to the container, not the host ❌

*  **Solution**: Docker provides a special hostname:✅
    ```
     host.docker.internal
    ```
    👉 This allows the container to communicate with the host machine ✅

---

### 🔹 Example

👉 Run a simple server on your host machine:

```
python -m http.server 3000
```
👉 Python code inside the container:
```
import requests

response = requests.get("http://host.docker.internal:3000")


print("Status:", response.status_code)
print("Content preview:", response.text[:50])
```
👉 Now the container can communicate with the host ✅

---

### 🔹 How It Works
```
Container → host.docker.internal → Host machine ✅
```
👉 Docker automatically resolves `host.docker.internal`
to the IP address of the host machine

---

### 🔹 Key Idea
👉 Inside a container:
```
localhost → container ❌  
host.docker.internal → host machine ✅  
```
👉 This is mainly used during development (e.g., accessing a local DB)

### 🔗 Container ↔ Container (Bridge Network)

👉 In real applications, Containers often need to communicate with other Containers  

👉 By default, containers are isolated ❌  

👉 Docker solves this using a **Bridge Network**, which connects containers ✅  


---

### 🔹 Ways to Connect Containers

👉 There are two approaches:

  
* **Use Container IP** ❌ (Not Recommended)
  - You can manually find the IP of another container  
    ```
    docker inspect <container_id>
    ```
    👉 Then use it:

    ```
    import requests

    requests.get("http://172.17.0.2:3000")
    ```
    👉 Problems:
    - ❌ IP addresses can change  
    - ❌ Hard to manage, Not reliable  

* **Use Docker Bridge Network** ✅ (Recommended)
  - Create a network and attach containers:
    ```
    docker network create my-net

    docker run --network my-net --name cont1 my-image
    docker run --network my-net --name cont2 my-other-image
    ```
    👉 Both cont1 and cont2 will be in the same Network.  
    👉 Now containers can communicate using **container names**:
    ```
    import requests

    requests.get("http://cont1:3000")
    ```
    👉 Docker automatically resolves: `cont1 → container IP ✅`

  - **How It Works** `cont2 → cont1` ✅   
    👉 Bridge network provides:

    - Internal communication
    - Built-in DNS
    - Name-based resolution
---

###✅ Final Takeaway
👉 Bridge Network is the default and most common way to connect containers  
👉 It simplifies communication by using container names instead of IPs ✅   
👉 This is the standard approach used in real-world applications ✅

## 🌐 Docker Compose

👉 Docker Compose is a tool used to define and run **multi-container applications** using a single file  

👉 It can also be used for single-container setups to simplify building and running containers  

👉 Instead of running multiple `docker` commands, you define everything in one file:

```
docker-compose.yml
```

---
### 🔧 Docker Compose – Key Commands
👉 There are two key commands:

* **`docker compose up`** →  👉 Starts all containers/services defined in the Compose file ✅

  * **`-d`** →  Run containers in detached mode (background)
  * **`--build`** → Force rebuild of images before starting

* **`docker compose down`** → 👉 Stops and removes all containers and networks ✅
  * **`-v`** → Remove all Volumes used for the Containers - otherwise they stay around, even if the Containers are removed


---

👉 **Note:** These are the most commonly used commands for everyday workflow ✅  
👉 For additional commands, refer to the [official documentation](  
https://docs.docker.com/reference/cli/docker/compose/)


---
### 📄 docker-compose.yml Structure

👉 A Compose file is used to define the entire application setup  

👉 It mainly consists of three sections:

- **services** → 👉 Defines all containers (app, db)  
- **volumes** → 👉 Used for data persistence (pgdata)
- **networks** → 👉 Enables container communication (my-net)  
- **depends_on** → 👉 Ensures correct startup order (app depends on db)



`````
version: "3.9"                       # Compose file version (check with: docker compose version)

services:                            # Define all containers (services) like : "app", "db"
  app:                               # Service name → creates a container named "app"
    build:
      context: .                     # Path where Dockerfile is located
      dockerfile: Dockerfile-web     # Custom Dockerfile name (no trailing dot)
      args:                          # Build-time arguments (used during image build)
        some-arg: 1                  # Example: passed to Dockerfile using ARG
    ports:
      - "8000:8000"                  # Map host port 8000 → container port 8000
    volumes:                         # (Optional) Define any required volumes / bind mounts
      - logs:/app/logs               # Named volume → store logs persistently
      # - ....                       # can add more volumes storage example bind mounts
    networks:                        # (Optional but recommended)
      - my-net                       # Connect app to custom network
    depends_on:                      # ✅ Start order (app depends on db)
      - db                           # Ensures DB container starts fir


  db:                                   # Service name → creates a container named "db" (Database service)
    image: postgres                     # Use official PostgreSQL image (not using dockerfile)
    environment:                        # Environment variables for container configuration
      POSTGRES_PASSWORD: postgres       # Set DB password (required for Postgres)

      # Alternative syntax (equivalent)
      # - POSTGRES_PASSWORD=postgres    # optional way of doing

    # Alternative syntax (equivalent)
    # env_file:                 # (Optional) Load variables from external file
    #   - ./env/postgres.env    # Example: store secrets in a file (better practice)

    volumes:
      - pgdata:/var/lib/postgresql/data # Named volume → Persist database data
    networks:
      - my-net                          # Same network → allows app ↔ db communication


volumes:                     # Define named volumes
  logs:                      # Volume for application logs
  pgdata:                    # Volume for PostgreSQL data


networks:                    # Define custom networks
  my-net:                    # (Optional) Custom bridge network for container communication
`````

**services**

* Defines all containers in the application  
* Each service = one container   
* Example:
  * app → FastAPI container  
  * db  → PostgreSQL container


**volumes**  
* Used for persistent storage  
* Ensures data is not lost when containers   restart   
* Example:
  * pgdata → stores database data ✅

**networks**   
* Enables communication between containers   
* Containers in the same network can talk to each other  
* Example:
  * app → db ✅

---

**🧠 Key Idea**
* docker-compose.yml is a blueprint of your application
* It defines:
  * What to run → services ✅  
  * How to store data → volumes ✅  
  * How to connect → networks ✅

---

✅ **Final Takeaway**  
👉 Docker Compose simplifies multi-container applications   
👉 Replace:
* multiple commands ❌
With:
* 1 file + 1 command ✅

---

🧠 **Final Mental Model**

* docker-compose.yml → defines system ✅  
* docker compose up → runs system ✅  
* docker compose down → stops system ✅

### 🧠 Healthcheck + Readiness

👉 In multi-container applications, services may start in the correct order but still not be **ready to use** ❌  

👉 Example:
* App starts → tries to connect to DB
* DB is still initializing → connection fails ❌

#### 🔹 Problem with `depends_on`

👉 `depends_on` only controls **startup order**, NOT readiness  

```yaml
app:
  depends_on:
    - db
```
👉 This ensures: db starts first  
👉 But NOT: db is ready to accept connections ❌

---

🔹 Solution: Healthcheck ✅   
👉 A healthcheck allows Docker to monitor if a container is actually ready

`````
services:
  db:
    image: postgres

    healthcheck:
      test: ["CMD-SHELL", "pg_isready -U postgres"]
      interval: 5s
      timeout: 5s
      retries: 5
`````

🔹 How It Works   
👉 Docker runs the healthcheck command periodically   
👉 If the command succeeds: `container = healthy` ✅  
👉 If it fails repeatedly: `container = unhealthy ` ❌

🔹 Using with depends_on (Improved)

`````
app:
  depends_on:
    db:
      condition: service_healthy
`````

👉 Now:  
db must be healthy ✅  
then app starts ✅

🔹 Common Healthcheck Fields

`````
healthcheck:
  test: ["CMD-SHELL", "command"]   # Command to check health
  interval: 5s                     # Time between checks
  timeout: 5s                      # Time to wait for response
  retries: 5                       # Retries before marking unhealthy
`````

🔹 Real Example (FastAPI + PostgreSQL)

`````
services:
  db:
    image: postgres
    healthcheck:
      test: ["CMD-SHELL", "pg_isready -U postgres"]
      interval: 5s
      retries: 5

  app:
    depends_on:
      db:
        condition: service_healthy
`````

🔹 Alternative (Application-Level Retry)
👉 Sometimes apps implement retry logic:
```
# retry DB connection if failed
```
👉 Used together with healthcheck for better reliability ✅


🧠 Key Idea   
* 👉 depends_on → controls startup order ✅   
* 👉 healthcheck → ensures service is ready ✅

✅ Final Takeaway   
* 👉 Always use healthchecks for critical dependencies (DB, cache, etc.) ✅   
* 👉 Prevents startup failures and improves system robustness ✅  

🧠 Final Mental Model
* depends_on → order ✅  
* healthcheck → readiness ✅  
* Both together → reliable system ✅

#### ✅ Example 3 — Production System (Scaling + LB + Nginx + RDS + HTTPS)

👉 Goal: Make your application **production-ready** with multiple EC2 instances + load balancing, scaling, security, and managed database

#### 🧠 Key Components
* **Scaling (Auto Scaling)**
  * 👉 Ability to **increase or decrease number of servers (EC2) automatically**
  * 👉 Example:
    * Low traffic → 1 EC2  
    * High traffic → 5 EC2 ✅
  * Prevents overload  ✅
  * Improves availability ✅
  * Optimizes cost ✅
* **Load Balancer (ALB)**
  * 👉 Traffic manager: Distributes incoming traffic across multiple EC2 instances
  * 👉 Takes incoming requests and distributes:
    ```
              🌐 User Traffic
                      ↓
        ┌──────────────────────────────┐
        │     Load Balancer (ALB)      │
        └──────────────────────────────┘
             ↓         ↓          ↓
        ┌────────┐ ┌────────┐ ┌────────┐
        │  EC2-1 │ │  EC2-2 │ │  EC2-3 │
        │  app   │ │  app   │ │  app   │
        └────────┘ └────────┘ └────────┘
    ```
  * Ensures no single server is overloaded ✅
  * Provides high availability ✅
  * Performs health checks ✅

* **Reverse Proxy (Nginx)**   
  💡 **Reverse Proxy**: A reverse proxy sits between user and application
  * It receives requests and forwards them to the backend ✅
    ```
    User → ALB → Reverse Proxy(Nginx) → App ✅
    ```
  * A reverse proxy is used to provide a single entry point that routes client requests to backend services while hiding internal details.

  * It improves usability (no ports), security, and allows centralized control over traffic.   

  💡 **Nginx**: Nginx is a web server that acts as a reverse proxy

  * Accepts incoming requests ✅
  * Forwards to backend services ✅
  * Handles routing, headers, caching ✅
  * Improves security and usability ✅
  * Acts as the entry point ✅
  
* **RDS (Managed Database)**
  * Replace the containerized database with a managed database (RDS)
  * 👉 Benefits:
    * No DB management ✅
    * Automatic backups ✅
    * High availability ✅
    * Scalable ✅
    
* **HTTPS (Security) + Domain**
  * 👉 Make your application secure and accessible via domain name
    ```
    example.com ✅
    instead of http://123.45.67.89 ❌
    ```
  * 👉 Encrypts communication between user and server
    ```
    HTTP ❌ → insecure
    HTTPS ✅ → secure (encrypted)
    ```
  * Protects user data   
  * Required for production systems ✅  
  * Enabled via AWS ACM (SSL certificates) ✅

#####🔹 **Step 4**: Create GitHub Actions Workflow
>👉 Automate build and push of Docker image ✅  
>📄 .github/workflows/deploy.yml

`````
name: CI/CD Pipeline

on:
  push:
    branches:
      - main

jobs:
  build-deploy:
    runs-on: ubuntu-latest

    steps:
      - name: Checkout
        uses: actions/checkout@v3

      - name: Login to AWS
        uses: aws-actions/configure-aws-credentials@v2
        with:
          aws-access-key-id: ${{ secrets.AWS_ACCESS_KEY_ID }}
          aws-secret-access-key: ${{ secrets.AWS_SECRET_ACCESS_KEY }}
          region: ${{ secrets.AWS_REGION }}

      - name: Login to ECR
        run: |
          aws ecr get-login-password --region ${{ secrets.AWS_REGION }} \
          | docker login --username AWS --password-stdin ${{ secrets.ECR_URI }}

      - name: Build Image
        run: docker build -t fastapi-app .

      - name: Tag Image
        run: docker tag fastapi-app:latest ${{ secrets.ECR_URI }}:latest

      - name: Push Image
        run: docker push ${{ secrets.ECR_URI }}:latest

      - name: Deploy to EC2
        uses: appleboy/ssh-action@v1.0.3
        with:
          host: ${{ secrets.EC2_HOST }}
          username: ${{ secrets.EC2_USER }}
          key: ${{ secrets.EC2_SSH_KEY }}
          script: |
            docker pull ${{ secrets.ECR_URI }}:latest
            docker compose up -d
`````
