# 📘 Introduction to Fine-Tuning
**Fine-Tuning**: Fine-tuning is the process of adapting a pre-trained LLM to a specific task using task-specific data.

👉 In simple terms:
* Pretraining → learns general language patterns
* Fine-tuning → learns task-specific behavior

---

## 🧠 Key Dimensions of Training

LLM training can be understood across three core dimensions:
* WHAT → Learning objective (SFT, RLHF, Pretraining)
* HOW → Training method (Full FT, LoRA)
* WHERE → Platform (HuggingFace, OpenAI, Bedrock)

---

## 📘 Types of Fine-Tuning

Fine-tuning adapts a pretrained model for specific tasks.

### ✅ 1. Supervised Fine-Tuning (SFT)
- Learn from labeled input-output data  
- Example: Q&A, summarization  


### ✅ 2. Instruction Tuning
- Specialized form of SFT  
- Learn to follow instructions  


### ✅ 3. Domain Adaptation
- Learn domain-specific knowledge  
- Example: medical, legal  


### ⚙️ Training Approaches (HOW)

- Fine-tuning can be performed using:

  - Full Fine-Tuning → update all model parameters  
  - **Parameter-Efficient Methods** → LoRA, QLoRA, Adapters

👉 In this notebook, we focus on:
```
Supervised Fine-Tuning (SFT)
→ using different approaches (Full FT, PEFT)
→ across platforms (HuggingFace, OpenAI, Bedrock)
```

A typical LLM training pipeline combines multiple stages:

1. Start with a **pretrained model**  
2. Apply **Supervised Fine-Tuning (SFT)**

4. Apply **efficient training methods (LoRA / QLoRA)** if needed  
5. (Optional) Apply **alignment methods (RLHF / DPO)**

# 📘 Supervised Fine-Tuning (SFT)

Supervised Fine-Tuning (SFT) is a training approach where a pre-trained model is trained further using labeled input-output pairs.

👉 The model learns to generate the correct output given an input.


👉 The goal is to teach the model how to generate correct responses for specific tasks.


👉 Internally, the model learns to predict the next token step-by-step, assigning higher probability to correct sequences of tokens.

👉 A pretrained model understands language, but SFT trains it to generate **correct, consistent, and task-oriented outputs** instead of generic responses.

<br>

---

## 🧠 Intuition

A pretrained model already understands language.

👉 SFT teaches the model:
"What kind of output should I generate for this input?"

**Example 1**: Question Answering
  * *`Input`*: What is machine learning?
  * *`Output`*: Machine learning is a method where systems learn patterns from data.

**Example 2**: Instruction Following
  * *`Instruction`*: Explain gravity in simple terms
  * *`Response`*: Gravity is a force that pulls objects toward each other.

**Example 3**: Summarization
  * *`Input`*: [Long paragraph]
  * *`Output`*: Short summary of the paragraph

<br>

---

📊 **Data Format**:
SFT requires **structured datasets**.
Common Formats:
* 1️⃣ Simple Q&A : `Question → Answer`

* 2️⃣ Instruction Format (most common):     `Instruction → Response`

* 3️⃣ Input + Context + Output : `Context + Question → Answer`

  Typical JSON Dataset Format

    ```json
    {
      "instruction": "Explain AI",
      "input": "",
      "output": "AI stands for Artificial Intelligence..."
    }


<br>

---

⚙️ **Training Approaches in SFT**

SFT defines what the model learns, but how weights are updated depends on the method used:

* Full Fine-Tuning

  → updates all model parameters

* Parameter-Efficient Fine-Tuning (LoRA / QLoRA)

  → keeps base model frozen and updates small adapter layers

👉 In practice, SFT is usually combined with LoRA.

<br>

---


🔗 **Where SFT Fits**

Typical training pipeline:

    Pretraining → SFT → Instruction Tuning → (Optional) RLHF

  👉 SFT is the foundational step in most LLM systems


<br>

---

✅ **Where SFT is Used**

* Chatbots
* Q&A systems
* Summarization
* Domain-specific assistants


<br>

---

✅ **Advantages of SFT**

* Simple and easy to implement
* Works well with small datasets
* Directly improves task performance
* Foundation for all advanced methods

<br>

---

❌ **Limitations of SFT**

* Requires labeled data
* Sensitive to data quality
* Can overfit small datasets
* Does not capture human preferences well

<br>

---

⚠️ **Common Mistakes**
* Poor data formatting
* Inconsistent instruction style
* Not enough data variety
* Training on noisy data
* Ignoring evaluation

<br>

---

🔍 **Relationship with Other Methods**

* Instruction Tuning = Special form of SFT
* Domain Adaptation = SFT on domain data
* RLHF = Improves SFT outputs
* LoRA = Efficient way to perform SFT

## 🔗 Note on Training Approaches

### Step 1: Load Dataset


````python

from datasets import load_dataset

# Load Dataset (we take 300 samples ) so we can split into:train )
dataset = load_dataset("tatsu-lab/alpaca", split="train[:300]")
print(dataset)

# we can split into: train 80%, test 10%, val 10%
train_test = dataset.train_test_split(test_size=0.2, seed=42)

train_dataset = train_test["train"]
temp_dataset = train_test["test"]

# Split Temp → Validation + Test
val_test = temp_dataset.train_test_split(test_size=0.5, seed=42)

val_dataset = val_test["train"]
test_dataset = val_test["test"]

# Check Sizes
print()
print("Train size:", len(train_dataset))
print("Validation size:", len(val_dataset))
print("Test size:", len(test_dataset))
````

### Step 2: Data Pre-processing
* Collect input-output pairs
* Clean and structure data
* Format into prompt style

````python
def format_example(example):
    return {
        "text": f"Instruction: {example['instruction']}\nResponse: {example['output']}"
    }

train_dataset = train_dataset.map(format_example)
val_dataset   = val_dataset.map(format_example)
test_dataset  = test_dataset.map(format_example)
````

### Step 3: Tokenization

Convert text into tokens
Example:

"Hello world" → [15496, 995]

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token


def tokenize(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset = val_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)


train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask"])
val_dataset.set_format(type="torch", columns=["input_ids", "attention_mask"])
````

### Step 4: Load model

````python

from transformers import AutoModelForCausalLM

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = AutoModelForCausalLM.from_pretrained("gpt2").to(device)
````

### Step 5: Training Loop

Model learns by minimizing prediction error

````python
import sys
import torch
import matplotlib.pyplot as plt

def progress_bar(step, total, epoch, total_epochs, train_loss, val_loss=0.0):
    percent = step / total
    bar_length = 20
    filled = int(bar_length * percent)
    bar = "#" * filled + "." * (bar_length - filled)

    sys.stdout.write(
        f"\rEpoch {epoch+1}/{total_epochs} "
        f"[{bar}] {percent*100:.0f}% "
        f"| Train Loss: {train_loss:.4f} "
        f"| Val Loss: {val_loss:.4f}"
    )
    sys.stdout.flush()


def plot_losses(train_losses, val_losses):

    epochs = range(1, len(train_losses) + 1)

    plt.figure(figsize=(8,5))

    plt.plot(epochs, train_losses, label="Train Loss", marker="o")
    plt.plot(epochs, val_losses, label="Validation Loss", marker="o")

    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")

    plt.legend()
    plt.grid(True)

    plt.show()


def train_model(model, train_loader, val_loader, optimizer, device, epochs):

    train_losses = []
    val_losses = []

    for epoch in range(epochs):

        # -------- TRAIN --------
        model.train()
        train_loss = 0

        for step, batch in enumerate(train_loader):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=input_ids
            )

            loss = outputs.loss
            train_loss += loss.item()

            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

            # ✅ Use external progress bar
            progress_bar(
                step + 1,
                len(train_loader),
                epoch,
                epochs,
                train_loss / (step + 1)
            )

        avg_train_loss = train_loss / len(train_loader)

        # -------- VALIDATION --------
        model.eval()
        val_loss = 0

        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch["input_ids"].to(device)
                attention_mask = batch["attention_mask"].to(device)

                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    labels=input_ids
                )

                val_loss += outputs.loss.item()

        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        print(
            f"\n✅ Epoch {epoch+1} Done | "
            f"Train Loss: {avg_train_loss:.4f} | "
            f"Val Loss: {avg_val_loss:.4f}"
        )

    return train_losses, val_losses
````

▶️ Run Training

````python

from torch.optim import AdamW
from torch.utils.data import DataLoader


batch_size =8
epoch=3
lr = 5e-5

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)

optimizer = AdamW(model.parameters(), lr=lr)

train_losses, val_losses = train_model(
    model,
    train_loader,
    val_loader,
    optimizer,
    device,
    epochs=epoch
)

plot_losses(train_losses, val_losses)
````

### Step 6: Evaluation

After training a model, we must evaluate:

👉 How well the model performs on unseen data

👉 Whether the model generates correct and useful outputs

<br>

---

🎯 **Types of Evaluation Metrics**
1. **Loss**
    - Measures prediction error  
    - Lower = better

2. **Perplexity**
    - Measures model confidence
    - Converts loss into an interpretable score
    - Measures how “confused” the model is
    - Formula: `Perplexity = exp(loss)`  
    - Lower = better

3. **ROUGE**
    - Measures overlap with reference text  
    - Used for summarization
    - Only measure text similarity
    - Do NOT fully capture meaning or reasoning
    - Higher = better  

4. **BLEU**
    - Measures n-gram overlap between prediction and reference
    - Used for translation-like tasks  
    - Only measure text similarity
    - Do NOT fully capture meaning or reasoning
    - Higher = better  

5. **Exact Match**
    - Checks if prediction exactly matches the reference
    - Good for short answers
    - Very strict metric

6. **Qualitative Evaluation**
    - Human inspection of model outputs
    - Check for: Relevance, Correctness, Structure, Instruction-following

🧠 **Mental Model**
```
    Loss → training quality
    Perplexity → confidence
    ROUGE/BLEU → similarity
    Exact Match → strict correctness
    Output → ✅ real performance
```

````python
import torch
from evaluate import load

# -----------------------------
# ✅ Load Metrics
# -----------------------------
rouge = load("rouge")
bleu = load("bleu")

# -----------------------------
# ✅ Metric Functions
# -----------------------------
def compute_perplexity(loss):
    return torch.exp(torch.tensor(loss)).item()

def compute_rouge(preds, refs):
    return rouge.compute(predictions=preds, references=refs)

def compute_bleu(preds, refs):
    return bleu.compute(predictions=preds, references=[[r] for r in refs])

def compute_exact_match(preds, refs):
    return sum(p.strip() == r.strip() for p, r in zip(preds, refs)) / len(preds)


# -----------------------------
# ✅ Generate Response
# -----------------------------
def generate_response(model, tokenizer, prompt, device):
    model.eval()

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    output = model.generate(
        inputs["input_ids"],
        max_length=100,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)


# -----------------------------
# ✅ Collect Predictions
# -----------------------------
def get_predictions(model, tokenizer, dataset, device, num_samples=50):
    preds = []
    refs = []

    for i in range(min(num_samples, len(dataset))):
        sample = dataset[i]

        prompt = f"Instruction: {sample['instruction']}\nResponse:"
        reference = sample["output"]

        prediction = generate_response(model, tokenizer, prompt, device)

        preds.append(prediction)
        refs.append(reference)

    return preds, refs


# -----------------------------
# ✅ Full Evaluation Function
# -----------------------------
def evaluate_all(model, tokenizer, test_loader, test_dataset, device):

    # -------- LOSS --------
    model.eval()
    test_loss = 0

    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=input_ids
            )

            test_loss += outputs.loss.item()

    avg_test_loss = test_loss / len(test_loader)

    # -------- PERPLEXITY --------
    ppl = compute_perplexity(avg_test_loss)

    # -------- GENERATE OUTPUT --------
    preds, refs = get_predictions(model, tokenizer, test_dataset, device)

    # -------- METRICS --------
    rouge_score = compute_rouge(preds, refs)
    bleu_score = compute_bleu(preds, refs)
    exact_match = compute_exact_match(preds, refs)

    # -------- PRINT --------
    print("\n✅ Evaluation Results")
    print("----------------------------")
    print(f"Test Loss: {avg_test_loss:.4f}")
    print(f"Perplexity: {ppl:.4f}")
    print(f"Exact Match: {exact_match:.4f}")
    print(f"ROUGE: {rouge_score}")
    print(f"BLEU: {bleu_score}")

    return preds, refs


# -----------------------------
# ✅ Show Sample Outputs
# -----------------------------
def show_samples(preds, refs, n=5):
    for i in range(min(n, len(preds))):
        print("\n----------------------------")
        print("Reference :", refs[i])
        print("Prediction:", preds[i])
````

▶️ Run Evaluation

````python
test_loader = DataLoader(test_dataset, batch_size=batch_size)

preds, refs = evaluate_all(
    model,
    tokenizer,
    test_loader,
    test_loader,
    device
)

show_samples(preds, refs)
````

# 📘 Parameter-Efficient Fine-Tuning (PEFT)

Parameter-Efficient Fine-Tuning (PEFT) is a set of techniques that enable fine-tuning large language models by updating only a **small subset of parameters** instead of the entire model.

👉 Instead of retraining all model weights:
- The base model is **frozen** ❄️  
- Only small additional components (e.g., LoRA, Adapters) are **trained** ✅  

---

## 🧠 Core Idea

- **Full Fine-Tuning** → update ALL parameters ❌  
  → expensive, slow, memory-intensive  

- **PEFT** → update a SMALL subset ✅  
  → efficient, fast, lightweight  

---
## 📘 LoRA / QLoRA

LoRA (Low-Rank Adaptation) and QLoRA are the most widely used PEFT techniques for efficient fine-tuning of large language models.

👉 They enable training large models:
- ✅ with limited memory  
- ✅ on smaller GPUs  
- ✅ at lower cost  
- ✅ with fewer trainable parameters  

👉 This is why most modern LLM fine-tuning uses **LoRA / QLoRA**

---

## 🧠 LoRA (Low-Rank Adaptation)
LoRA fine-tunes a model by learning **low-rank updates** instead of modifying full weight matrices.

* Instead of updating the full weight matrix **W** (original weights),
* LoRA learns a small correction:
  ```
  W + (A × B)

  Where:
  - W → frozen base weights (original weights)❄️  
  - A, B → small trainable matrices ✅
  ```
---

## ⚡ QLoRA (Quantized LoRA)
QLoRA extends LoRA by combining it with **quantization**, which reduces the memory footprint of the base model.

* Instead of loading the model in full precision:
  - The base model is loaded in **lower precision (e.g., 4-bit, 8-bit)** ❄️  
  - The model remains **frozen and compressed**
* Then, just like LoRA, it learns a small correction:
  ```
  W (quantized) + (A × B)

  Where:
  - W → compressed (low-precision) base weights ❄️  
  - A, B → small trainable matrices ✅
  ```

---

## ⚙️ Intuition
- **Full Fine-Tuning** → update entire matrix **W** ❌  
  → expensive, memory-heavy  

- **LoRA** → keep **W** frozen  
  → train only small adapters **(A × B)** ✅  

- **QLoRA** →  
  → compress model + apply LoRA ✅✅  

👉 The base model retains its knowledge, while:

- **LoRA** learns task-specific adjustments  
- **QLoRA** makes the model lightweight and applies the same adjustments  

👉 This enables fine-tuning **large models even on limited hardware**.

### Step 1: Load Dataset

````python
from datasets import load_dataset

# Load dataset (small subset for demo)
dataset = load_dataset("tatsu-lab/alpaca", split="train[:300]")

# Split into train (80%) and temp (20%)
train_test = dataset.train_test_split(test_size=0.2, seed=42)

train_dataset = train_test["train"]
temp_dataset = train_test["test"]

# Split temp into validation (10%) and test (10%)
val_test = temp_dataset.train_test_split(test_size=0.5, seed=42)

val_dataset = val_test["train"]
test_dataset = val_test["test"]

print("Train size:", len(train_dataset))
print("Validation size:", len(val_dataset))
print("Test size:", len(test_dataset))
print()
print(test_dataset)
````

### Step 2: Data Pre-processing
* Collect input-output pairs
* Clean and structure data
* Format into prompt style

````python
def format_example(example):
    return {
        "text": f"Instruction: {example['instruction']}\nResponse: {example['output']}"
    }



COL_TO_REMOVE =["instruction", "input", "output"]
train_dataset = train_dataset.map(format_example,remove_columns=COL_TO_REMOVE)
val_dataset   = val_dataset.map(format_example,remove_columns=COL_TO_REMOVE)
test_dataset  = test_dataset.map(format_example,remove_columns=COL_TO_REMOVE)
test_dataset[0]
````

### Step 3: Tokenization

Convert text into tokens
Example:

"Hello world" → [15496, 995]

````python
from transformers import AutoTokenizer

BASE_MODEL ='gpt2' #"meta-llama/Llama-3.2-3B" #"gpt2" #
# ✅ Load tokenizer for your model (must match model you train)
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL,use_fast=True)

# ✅ GPT-2 does not have a pad token, so we set it to EOS (end-of-sequence)
tokenizer.pad_token = tokenizer.eos_token


def tokenize(example):
    tokens =  tokenizer(
        example["text"],        # ✅ input text to tokenize
        truncation=True,        # ✅ cut text if longer than max_length
                                # prevents too long inputs → memory issue
        padding="max_length",   # ✅ pad shorter sequences to fixed size
                                # needed for batching (all inputs same length)
        max_length=512          # ✅ fixed sequence length
                                # every example will have exactly 512 tokens
    )
    tokens["labels"] = tokens["input_ids"].copy()

    return tokens

# ✅ Apply tokenization to entire dataset
train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset = val_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)


train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask","labels"])
val_dataset.set_format(type="torch", columns=["input_ids", "attention_mask","labels"])
val_dataset
````

### Step 4: Load Model (QLoRA-enabled)
We load the pretrained base model and optionally apply **quantization** to reduce memory usage.  

---

⚙️ **Important Hyperparameters**

    - Loads the model in **low precision (4-bit)**  → reduces memory ✅
    - Enables training large models on **smaller GPUs**  
    - 👉 Alternatives: `load_in_8bit=True`

<br>

2. `device_map="auto"`
    - Automatically distributes model layers across available devices GPU/CPU seamlessly.
    - Essential for handling large models  
    - Avoids manual device placement

<br>

3. `bnb_4bit_compute_dtype="float16"`
    - Defines computation precision during training
    - Controls how calculations are performed
    - Common values:
      - "float16" ✅ → faster, widely used ✅
      - "bfloat16" ✅ → more stable on newer GPUs

<br>

4. `bnb_4bit_quant_type="nf4"`
    - Defines the quantization type
    - "nf4" (Normal Float 4) is recommended ✅
    - Provides better performance than standard int4

<br>

5. `bnb_4bit_use_double_quant=True`

    - Applies **double quantization**
    - Further reduces memory usage ✅
    - Improves efficiency with minimal performance loss

<br>

6. `low_cpu_mem_usage=True` *(optional)*

    - Reduces CPU memory usage during model loading
    - Useful for very large models

<br>

7. `trust_remote_code=True` *(optional)*

    - Allows loading custom architectures from model repositories
    - Required for some advanced or newer models

<br>

---

📚 **Additional Resources**

👉 For more details on quantization and QLoRA setup:

- 🔗 Hugging Face Quantization Docs: https://huggingface.co/docs/transformers/main_classes/quantization

````python
from transformers import AutoModelForCausalLM, BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype="float16",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    quantization_config=bnb_config,
    device_map="auto"
)
````

### Step 5: Apply LoRA

We apply LoRA (Low-Rank Adaptation) to enable **parameter-efficient fine-tuning** by training only a small subset of parameters.

⚙️ **Important Hyperparameters**
1. `r=8` (rank)
    - Defines the size of the low-rank matrices (A and B)
    - Controls how much the model can adapt to new tasks
    - Typical values:
      - 4 → very lightweight (lower capacity)
      - 8 → balanced ✅
      - 16+ → more expressive (higher capacity)
<br>

2. `lora_alpha = 16` (scaling factor)
    - Controls the strength of LoRA updates
    - Scales the contribution of `(A × B)`
    - Rule of thumb: `lora_alpha ≈ 2 × r`
<br>

3. `target_modules=["c_attn"]` **(VERY IMPORTANT)**
    - Specifies which layers LoRA modifies
    - Critical for model performance ✅
    - Wrong choice → ineffective learning ❌
    - Model-specific choices:
      ```
      - GPT-2 :["c_attn"]
      - LLaMA : ["q_proj", "v_proj"]
      ```
<br>

4. `lora_dropout = 0.1` (lora_dropout)
    - Adds **dropout regularization** to LoRA layers
    - Helps prevent overfitting
    - Especially useful for small datasets

<br>

5. `bias = "none"` (bias)
    - Controls whether bias terms are trained
    - Usually kept as "none" for efficiency ✅
<br>

6. `task_type = "CAUSAL_LM"`
    - Defines the training objective
    - Required for text generation tasks

📚 **Additional Resources**
👉 For advanced configurations and full parameter details:

- 🔗 Hugging Face PEFT Docs: https://huggingface.co/docs/peft
- 🔗 LoRA Config Reference: https://huggingface.co/docs/peft/package_reference/lora


````python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["c_attn"],  # GPT-style attention
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"

)

model = get_peft_model(model, lora_config)
````

### Step 10 : Evaluate


---

## ✅ 1. Loss

- Measures how well the model predicts the next token  
- Computed during training and evaluation  

👉 Interpretation:
- Lower loss → better model ✅  
- High loss → model is making many mistakes  

---

#### ✅ 2. Perplexity

- Derived from loss:  `Perplexity = exp(loss)`
- Measures how **confident** the model is in its predictions  

👉 Interpretation:
- Lower perplexity → better ✅  
- High perplexity → model is “confused”  

---

#### ✅ 3. ROUGE (Recall-Oriented Metric)

- Measures overlap between generated text and reference text  
- Compares sequences of words  

👉 Common variants:
- ROUGE-1 → unigram overlap  
- ROUGE-2 → bigram overlap  
- ROUGE-L → longest common subsequence  

👉 Best for:
- summarization  
- content similarity  

✅ Higher → better  

---

#### ✅ 4. BLEU (Precision-Oriented Metric)

- Measures n-gram overlap between prediction and reference  
- Focuses more on precision than recall  

👉 Best for:
- translation tasks  
- structured outputs  

✅ Higher → better  

⚠️ Limitation:
- Does NOT capture meaning or reasoning  

---

#### ✅ 5. Exact Match

- Checks if prediction exactly matches the reference  

👉 Example: **Prediction = "AI is intelligence"
Reference  = "AI is intelligence" ✅ → Match**

👉 Use case:
- QA systems  
- short answers  

✅ Higher → better  

⚠️ Very strict metric  

---

#### ✅ 6. Token-Level Accuracy

- Measures how many tokens are predicted correctly  

👉 Computed as: `correct tokens / total tokens`

👉 Use case:
- low-level model performance  
- debugging training quality  

✅ Higher → better  

---

#### ✅ 7. Diversity Metrics (Distinct-n)

- Measures how diverse or repetitive the output is  

👉 Example:

      Distinct-1 → unique words
      Distinct-2 → unique word pairs

👉 Interpretation:
- Higher → more diverse ✅  
- Lower → repetitive model  

---

#### ✅ 8. Qualitative Evaluation (Most Important)

- Manual inspection of model outputs  

👉 Check:
- correctness  
- fluency  
- coherence  
- instruction-following  

---
## ✅ Summary

| Metric            | Measures                      |
|------------------|------------------------------|
| Loss             | Prediction error             |
| Perplexity       | Model confidence             |
| ROUGE            | Text overlap (recall)        |
| BLEU             | Text overlap (precision)     |
| Exact Match      | Strict correctness           |
| Token Accuracy   | Token-level prediction       |
| Diversity        | Output variation             |
| Qualitative Eval | Real-world performance ✅     |



````python
import torch
import math
from evaluate import load

# ✅ Load metrics
rouge = load("rouge")
bleu = load("bleu")

# -----------------------------
# ✅ Generate Predictions
# -----------------------------
def generate_predictions(model, tokenizer, dataset, num_samples=20):
    model.eval()

    preds = []
    refs = []

    for i in range(min(num_samples, len(dataset))):
        sample = dataset[i]

        # Extract prompt + reference
        full_text = sample["text"]
        prompt = full_text.split("Response:")[0] + "Response:"
        reference = full_text.split("Response:")[1].strip()

        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        with torch.no_grad():
            output = model.generate(
                **inputs,
                max_length=150,
                temperature=0.7
            )

        prediction = tokenizer.decode(output[0], skip_special_tokens=True)

        preds.append(prediction)
        refs.append(reference)

    return preds, refs


# -----------------------------
# ✅ Exact Match
# -----------------------------
def exact_match(preds, refs):
    return sum(p.strip() == r.strip() for p, r in zip(preds, refs)) / len(preds)


# -----------------------------
# ✅ Token Accuracy
# -----------------------------
def token_accuracy(model, dataset):
    model.eval()
    total = 0
    correct = 0

    for sample in dataset:
        input_ids = torch.tensor(sample["input_ids"]).unsqueeze(0).to(model.device)
        labels = torch.tensor(sample["labels"]).unsqueeze(0).to(model.device)

        with torch.no_grad():
            outputs = model(input_ids)

        preds = outputs.logits.argmax(dim=-1)

        correct += (preds == labels).sum().item()
        total += labels.numel()

    return correct / total


# -----------------------------
# ✅ Diversity Metric (Distinct-2)
# -----------------------------
def distinct_n(texts, n=2):
    tokens = []
    for t in texts:
        tokens.extend(t.split())

    if len(tokens) < n:
        return 0

    ngrams = set(zip(*[tokens[i:] for i in range(n)]))
    return len(ngrams) / len(tokens)


# -----------------------------
# ✅ Full Evaluation Pipeline
# -----------------------------
def evaluate_model(model, tokenizer, trainer, test_dataset):

    print("\n📊 Running Evaluation...\n")

    # ✅ 1. Loss
    eval_results = trainer.evaluate()
    eval_loss = eval_results["eval_loss"]

    # ✅ 2. Perplexity
    perplexity = math.exp(eval_loss)

    # ✅ 3. Generate Predictions
    preds, refs = generate_predictions(model, tokenizer, test_dataset)

    # ✅ 4. ROUGE
    rouge_scores = rouge.compute(predictions=preds, references=refs)

    # ✅ 5. BLEU
    bleu_score = bleu.compute(predictions=preds, references=[[r] for r in refs])

    # ✅ 6. Exact Match
    exact = exact_match(preds, refs)

    # ✅ 7. Token Accuracy
    token_acc = token_accuracy(model, test_dataset)

    # ✅ 8. Diversity
    diversity = distinct_n(preds)

    # ✅ Print Results
    print("✅ Evaluation Results")
    print("--------------------------------")
    print(f"Loss           : {eval_loss:.4f}")
    print(f"Perplexity     : {perplexity:.4f}")
    print(f"Exact Match    : {exact:.4f}")
    print(f"Token Accuracy : {token_acc:.4f}")
    print(f"ROUGE          : {rouge_scores}")
    print(f"BLEU           : {bleu_score}")
    print(f"Diversity      : {diversity:.4f}")

    return {
        "loss": eval_loss,
        "perplexity": perplexity,
        "exact_match": exact,
        "token_accuracy": token_acc,
        "rouge": rouge_scores,
        "bleu": bleu_score,
        "diversity": diversity
    }


# -----------------------------
# ✅ Run Evaluation
# -----------------------------
results = evaluate_model(
    model,
    tokenizer,
    trainer,
    test_dataset
)
````
