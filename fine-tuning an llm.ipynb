# 📘 Introduction
**Fine-Tuning**: Fine-tuning is the process of adapting a pre-trained LLM to a specific task using task-specific data.

👉 In simple terms: LLM training can be understood based on what the model learns, how it learns, and how efficiently it is trained.

---

## 📘 Types of LLM Training

LLM training techniques can be categorized based on how the model is trained, the learning signal used, and how model parameters are updated.

These categories represent different dimensions of how LLMs are trained, and they are often combined in practice.


### 1. **Supervised Training** (learn from labeled data)
- Supervised Fine-Tuning (SFT)
- Instruction Tuning
- Domain Adaptation



### 2. **Feedback-Based Training** (learn from preferences)
- Reinforcement Learning (RLHF)
- Direct Preference Optimization (DPO)



### 3. **Parameter-Efficient Training** (update small number of parameters)
- LoRA / QLoRA
- Adapters
- Prefix Tuning
- Prompt Tuning



### 4. **Large-Scale Training** (update all model parameters)
- Pretraining (Self-Supervised Learning)
- Full Fine-Tuning



### 5. **Extended Training Strategies** (improve or extend capabilities)
- Multi-Task Learning
- Continual Learning
- Contrastive Learning

---

## 🧭 How These Training Types Work Together

A typical LLM training setup combines multiple dimensions:

- **Supervised Training (SFT / Instruction)** → defines what the model learns  
- **Parameter-Efficient Training (LoRA / QLoRA)** → determines how parameters are updated efficiently    
- **Large-Scale Training** → provides the pretrained base model
- **Feedback-Based Training (RLHF / DPO)** → refines and aligns model outputs

---

## 🚀 Example Training Pipeline

The following pipeline shows how these training types are typically applied in sequence:

1. Start with a **pretrained model**  
2. Apply **Supervised Fine-Tuning (SFT)**  
3. Use **Instruction Tuning** for better task understanding  
4. Apply **LoRA / QLoRA** for efficient training  
5. (Optional) Apply **RLHF / DPO** for alignment and quality

# 📘 Supervised Fine-Tuning (SFT)

Supervised Fine-Tuning (SFT) is a training approach where a pre-trained model is trained further using labeled input-output pairs.

👉 The model learns to generate the correct output given an input.

👉 Internally, the model learns to predict the next token step-by-step, assigning higher probability to correct sequences of tokens.

👉 A pretrained model knows language, but SFT trains it to give correct, task-specific and consistent responses instead of generic one

<br>

---

**Example 1**: Question Answering
  * *`Input`*: What is machine learning?
  * *`Output`*: Machine learning is a method where systems learn patterns from data.

**Example 2**: Instruction Following
  * *`Instruction`*: Explain gravity in simple terms
  * *`Response`*: Gravity is a force that pulls objects toward each other.

**Example 3**: Summarization
  * *`Input`*: [Long paragraph]
  * *`Output`*: Short summary of the paragraph

  → keeps base model frozen and updates small adapter layers

````python

from datasets import load_dataset

# Load Dataset (we take 300 samples ) so we can split into:train )
dataset = load_dataset("tatsu-lab/alpaca", split="train[:300]")
print(dataset)

# we can split into: train 80%, test 10%, val 10%
train_test = dataset.train_test_split(test_size=0.2, seed=42)

train_dataset = train_test["train"]
temp_dataset = train_test["test"]

# Split Temp → Validation + Test
val_test = temp_dataset.train_test_split(test_size=0.5, seed=42)

val_dataset = val_test["train"]
test_dataset = val_test["test"]

# Check Sizes
print()
print("Train size:", len(train_dataset))
print("Validation size:", len(val_dataset))
print("Test size:", len(test_dataset))
````

## HuggingFace SFT (Full Fine-Tuning)

### Step 1: Data Pre-processing
* Collect input-output pairs
* Clean and structure data
* Format into prompt style

````python
def format_example(example):
    return {
        "text": f"Instruction: {example['instruction']}\nResponse: {example['output']}"
    }

train_dataset = train_dataset.map(format_example)
val_dataset   = val_dataset.map(format_example)
test_dataset  = test_dataset.map(format_example)
````

### Step 2: Tokenization

Convert text into tokens
Example:

"Hello world" → [15496, 995]

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token


def tokenize(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

train_dataset = train_dataset.map(tokenize, batched=True)
val_dataset = val_dataset.map(tokenize, batched=True)
test_dataset = test_dataset.map(tokenize, batched=True)


train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask"])
val_dataset.set_format(type="torch", columns=["input_ids", "attention_mask"])
````

### Step 3: Load model

````python

from transformers import AutoModelForCausalLM

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = AutoModelForCausalLM.from_pretrained("gpt2").to(device)
````

### ### Step 4: Training Loop

````python
import sys
import torch
import matplotlib.pyplot as plt

def progress_bar(step, total, epoch, total_epochs, train_loss, val_loss=0.0):
    percent = step / total
    bar_length = 20
    filled = int(bar_length * percent)
    bar = "#" * filled + "." * (bar_length - filled)

    sys.stdout.write(
        f"\rEpoch {epoch+1}/{total_epochs} "
        f"[{bar}] {percent*100:.0f}% "
        f"| Train Loss: {train_loss:.4f} "
        f"| Val Loss: {val_loss:.4f}"
    )
    sys.stdout.flush()


def plot_losses(train_losses, val_losses):

    epochs = range(1, len(train_losses) + 1)

    plt.figure(figsize=(8,5))

    plt.plot(epochs, train_losses, label="Train Loss", marker="o")
    plt.plot(epochs, val_losses, label="Validation Loss", marker="o")

    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")

    plt.legend()
    plt.grid(True)

    plt.show()


def train_model(model, train_loader, val_loader, optimizer, device, epochs):

    train_losses = []
    val_losses = []

    for epoch in range(epochs):

        # -------- TRAIN --------
        model.train()
        train_loss = 0

        for step, batch in enumerate(train_loader):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=input_ids
            )

            loss = outputs.loss
            train_loss += loss.item()

            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

            # ✅ Use external progress bar
            progress_bar(
                step + 1,
                len(train_loader),
                epoch,
                epochs,
                train_loss / (step + 1)
            )

        avg_train_loss = train_loss / len(train_loader)

        # -------- VALIDATION --------
        model.eval()
        val_loss = 0

        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch["input_ids"].to(device)
                attention_mask = batch["attention_mask"].to(device)

                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    labels=input_ids
                )

                val_loss += outputs.loss.item()

        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        print(
            f"\n✅ Epoch {epoch+1} Done | "
            f"Train Loss: {avg_train_loss:.4f} | "
            f"Val Loss: {avg_val_loss:.4f}"
        )

    return train_losses, val_losses
````

````python

from torch.optim import AdamW
from torch.utils.data import DataLoader


batch_size =8
epochs=3
lr = 5e-5

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)

optimizer = AdamW(model.parameters(), lr=lr)

train_losses, val_losses = train_model(
    model,
    train_loader,
    val_loader,
    optimizer,
    device,
    epochs=epoch
)

plot_losses(train_losses, val_losses)
````

### Step 5: Evaluation

After training a model, we must evaluate:

👉 How well the model performs on unseen data

👉 Whether the model generates correct and useful outputs

<br>

---

🎯 **Types of Evaluation Metrics**
1. **Loss**
    - Measures prediction error  
    - Lower = better

2. **Perplexity**
    - Measures model confidence
    - Converts loss into an interpretable score
    - Measures how “confused” the model is
    - Formula: `Perplexity = exp(loss)`  
    - Lower = better

3. **ROUGE**
    - Measures overlap with reference text  
    - Used for summarization
    - Only measure text similarity
    - Do NOT fully capture meaning or reasoning
    - Higher = better  

4. **BLEU**
    - Measures n-gram overlap between prediction and reference
    - Used for translation-like tasks  
    - Only measure text similarity
    - Do NOT fully capture meaning or reasoning
    - Higher = better  

5. **Exact Match**
    - Checks if prediction exactly matches the reference
    - Good for short answers
    - Very strict metric

6. **Qualitative Evaluation**
    - Human inspection of model outputs
    - Check for: Relevance, Correctness, Structure, Instruction-following

🧠 **Mental Model**
```
    Loss → training quality
    Perplexity → confidence
    ROUGE/BLEU → similarity
    Exact Match → strict correctness
    Output → ✅ real performance
```

````python
import torch
from evaluate import load

# -----------------------------
# ✅ Load Metrics
# -----------------------------
rouge = load("rouge")
bleu = load("bleu")

# -----------------------------
# ✅ Metric Functions
# -----------------------------
def compute_perplexity(loss):
    return torch.exp(torch.tensor(loss)).item()

def compute_rouge(preds, refs):
    return rouge.compute(predictions=preds, references=refs)

def compute_bleu(preds, refs):
    return bleu.compute(predictions=preds, references=[[r] for r in refs])

def compute_exact_match(preds, refs):
    return sum(p.strip() == r.strip() for p, r in zip(preds, refs)) / len(preds)


# -----------------------------
# ✅ Generate Response
# -----------------------------
def generate_response(model, tokenizer, prompt, device):
    model.eval()

    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    output = model.generate(
        inputs["input_ids"],
        max_length=100,
        do_sample=True,
        temperature=0.7
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)


# -----------------------------
# ✅ Collect Predictions
# -----------------------------
def get_predictions(model, tokenizer, dataset, device, num_samples=50):
    preds = []
    refs = []

    for i in range(min(num_samples, len(dataset))):
        sample = dataset[i]

        prompt = f"Instruction: {sample['instruction']}\nResponse:"
        reference = sample["output"]

        prediction = generate_response(model, tokenizer, prompt, device)

        preds.append(prediction)
        refs.append(reference)

    return preds, refs


# -----------------------------
# ✅ Full Evaluation Function
# -----------------------------
def evaluate_all(model, tokenizer, test_loader, test_dataset, device):

    # -------- LOSS --------
    model.eval()
    test_loss = 0

    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=input_ids
            )

            test_loss += outputs.loss.item()

    avg_test_loss = test_loss / len(test_loader)

    # -------- PERPLEXITY --------
    ppl = compute_perplexity(avg_test_loss)

    # -------- GENERATE OUTPUT --------
    preds, refs = get_predictions(model, tokenizer, test_dataset, device)

    # -------- METRICS --------
    rouge_score = compute_rouge(preds, refs)
    bleu_score = compute_bleu(preds, refs)
    exact_match = compute_exact_match(preds, refs)

    # -------- PRINT --------
    print("\n✅ Evaluation Results")
    print("----------------------------")
    print(f"Test Loss: {avg_test_loss:.4f}")
    print(f"Perplexity: {ppl:.4f}")
    print(f"Exact Match: {exact_match:.4f}")
    print(f"ROUGE: {rouge_score}")
    print(f"BLEU: {bleu_score}")

    return preds, refs


# -----------------------------
# ✅ Show Sample Outputs
# -----------------------------
def show_samples(preds, refs, n=5):
    for i in range(min(n, len(preds))):
        print("\n----------------------------")
        print("Reference :", refs[i])
        print("Prediction:", preds[i])
````

````python
preds, refs = evaluate_all(
    model,
    tokenizer,
    test_loader,
    test_dataset,
    device
)

show_samples(preds, refs)
````


 How SFT Works (Step-by-Step)

Step 1: Prepare Dataset

Collect input-output pairs
Clean and structure data
Format into prompt style


Step 2: Tokenization

Convert text into tokens
Example:

"Hello world" → [15496, 995]


Step 3: Model Prediction

Model takes input
Generates output tokens


Step 4: Loss Calculation

Compare predicted output with true output
Compute loss (error)

👉 Typically:
Cross-Entropy Loss


Step 5: Backpropagation

Adjust weights to reduce error


Step 6: Iterate Over Dataset

Repeat for multiple epochs


🔧 Training Objective
SFT uses:
👉 Next-token prediction
The model learns:

“Given previous tokens, what is the next correct token?”


Example:
Input: "The sky is"
Target: "blue"


🧠 Important Insight
Even though SFT looks like:
Input → Output

Internally it is:
Token → Next token prediction

👉 This is very important to understan
